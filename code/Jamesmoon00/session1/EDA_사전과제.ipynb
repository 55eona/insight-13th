# EDA 란?

##### Exploratory Data Analysis (탐색적 데이터 분석)

*** CDA Comfirmatory Data Analysis (확증적 데이터 분석, 가설 검증을 위한 데이터 분석)와 다르게 

**데이터 덩어리를 분석하여 인사이트를 도출하는 과정임!!**

분석을 할 때에는, 탐색적인 태도를 끊임없이 가지고 데이터의 확인해야한다.

### EDA 대상

일변량 / 다변량

> 일변량 : 파악하려는 변수가 1개, 해당 변수에 존재하는 패턴을 분석하는 것에 집중

> 다변량 : 파악하려는 변수가 2개 이상, 변수들끼리의 관계를 파악하기, (변수들끼리의 관계보다 각 변수 자체를 먼저 파악하는 것이 좋음)

### EDA 종류

> 시각화 : 데이터에 대해 시각적으로 나타내어 분석하기, 시각화에 특화된 파이썬 라이브러리들이 많음 (matplotlib, seaborn, plotly가 대표적임)

> 비시각화 : 데이터들에 대해 요약적으로 분석한 결과를 볼 수 있음. 한눈에 파악은 못하지만, **정확한 수치를 통해 각 변수나 변수들 간의 관계를 파악할 수 있음**

### EDA 유형

EDA 대상 2가지(일변량/다변량), EDA 종류 2가지(시각화/비시각화)에 대해 총 4가지로 나눠서 유형을 분류함

> **Uni-Non Graphic(일변량 비시각화) :** 주어진 데이터의 Distribution을 확인하는 것이 주목적.

> **Uni-Graphic(일변량 시각화)** : 주어진 데이터를 _전체적으로_ 살펴보는 것이 주목적.

> **Multi-Non Graphic(다변량 비시각화)** : 주어진 둘 이상의 변수 간 관계를 확인하는 것이 주목적.

> **Multi-Graphic(다변량 시각화)** : 주어진 둘 이상의 변수 간 관계를 _전체적으로_ 살펴보는 것이 주목적.

# 데이터 확인

### 데이터 Load

판다스에서는 여러 가지 파일 형식을 지원하기 때문에, 판다스 내장 메소드를 활용하여 해당 파일을 읽고 판다스 자료형으로 변환할 수 있다.

> (ex. CSV, EXCEL, JSON, SQL)

또한, 파일이 아닌 외부 프로그램의 api 형태로 데이터를 제공한다면, 해당 프로그램에서 안내해주는 api 사용법을 활용하여 load하면 된다.

> `import kagglehub` -> kaggle 플랫폼에서 제공하는 데이터셋을 판다스 라이브러리 kagglehub를 통해 load하기 위해 해당 라이브러리를 임포트하여 load한다면, 데이터셋을 쉽게 다운 받을 수 있음

다만, 데이터셋의 파일 형태가 다양할 수도 있고, 수치형인지 범주형인지에 따라 파일을 구성하고 있는 텍스트 형태가 매우 다양할 것이다. 파일을 읽고 쓰는데, 인코딩/디코딩 하는 형태가 다르기 때문에 이를 유의해야함!!

> 한국어 데이터셋을 활용하는 일이 많이 있기 때문에, 한국어를 지원하는 인코딩 방식이 무엇인지 잘 파악하여 데이터셋을 정상적으로 로드할 수 있게 해야함!!

### 데이터셋 확인

판다스 DataFrame 형식으로 데이터를 불러오게 된다면 우선적으로 raw data를 한번 스캔하는 것이 좋다!!

>데이터 수가 지나치게 많다면, `dataset.head()`와 같은 메소드를 통해 일부분을 보고 파악해도 된다. 

>(그러나 해당 데이터가 전체 데이터를 무조건 대표하는 건 아니라는 것을 인지하고 있기)

#### 데이터의 종류

- ##### 범주형 : 여러 개의 범주로 나눌 수 있는 데이터 (연산이 안됨!!)

> 명목형 : 순서가 없는 범주형 데이터 (색상 - 빨, 주, 노, 초, 파)

> 순서형 : 순서가 있는 범주형 데이터 (학년 - 초 -> 중 -> 고)

- ##### 수치형 : 수치로 표현이 되고 연산이 가능한 데이터

> 이산형 : 특정 값을 가져야하는 데이터 (나이 - 23살, 24살, 25살, 26살)

> 연속형 : 무한으로 수치를 나눌 수 있는 데이터 (몸무게 - 53kg, 54kg, 54.32344kg, 54.323449012123kg)

### 데이터 분포 확인

판다스 메소드 `info()`를 활용해서 데이터 유형이나 결측치(null)를 파악하거나 전체 데이터의 양도 파악할 수 있다.

판다스 메소드 `describe()`를 통해서는 수치형 데이터들의 양, 평균, 표준편차 등등 통계적인 정보를 대략적으로 파악할 수 있다.

시각화 라이브러리를 통해, 하나의 변수/2개 이상의 변수에 대해 어떤 분포를 띄고 있는지, 변수 간의 관계가 어떤지를 파악할 수 있다.

# 전처리

> 데이터를 분석할 때 되도록 전체 데이터를 대표할 수 있게 분석을 해야한다!!

> 그렇기 때문에, 대표성을 무너뜨리는 데이터는 없애는 것이 좋거나 쓸모없는 데이터가 있다면 과감히 없애는 것도 정확한 데이터 분석에 도움이 될 것이다

##### 결측치 처리



위 그림에서 NaN 값은 수치적으로 없는 값이라고 생각하면 된다. 숫자 0도 아니라 그냥 없는 값인 것이다.

이러한 데이터를 **결측치**라고 하며, 해당 데이터가 수치 분석에 대해 영향이 미미하다면 처리하지 않아도 되지만, 

통계 분석에 있어서 문제가 되는 경우라면 0으로 대체하거나 과도하게 결측치가 많은 경우는 아예 분석에서 제외하기도 한다.

##### 이상치 처리



그림에서 노란색 bar에 해당하는 값이 **이상치**이다.

전체 데이터들이 이루고 있는 분포와 다른 경우가 이상치라고 생각하면 된다. 이는 통계 분석에 있어서 대표성을 떨어뜨릴 수도 있기 때문에, 적절한 값으로 대체하여 분석하는 것이 정확한 데이터 분석에 도움이 된다.

##### Feature Engineering (변수 가공하기)




데이터 분석을 하여, ML/DL 모델에 해당 데이터를 학습을 시켜 데이터 예측을 하는 경우,

Feature : 독립 변수, 원인

target : 종속 변수, 결과

이렇게 변수를 나눠서 생각을 해야한다. Feature Engineering을 하는 이유는 

1. 모델 성능을 개선하기 위해 더 나은 형태로 데이터를 가공을 하여 학습시키는 목적도 있고, 

2. 변수 가공을 통해 데이터 자체에 대해 더 유용하게 분석할 수 있기 때문이다.
