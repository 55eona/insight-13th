## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 선형 회귀 </span>



데이터를 가장 잘 설명하는 <U>회귀선을 찾고 이를 통해 예측하는 과정</U>

회귀 모델을 사용하기 위해서는 대표적으로 다음과 같은 가정이 요구된다.

>선형성 : 종속변수(y)와 독립변수(x)의 관계가 선형적이어야한다. (독립 변수가 종속변수한테 선형적인 영향을 줘야한다.)

>독립성 : 각각의 관측치는 서로 독립적이어야한다. (관측치끼리 영향이 미치면 안된다.)

>등분산성 : 종속 변수와 독립 변수들의 관계가 일정하다.<BR>
(독립 변수들의 수준에서 오차의 분산이 일정해야한다. 그게 아니라면, 일부 데이터에는 다른 데이터에 비해 상대적으로 큰 오차를 갖게 되는데, 여기서 신뢰성이 저하된다.)

>정규성 : 오차가 정규 분포를 따라야한다.

- ## 단순 선형 회귀 : 하나의 독립 변수인 경우
- ## 다중 선형 회귀 : 두 개 이상의 독립 변수인 경우

우리의 목적은 <U>데이터를 가장 잘 설명하는 직선 y = ax + b</U> 를 찾는 것이다!!

##### 최소제곱법



잔차의 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식

-> 회귀선과 실 데이터의 차이를 최소화하기 위한 목적

*** 잔차 : 실제값과 예측값의 차이

##### 경사하강법

1. 목적 함수<BR>
    최적화(Optimization)를 위해 오차를 최소화하는 것이 목적인 함수<BR>
    최적화 대상을 수학적으로 표현함 (ex. 성적을 최대화, 실수를 최소화)

2. 손실 함수<BR>
    개별 데이터 샘플에 대한 오차를 측정하는 함수<BR>
    데이터 포인트 하나에 대해 모델이 얼마나 틀렸는지 평가 (예측값과 실제값의 차이를 계산)

3. 비용 함수<BR>
    모든 데이터 포인트를 손실 함수 적용하여 계산하여 평균/합계를 계산한 값<BR>
    <U>모델 학습에서, 비용 함수를 최소화하는 것이 목표!!</U>



독립 변수가 2개 이상인 경우 -> 다중선형회귀분석

최소제곱법과 경사하강법 모두를 사용할 수 있음!!! 

##### 다중공선성

다중 선형 회귀 과정에서 확인해야하는 중요한 사항이다!!

독립 변수들 간의 상관관계가 높은 경우, <U>어떤 독립 변수가 종속 변수에 얼마나 많은 영향을 미치는지 정확하게 구분하기 어려워</U>지는 문제가 발생한다.

-> 이 문제로 회귀 분석 정확도가 낮아진다...



이런 히트맵을 통해 <U>독립 변수들 간의 상관계수를 측정</U>해서 다중공선성을 파악할 수 있음!!

추가적으로, VIF 지수 (분산 팽창 인수)<BR>

이는 회귀 모델의 결정계수 R square를 사용하여 계산하는데, <U>VIF 지수가 높으면 다중공선성이 존재한다고 판단</U>한다.

다중공선성을 대처하기 위해서는

1. 변수 선택법 : 상관 관계가 높은 변수들 중에서 feature로 사용할 변수를 선택하는 것이다.

2. 변수 변환 : 변수들끼리 더하거나 빼도 문제가 없으면, 해당 처리를 하여 새로운 변수를 생성하는 것이다.

3. 규제 선형 모델 : 릿지, 라쏘, 엘라스틱넷 등의 방법으로 모델의 복잡도를 줄이는 방법이다.

4. 주성분분석(PCA) : 데이터의 차원을 축소하여 고차원 데이터의 중요 정보는 최대한 보존하면서 데이터의 복잡성을 줄이는 기법이다.<BR>
    다중공선성이 존재하는 변수들끼리 묶고 더 적은 수의 대표 축으로 바꾸는 것이 핵심이다.<BR>
    (ex. `['Size','Number of Rooms', 'Number of Bathrooms'] -> ['Size Feature']` 이 경우는 PCA를 통해 3개의 차원을 1개 차원으로 줄인 것)

##### 규제선형모델

모델이 학습데이터에 대하여 과적합 상태가 되지 않도록 규제를 더하고자 만든 모델

-> 선형 회귀 모델에 관해서는 특성에 곱해지는 기울기의 크기를 조정하는 것이다!

규제 선형 모델의 종류

1. L2 : 릿지 회귀, W(회귀 계수)의 제곱에 대해 페널티 부여<BR>
    w의 제곱에 대한 페널티이기 때문에, 크기가 큰 회귀계수일수록 더 강한 페널티를 받음
    <U>비중이 높은 변수는 줄어들게 되고, 모든 변수는 유지함</U>

2. L1 : 라쏘 회귀, W의 절댓값에 대해 페널티 부여<BR>
    w의 절댓값을 페널티로 주기 때문에, 회귀계수의 크기와 상관없이 페널티를 받음<BR>
    <U>비중이 낮은 변수는 사라지게 되고, 중요 변수들만 남음!!</U>

3. L2 + L1 : 엘라스틱넷 회귀<BR>
    L2 & L1을 함께 사용해서, <U>모든 변수가 균형있게 줄고, 변수가 제거되는 효과가 난다.</U>

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 모델 평가 방법 </span>

##### 성능 평가 지표

- MSE(Mean Squared Error, 평균 제곱 오차) :<BR>
    <U>모델의 예측값과 실제 관측값의 오차 제곱의 평균</U>, 예측 모델의 성능 평가 지표들 중 하나

- MAE(Mean Absolute Error, 평균 절대 오차) :<BR>
    <U>모델의 예측값과 실제 관측값의 오차 절댓값의 평균</U>, 예측 모델의 성능 평가 지표들 중 하나    

##### 변수 유의성 평가

- T 검정

        독립 변수의 회귀 계수가 유의미한지 검정하기 위해 사용됨

        해석 방법<BR>
        | 회귀 계수가 0 -> 해당 독립 변수는 의미가 없음<BR>
        | 독립 변수의 p value가 작을수록, 종속 변수에 더 큰 영향을 줆<BR>
        | t값이 크면, 귀무 가설을 기각할 가능성이 높다<BR>
