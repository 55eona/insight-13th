### 기술 통계 vs 추론 통계에 관해서...

우리가 여태까지 판다스를 공부해서 이를 데이터셋에 적용하여 EDA를 진행했던 것은 기술 통계(Descriptive Statistics)에 해당한다.<BR>
표본으로 뽑은 데이터셋에 대해 다양한 방식으로 요약하기 때문에 <U>통계적으로 기술한다</U>라고 생각하면 된다.

### 왜 통계를 공부하는가?

Python, R처럼 통계 분석에 특화된 프로그램들을 활용해서 통계 분석 기술을 익히는 것은 중요하다.<BR>
그러나, 이러한 기술 통계 분야뿐만 아니라 분석하고자 하는 데이터셋의 모집단에 대해 정보를 추출하거나 추측을 하여 데이터의 인사이트를 찾는 추론 통계 분야도 매우 중요하다.<BR>
그래서 ML/DL을 다루려면 통계 공부는 필수적이라고 생각해야한다.



### 데이터의 척도

> 관찰 / 조사를 통해 얻어낸 수치적 or 비수치적 자료

#### 양적 자료 (수치적 자료)

- 구간 척도

        연속적 수이며, 대소 관계를 표현할 수 있음 (ex. 온도 : xx도)

- 비율 척도

        연속적인 수이며, 절대적인 원점이 존재, 사칙연산도 가능! (ex. 몸무게 : 0g -> 무게가 없다는 뜻으로 간주, xxg)

#### 질적 자료 (비수치적 자료)

- 명목 척도

                단순 범주만 표현하는 척도, 숫자는 분류의 기능만 수행함 (ex. 색상 : 빨, 주, 노, 초)

- 서열 척도

                순위 / 서열이 있는 자료, 사칙연산 수행은 안됨 (ex. 순위 : 1위, 2위, 3위)


## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 기술 통계 </span>

데이터의 주요한 특징을 요약하는 방식을 다루는 통계 분야.<BR>
<U>평균, 중앙값, 표준편차</U>등의 개념들이 이에 해당함

```python
list = [1,2,3,4,2,1,2,4,5000]

import numpy as np

import pandas as pd

plist = pd.Series(list)

print(plist.mode())
```

- 평균 : 데이터의 총합을 개수로 나눈 값이다. 그러나 이상치에 민감함!! (ex. `[1,2,3,4,2,1,2,4,5000]`인 경우, 평균은 __557.67__ )
- 중앙값 : 데이터들을 크기를 기준으로 정렬했을 때, 중앙에 위치한 값. 이상치에 덜 민감함 (ex. 예시 상황에서 중앙값은 __2__ )
- 최빈값 : 데이터셋에서 가장 많이 나타난 값, _이산형 데이터에 자주 사용됨_ (ex. 예시 상황에서 최빈값은 __2__ )



왜도 : 데이터의 분포를 파악하기 위해 활용됨, 중심 경향성을 보여주지는 않는다!

    왜도 > 0
        평균과 중앙값이 최빈값보다 <U>크다</U>

    왜도 = 0
        평균과 중앙값이 최빈값과 <U>같다</U>

    왜도 < 0
        평균과 중앙값이 최빈값보다 <U>작다</U>

분산 : 데이터들이 평균을 기준으로 얼마나 퍼져있는지 나타내는 통계 지표

표준편차 : 분산에 제곱근을 취한 값

범위 : 데이터의 최솟값과 최댓값의 차이, 데이터의 변동 범위를 나타냄

사분위수 (quartile) : 표본을 4개의 동일한 구간으로 나눈 값 (<U>이상치를 파악하기 유용하다!</U>)





변동계수 : 데이터의 상대적 변동성을 표현하는 값

### 추론통계란?

수집한 데이터들을 토대로 미래에 대한 데이터를 예측 / 추론하는 통계 분야 (통계적 가설 검정, 신뢰구간 등이 대표적인 개념) <BR>
쉽게 생각해서, 표본집단으로 모집단을 추론하는 과정이라고 생각하기

### 추정

#### 확률분포

어떠한 확률 변수가 특정값을 갖게 될 확률을 나타내는 함수!!<BR>
확률 변수가 <U>이산적인 값이면, 이산확률분포에 해당</U>하고,<BR> 확률변수가 <U>연속적인 값이면, 연속확률분포에 해당</U>한다.

- 이산확률분포 (베르누이분포, 이항분포, 기하분포, 음이항분포, 초기하분포, 포아송분포)

- 연속확률분포 (정규분포, T분포/F분포, 감마분포, 지수분포, 카이제곱분포, 균등분포)


##### 대표적인 확률 분포 모형들



정규분포 : 표준편차에 따른 분포를 나타낸 모형



T분포 : 자유도에 따른 분포를 나타낸 모형 (자유도 : 독립적으로 변할 수 있는 데이터들이 얼마나 많은가?를 나타냄)



카이제곱분포 : 자유도에 따른 분포를 나타낸 모형, 독립적인 표준 정규 분포에 대한 변수들의 제곱 합이다. 그렇기 때문에, 독립성 검정에 자주 쓰임

### 검정

가설 검정은 표본집단을 통해 모집단의 특정 값을 예측할 때, 해당 가설의 합당성 여부를 판단하는 과정이다.

#### 가설 검정 절차

가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단


1. 기설 설정
    귀무 가설 = 기각하고자 하는 명제 / 대립 가설 = 주장하고자 하는 명제

2. 유의 수준 설정
    결과 판단에 있어서 허용할 수 있는 오류의 최대치, <U>얼마나 신뢰할 수 있는 지에 대한 판단 기준</U>

3. 검정통계량 산출
    결과 판단에 사용되는 지표, 데이터의 분포/특징에 따라 Z값, T값, F값 등 올바른 통계량을 선택하여 판단함

4. 가설 기각/채택 판단
    유의 확률 : 검정통계량에서 유도할 수 있는 지표이며, <U>유의 확률이 유의 수준보다 작으면, 귀무가설은 기각이 되고, 크다면 기각하지 못함</U>


#### 데이터 형태에 따른 검정 방법들

> 수치형 - 수치형<BR>

(피어슨) 상관 분석 : 두 변수 간 피어슨 상관 계수가 얼마나 유의미한지 검정

> 범주형 - 수치형

T-test, F-test : 변수가 2개 (1:1)이면 T-test, 변수가 3개 이상이면 F-test

> 범주형 - 범주형

Chi2-test : 두 변수 간의 독립성을 검정하는 지표, 데이터에 대한 기대빈도와 관측빈도의 차이가 크다면 -> 둘이 독립적이지 않다는 확률이 높다는 것!

> 수치형 - 범주형

대표적인 지표가 따로 있지 않아서, _범주형을 target, 수치형을 feature로 두고 모델을 활용하여 feature-target에 대한 유의성 검정하여 판단하는 방법을 사용_
