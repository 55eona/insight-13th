### 회귀 분석 프로세스

1. 사전 검증<BR>
    전처리 과정
2. 모델 생성 및 학습<BR>
    모델 생성, 하이퍼 파라미터 설정, 모델 학습
3. 모델 평가<BR>
    검증 가능한 가설 검토
4. 모델 성능 개선<BR>
    불필요 변수 제거, 비선형 모델 활용

$$
y=w_0x_0+w_1x_1+\cdot\cdot\cdot+w_mx_m=\sum_{i=0}^mw_ix_i=w^Tx
$$

회귀 식은 feature(x)들의 가중 평균이다!!

다중 선형 회귀 기본 가정

1. 선형성<BR>
종속 변수와 독립 변수의 관계가 선형적이어야함
2. 독립성<BR>
각각의 feature들이 서로 선형독립적이어야함 (다중공선성의 반대)
3. 오차항의 평균은 0이다.(0에 가까워진다.)<BR>
오차항 = 실제값과 예측값의 차이
4. 등분산성<BR>
오차항과 분산이 일정해야함
5. 오차항은 자기상관되어 있지 않다.<BR>
오차항의 공분산이 항상 0이어야 함
6. 정규성<BR>
오차항이 정규분포를 따라야 함


> 2, 3, 6은 회귀 분석 전에는 판별하지 못함...!!

### 회귀 분석 평가 방법

#### <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 시각화 </span>

선형 회귀 분석 결과를 시각화하여 데이터들의 대표성을 띄는지 파악하기!<BR>
그러나 정확하게 분석을 할 수 없기 때문에, 통계 분석이랑 병행해야한다!!

#### <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 통계 지표 </span>

R2 Score, MSE, RMSE, MAE 등을 활용하여 회귀 모델의 예측 결과를 평가할 수 있다.

F 검정이랑 T 검정을 통해서 모델의 유의성을 검정할 수도 있는데,<BR>


F 검정은 <U>회귀 모델 자체가 유의미한지</U>를 판별한다.<BR>
> p value가 0.05 이하인 경우, 회귀 모델이 유의미하다고 검정한다!

T 검정은 <U>특정 독립 변수가 유의미한지</U>를 판별한다.<BR>
> p value가 0.05 이하인 경우, 해당 변수가 유의미하다고 검정한다!

#### <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 모델 성능 평가 </span>


회귀 모델 성능 평가에는 $R^2$(결정계수)를 자주 사용한다.

회귀 분석에서 모델이 설명하는 데이터의 총 변동(평균과의 차이)중에서 설명된 비율을 나타냄!!

$\bar{Y}$: 평균 / $\hat{Y}$: 예측 값 / $Y$: 실제 값

노란색 선은 평균을 의미한다.

또한, 초록색 표시는 SST를 나타내고, 보라색 표시는 SSR, 빨간색 표시는 SSE를 나타낸다.

해당 값들을 활용하여 $R^2$값을 아래의 그림처럼 도출할 수 있다.

$R^2$ 값은 모델이 데이터를 얼마나 잘 설명하는지 측정하는 지표이며, 0부터 1사이의 값을 가진다.

값이 1에 가까워질수록, SSR과 SST의 값이 비슷해지며 SSE는 0에 가까워진다. // 이렇게 되면 모델이 데이터를 잘 설명한다고 판별할 수 있음!!

값이 0에 가까워지면, 모델이 데이터를 잘 설명하지 못한다고 판별...

$R^2$ 값이 높으면 무조건 좋은 의미인가?

-> 아님. X

why?

-> 독립 변수의 개수가 증가할수록 $R^2$값도 증가하기 때문에, 유의해야한다!

## <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> 비선형 회귀 </span>

#### 다항식 회귀모델

$Y=a+bX+cX^2+dX^3+\cdot\cdot\cdot$ 의 형태로, 독립 변수를 거듭 제곱 꼴로 변형하여, 데이터의 곡선적(비선형적)인 특성을 모델링함

선형 모델에 비해 회귀선을 더 fit하도록 그릴 수 있다.

그러나, 너무 많은 feature를 사용하게 되면, 과적합 상태가 될 수도 있다.


> 직선적인 관계보다, 곡선을 그리는 데이터 패턴일 때 더 적합함

> 성장률, 감소율 등이 변화 속도가 시간에 따라 다른 경우 더 적합함

#### 지수 회귀모델

종속 변수가 지수적으로 변화하는 관계를 모델링 할 때 사용

$Y$에 로그 함수를 적용하고, 이에 대해 선형 회귀를 적용함<BR>
그 이후에 모델이 예측한 값에 지수 함수를 적용하여 로그 변환 전의 형태로 다시 되돌리면 됨

> 시간에 따라 값이 지수 함수 형태로 변화할 때 적합함 (기술 발전, 투자 수익률 증가 등등)

#### 로그 회귀모델

독립 변수와 종속 변수의 관계가 로그함수로 더 표현이 잘 될 때 사용

독립변수 $X$에 로그함수를 적용함<BR>
그리고 선형 회귀를 적용하면 됨 ($Y$값에 별도의 처리를 할 필요가 없다)

> 데이터가 초기에 빠르게 변화하고, 점차 변화율이 줄어들 때 적합함 (인구 성장, 감염병 확산 등등)

#### 스플라인 회귀모델

데이터를 구간 별로 나눈 다음, 각 구간 별로 각기 다른 선형 / 비선형 모델을 적용하여 예측하는 방식

전체 데이터의 범위를 여러 구간으로 나누고, 각 구간에 대한 회귀 모델을 적용함<BR>
<U>구간의 경계에서 연속성을 유지해야함!!!</U>

> 데이터 패턴이 여러 구간에서 다르게 나타날 때 적합함
> 계절에 따라 판매량이 변화하는 경우 등에 적합함
