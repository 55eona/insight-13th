다중선형회귀의 기본 가정

선형회귀는 굉장히 강력한 툴이지만, 모든 경우에 적용 가능한 것은 아니고, 다음과 같은 가정을 만족해야함


1. 선형성 - 종속변수와 설명변수 간의 관계가 선형적이어야 합니다.
2. 독립성 - 각각의 설명변수가 서로 선형독립적이어야 합니다
    - 선형독립적이지 않은 경우 다중공산성이 있어 문제를 해결해줘야합니다.
3. 오차항의 평균은 0 - 오차항은 백색잡음, 즉 특정 패턴 없이 랜덤하게 변하는 신호를 의미하는데, 어느 한 쪽으로도 치우처지지 않아야합니다
4. 등분산성 - 오차항의 분산이 일정해야함 / 그렇지 않을 경우 모델이 모든 구간에서 일정한 예측력을 가지지 않음
5. 오차항이 자기상관 x - 오차항의 공분산이 0이어야함
6. 정규성 - 오차항이 정규분포를 따른다는 가정(위배되어도 결과에 큰 영향을 주진 않지만, p값의 정확도와 유의미한 해석을 위해 필요)

모든 가정을 회귀분석 전에 검증해야 하진 않고, 선형성, 독립성, 자기상관무 3개만 체크하고, 나머지는 분석 후에 해도 됨




회귀 분석 평가 방법

1. 시각화
회귀선이 데이터 전반을 잘 요약하는지 큰 그림으로 확인 가능, 하지만 어떤 회귀선이 더 좋은지 비교하기엔 어려움 -> 통계지표 활용

2. 통계지표

모델의 유의성 검정 : F-검정, T-검정
- f검정은 / 전체 모델에 대해 / 회귀모델이 유의미한지 검정 / 가설로 모든 회귀계수가 0이다 하고 검정
- t검정은 / 개별 변수에 대해 / 특정 독립 변수가 유의미한지 검정하는 것 / 가설로 해당 변수의 회귀계수가 0이다 하고 검정

모델의 성능 평가 : R-squared, Adjusted R-squared, AIC, BIC
- R-squard(결정계수)- 데이터의 총 변동에서 설명된 비율(평균보다 모델 사용시 얼마나 더 잘 예측하는지), 1에 가까울수록 잘 설명


- Adjusted R-squard- 독립변수의 개수 증가에 덜 민감하도록 조정한 지표


- AIC, BIC - 둘 다 정보기준이라 불리며, 값이 작을수록 좋음, AIC가 BIC보다 여유있는 기준




<h2>2. 비선형 회귀</h2>

데이터셋의 분포가 꼭 선형이라는 법은 없지요! 이런 경우에는 선형적이지 않은 식을 작성하면 됩니다

1. 다항식 회귀모델 - 기존의 선형모델 y= a+ bx 에서 x^2, x^3등 거듭제곱 항을 추가해 모델링
    - 선형 모델에 비해 회귀선을 핏하게 잘 그릴 수 있으나, 너무 많이 넣으면 오버피팅 생김
    - 적합한 상황 : 관계가 직선적이지 않고 곡선을 그리는 데이터에서 유용, 성장률 감소율 등 시간에 따라 변화하는 속도가 다른 경우 


2. 지수 회귀모델 - 종속변수가 지수적으로 변화할 때 사용
    - 사용 방법 : 지수함수형태의 데이터는 y=ae^bx의 수식을 따르는데, 양변에 로그를 취해서 ln y = ln a + bx 형태로 만들고, 여기에 선형회귀를 적용한 후, 원래의 데이터 형태로 복구하기 위해 예측값에 지수함수(e^x)를 적용
    - 적합한 상황 : 값이 시간에 따라 지수적으로 증가하거나 감소하는 경우 적합(기술 발전, 투자 수익률 등)



3. 로그 회귀모델 - 종속변수와 독립변수간의 관계가 로그함수 형태일 때
    - 사용 방법 : 로그함수 형태의 데이터는 y= a+blog(X)형태인데, 선형회귀식에서 X대신 log(X)를 사용한 것!, 순서는 독립변수 X에 로그함수를 적용한 후, 여기에 선형회귀를 적용함, 지수함수와는 달리, 예측값에 로그를 취하거나 한 게 아니라, 그대로 사용 가능함
    - 적합한 상황 : 데이터가 초기에 빠르게 증가하고 점차 증가율이 줄어드는 패턴을 보일때 좋음(인구성장, 감염병 확산 등)


4. 스플라인 회귀 - 데이터를 구간별로 나눠서 각 구간별로 선형 또는 비선형 함수 적용해 예측
    - 구간 경계에서 연속성을 유지하는것이 중요함
    - 적합한 상황 : 데이터 패턴이 여러 구간에서 서로 다르게 나타날 때 유용, 계절따라 판매량 변화하는 소매데이터 분석에 유용
