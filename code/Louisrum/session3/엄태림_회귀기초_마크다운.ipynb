<h2> 머신러닝과 모델링

<h3>
    1. 머신러닝
</h3>

데이터에서 패턴을 찾아내고, 이를 바탕으로 새로운 데이터를 예측하거나 결정을 내릴 수 있도록 하는 기술,

+ 비지도학습 : 정답이 없는 데이터로 학습하는 방식
+ 지도학습 : 정답이 있는 데이터로 학습하는 방식


<h3>모델링




데이터를 훈련데이터와 테스트 데이터로 분할
+ 훈련데이터(train_set): 모델을 학습시킬 때 쓰는 데이터
+ 테스트데이터(test_set): 학습된 모델을 검증하기 위해 사용하는 데이터 - unseen data

데이터를 쪼개는 이유 : 훈련데이터로 학습시킨 모델이 새로운 데이터, 즉 테스트 데이터에 대해 얼마나 예측을 잘 수행하는지 공정하게 평가하기 위해 데이터를 분리함, 분리하지 않고 훈련시킨 데이터를 평가에 쓰면 모델이 데이터를 단순 암기한 결과일 수 있기 때문에 실제 성능을 제대로 평가할 수 없음


<h2> 회귀(Regression)와 모델링

<h3> 회귀란?
    





데이터에서 패턴을 찾아내어 미래 값을 예측하는 것

통계나 머신러닝의 주요 개념

목표변수가 연속적일 때 사용 (불연속적일 땐 분류를 이용) - 목표변수의 연속성 불연속성 따라 사용 툴 다름

<h3> 회귀의 종류

선형회귀, 비선형 회귀, 릿지 회귀, 라쏘 회귀, 다항회귀 등 다양한 기법이 존재

<h3> 선형회귀

+ 데이터를 가장 잘 설명하는 하나의 선, 회귀선을 찾아 예측하는 과정
+ 우리의 관심사이자, 예측하고 싶은 것을 종속변수 y, 예측을 위해 사용하는 변수를 독립변수 x로 설정
+ 함수의 모형이 꼭 직선일 필요는 없음, 단지 계수들이 선형적으로 결함하면 선형회귀로 포함됨


선형회귀 기본 가정(회귀 심화세션에서 자세히 학습)
1. 선형성 - 종속변수와 독립변수의 관계가 선형적이어야함
2. 독립성 - 각각의 관측치가 서로 독립적이어야
3. 등분산성 - 모든 독립변수들의 수준에서 오차항의 분산이 일정해야
4. 정규성 - 오차항이 정규분포를 따라야



선형회귀의 종류 : 단순 선형 회귀(하나의 독립변수로) / 다중 선형 회귀(두 개 이상의 독립변수로)

(1) 단순선형회귀분석
- 하나의 독립변수만 존재
- 데이터를 가장 잘 설명하는 직선 y = ax+b를 찾는 과정
- a는 회귀계수(기울기), b는 y절편

최소제곱법(최소자승법, ols) - 최적의 회귀선을 찾는 방식!!!
- 잔차(실과 예측값의 차이)의 제곱합이 최소가 되는 지점으로 회귀선 구하는 방식
- 회귀선과 실제 데이터와의 차이를 최소화하기 위한 시도
- 이걸로 했을 때 복잡하면 경사하강법 사용(함수가 닫힌 상태, 미분계수 구하기 어려운 상태, 데이터 양이 너무 많은 경우)

경사하강법 - 최소제곱법을 통해 구한 함수가 복잡해서 최솟값을 구하기 어려운 경우 사용
1. 함수가 닫힌 상태인 경우
2. 함수가 너무 복잡해 미분 계수를 구하기 어려운 경우
3. gradient descent를 구현하는게 미분 계수를 구하는 것 보다 더 쉬운 경우
4. 데이터 양이 너무 많은 경우 효율적으로 계산하기 위해

목적함수 : 최적화를 위해 오차를 최소화하는 것이 목적인 함수, 최적화하려는 대상을 수학적으로 표현한 함수, 머신러닝에서 모델의 예측이 얼마나 잘 맞는지 평가하는 기준이 필요

손실함수 : 개별 데이터 샘플에 대한 오차를 측정하는 함수, 예측값과 실제 값 차이를 계산

비용함수 : 전체 데이터셋에서 평균적인 손실을 측정하는 함수, 모든 데이터포인트에 대해 계산한 후 평균 또는 합산한 값, 이걸 최소화하는게 보통 목표(이러면 비용함수 = 목적함수로 봐도 무방)

(1) 손실함수 :잔차를 제곱한 뒤 합하는 형태



독립변수 W와 b에 대해, 종속변수 E가 변화하는 양상을 보기 위해, W와 E, b와E의 관계를 이차함수로 표현

만약 현재 값이 최솟값의 왼쪽이면 기울기가 음수이므로, "현재값 - 미분값" 구조에 의해 현재 값보다 커짐, 그래프의 오른쪽으로 이동, 만약 현재 값이 최솟값의 오른쪽이면 기울기가 양수이므로, "현재값 - 미분값" 구조에 의해 현재 값보다 커짐, 그래프의 왼쪽으로 이동, 최솟값에 도달하면 변화하지 않고 멈춤


(2)학습률 : 얼만큼 이동을 할 것인가(마치 보폭)

학습률이 너무 작으면 학습 속도가 너무 느리고, 너무 크면 모델이 수렴하지 못하고 발산함 
> 작은 값부터 시도해가면서 속도 보면서 조정해준다!(자동으 찾는 방법도 있음)

(3)Local Minima 문제 (전체 모델에서의 최솟값이 아닌, 일부 구역의 최솟값인데도 빠져나오지 못하고 갖히는 경우)

모든 비용함수가 매끈한 곡선인 것이 아니라 발생

(4)해결법 - 모멘텀

경사하강법은 현재 기울기만을 기준으로 하기 때문에, 이전 기울기는 잊음
-> 이전의 기울기와 이동하던 방향을 고려하도록 관성 부여!!
-> 지역최소값 탈출 가능

<h3> 다중선형회귀분석</h3>


독립변수가 2개 이상인 경우의 선형회귀 -> 회귀선이 평면으로 표현됨

단순과 마찬가지로 최소제곱법을 사용할 수 있고, 관측치와 평면간의 차이가 잔차가 된다

<b>다중공산성

여러 변수들이 존재하는데, 이 각 변수들이 서로 강한 상관관계를 가지면, 각각의 독립변수가 종속변수에 어떤 영향을 미치는지 정확하게 구분하기 어렵기 때문에, 회귀모델의 정확도가 낮아짐

<b>다중공산성 확인
1. 상관계수 확인
2. VIF(분산 팽창 인수) - 결정계수를 활용해 체크 (회귀 심화에서 더 다룸)

<b>다중공산성 대처 방법
1. 변수 제거(변수선택법) : 독립변수로 사용할 변수를 선택하는 방법
2. 변수 변환 : 변수들을 더하거나 뺴서 새로운 변수 생성(당연히 이렇게 해도 문제가 없는 경우에 사용)
3. 규제 선형 모델 활용 : 릿지, 라쏘, 엘라스틱넥 등으로 모델의 복잡도를 줄이는 방법 사용
4. PCA(주성분 분석) : 데이터의 차원을 축소하는데 사용, 별로 영향을 주지 않는 차원을 없앰, 상관이쓴ㄴ 변수들을 묶어 더 적은 수의 대표축으로 바꾸는 것(심리학에서도 자주 사용한 기법)
    1. 데이터 전처리(정규화)- 평균을 0으로 조정하고, 표준화함
    2. 주성분 찾기 - 공분산 행렬 계산, 고유벡터, 고유값 추출
    3. 차원 축소 - 주성분 축에 데이터 투영
    4. 복원 가능(넣어도 되고 안넣어도 되고)



<h3> 규제 선형 모델

+ 모델이 과적합되지 않도록 하는 방식, 계수의 크기를 조정하는 방식으로 사용
+ 특히 더 복잡한 다중회귀에서 주의해야함
+ RSS를 최소화하는 방식을 쓰면 회귀계수가 쉽게 커지게 되어 과적합 발생



비용함수는 최적화를 하는 방법과, 과적합 방지법이 균형을 유지해야함
-> 오류는 최대한 줄이면서 특정 변수에 집착하지 않도록 균형을 맞춤


규제선형모델의 종류
1. L2규제 : 획귀계수의 제곱에 대해 페널티를 부여하는 방식, 릿지(Ridge)회귀
2. L1규제 : 회귀계수의 절댓값에 대해 페널티를 부여하는 방식, 라쏘(Lasso)회귀
3. L1규제 +L2규제 결합 : 엘라스틱넷 회귀






<h3> 릿지 회귀(L2규제)</h3>

기둥 크기에 따라 다른 힘으로 누르는 망치, 회귀계수의 제곱을 패널티로 주기 때문에, 큰 회귀계수일수록 더 강한 패널티 받음, 전체적으로 작아지지만, 완전히 사라지지는 않아서 변수가 제거되지 않음


<h3> 라쏘 회귀(L1규제)</h3>

모든 기둥을 같은 힘으로 누르는 망치, 일부 변수가 완전히 사라질 수 있다 -> 이걸 이용해서 비중 낮은 변수를 날리고, 중요한 변수만 남기기도 가능

<h3> 엘라스틱넷(L1 + L2규제)</h3>

망치 두 개로 내려쳐서 안정적인 예측과 희소성을 동시에 달성 가능

<h3> 정리</h3>



+ 릿지회귀 : 예측변수가 많을 때, 다중공산성 존재할 때 사용
+ 라쏘회귀 : 예측변수가 많은데, 일부만 중요할 때, 모델의 해석을 간단하게 유지하고자 할 때
+ 엘라스틱넷 : 위 둘의 기능이 모두 필요할 때


<h3> 모델 평가방법

<b> (1) 성능 평가 지표
1. 평균 제곱 오차(MSE)
    + 모델의 예측값과 실제 관측값 사이의 오차 제곱의 평균을 의미

2. 평균 절대 오차(MAE)
    + 모델의 예측값과 실제 관측값 사이의 절대값 오차의 평균을 의미

둘의 차이 : MSE는 이상치에 민감, MAE는 전반적인 예측 성능 유지 가능

1. t검정
