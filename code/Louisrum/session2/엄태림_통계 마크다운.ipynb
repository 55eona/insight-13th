통계

통계의 종류
기술통계-표본을 요약하여 표현
추론통계-추정과 검정을 통해 표본에서 얻은 정보를 바탕으로 모집단을 추정

통계 자료의 유형
1.질적자료
명목척도 - 이름 등 순서가 없는 범주 데이터
서열척도 - 성적 등급 등 순서가 있는 범주 데이터, 등급 간 차이가 존재하지만, 정확히 얼마나 차이나는지는 알 수 없음
2.양적자료
등간척도 - 동일한 간격으로 차이가 발생하는 데이터, 이를 통해 얼마나 크고 작은지 구분 가능 // 하지만 0이 절대적인 숫자 0이 아닌, 편의상 구분한 원점이라고 볼 수 있음
비율척도 - 등간척도와 비슷하나, 절대적인 0 이 존재함

기술통계:
데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야, 평균 중앙값, 표준편차 등의 개념이 포함되고, 이를 통해 데이터의 기본 특성을 파악 가능
수치기술통계(중심경향성, 산포도)와 시각화기술통계(EDA에서 배운 내용)

중심경향성 - 평균, 중앙값, 최빈값
평균 - 전체 데이터의 총합을 데이터의 개수로 나눔, 가장 대표적인 중심경향치, 하지만 아웃라이어에 민감함
중앙값 - 전체 데이터를 나열했을 때 가운데 위치한 값, 평균과 달리 이상치에 덜 민감함, 또한 양적자료 뿐 만 아니라, 일부 서열 척도에도 사용 가능(성적 등급의 중앙값 등)
최빈값 - 가장 자주 나타나는 값, 주로 이산형 데이터에서 사용, 분포를 이해하는데 도움

왜도 : 기울어진 정도, 데이터의 분포의 특징을 설명할 때 사용
최빈값<중앙값<평균 : Right-skewed
최빈값=중앙값=평균 : 좌우대칭(ex 정규분포)
평균<중앙값<최빈값 : Left-skewed

산포도 :데이터가 퍼져있는 정도
1. 분산 : 데이터가 평균으로부터 떨어진 거리의 제곱의 평균, 편차의 제곱의 합의 평균, 단위도 제곱됨
2. 표준편차 : 분산의 제곱근, 평균으로부터 떨어진 거리의 평균이라고 이해하면 편함, 단위가 데이터와 동일
3. 범위 : 데이터의 최댓값과 최솟값의 차이, 전체 범위
4. 사분위수 : 데이터의 표본을 4개의 동일한 크기로 나눈 값
5. 사분위간 범위(IQR) : Q3에서 Q1을 제외한 범위, 데이터의 중앙 50%가 퍼진 범위, 이상치 탐지용
6. 변동계수 : 표준편차를 평균으로 나누고 100을 곱해 상대적인 변동성을 확인

추론통계 : 일부 표본에서 수집한 데이터를 바탕으로 전체 모집단을 추정하고, 일반화하는 과정

확률분포 : 확률변수가 특정한 값을 가질 확률을 나타내는 함수
확률변수가 셀 수 있는 특정 값을 가지면 이산확률분포, 연속적으로 이어지는 값을 가지면 연속확률분포임

정규분포 : 종형, 평균을 기준으로 대칭, 표준편차가 크면 데이터가 평평해지고, 표준편차가 작으면 뾰족해짐, 가장 대표적
(중심극한정리에 의해 n>=30이면 정규분포에 근사함)

T분포 : 정규분포와 형태가 유사하나, 꼬리가 더 두꺼움, 자유도에 따라 모양이 결정되고, 자유도가 커지면 표준정규분포에 수렴, 추정에서 사용, 표본의 크기가 작을 때(N = 30미만)사용
자유도: n-1, 왜 n-1이냐면 마지막 하나는 나머지가 다 정해지고 나면 값이 확정되기 때문, 자유도가 너무 커지면 오버피팅과 같은 문제가 발생

카이제곱 분포 : 0에서 시작해 오른쪽으로 긴 꼬리를 가진 비대칭 분포, 자유도가 증가하면 점점 정규분포에 가까워짐, 독립성 검정 등에 사용

추론통계 - 검정

가설검정 절차
1. 가설 설정 : 귀무가설 H0과 대립가설 H1을 설정, H1이 우리가 원하는 명제고, H0은 이거의 반대, H0을 날려 우리의 주장이 그럴듯함을 보이는 과정
2. 유의수준 설정 : 알파값이고, 보통 0.05를 사용, 결과를 판단할 때, 허용할 수 있는 오류의 최대치, 제1종오류의 범위로 실제로 참인데 기각할 확률을 의미함
3. 검정통계량 산출 : 가설 검정의 결과를 판단하는 데 사용하는 값, Z, T, F값 등 사용
4. 가설 기각 채택 판단 : 유의확률(p-value)를 계산하고, 이를 유의수준과 비교해 가설 채택/가각 여부를 결정
5. 결과 해석 및 보고 : 결과를 보고하며 해석함

데이터의 자료형에 따른 검정 방법

피어슨 상관계수 분석 : 두 변수가 선형적인 관계를 가지고 있는지 확인함, 30% 언더는 약함, 50%까진 중간, 50% 이상은 강한 관계임
F-test, Chi2-test 등이 있음
