EDA & 전처리 마크다운 사전과제

데이터 분석 프로세스 : 수집 - 탐색(EDA) - 전처리 - 모델링

데이터 탐색(EDA)는 수집한 데이터를 여라가지 방식으로 파악하는 과정, 
전처리는 데이터를 적절한 방식으로 손질하는 과정

EDA(Exploratory Data Analysis), 탐색적 데이터 분석 과정

1. 대상 - 일변량/다변량 : 하나의 변수를 자세히 볼 것인가, 아니면 여러 변수들이 어떤 관계를 가지는지 볼 것인가
    - 일변량(univariate) 파악하려는 변수가 하나로, 데이터 안에 존재하는 패턴을 찾는 것이 목적
    - 다변량(Muliti-variate) 여러 변수를 다뤄서, 변수 간의 관계를 보려고 함(이거 돌리기 전에 돌릴 친구들을 미리 일변량을 돌려서 확인해볼것)

3. 종류 - 시각화/비시각화 : 시각화를 통해 대략직인 형태를 파악하고, 정확한 값은 비시각화를 통해 파악
    - 시각화(Graphic) 차트 혹은 그림 등을 이용하여 데이터를 확인
    - 비시각화(Non-Graphic) Summary Statistics를 통해 데이터를 확인

4. 유형 : 대상(일변량/다변량), 종류(시각화/비시각화)에 따라 분류
    - Uni-Non Graphic(일변량 비시각화) : 주어진 데이터의 Distribution을 확인하는 것이 주목적.
          ex)빈도표, 기술통계량
    - Uni-Graphic(일변량 시각화) : 주어진 데이터를 전체적으로 살펴보는 것이 주목적.
          ex)파이차트, 막대그래프, 히스토그램, 박스플롯
    - Multi-Non Graphic(다변량 비시각화) : 주어진 둘 이상의 변수 간 관계를 확인하는 것이 주목적.
          ex)교차표, 상관계수
    - Multi-Graphic(다변량 시각화) : 주어진 둘 이상의 변수 간 관계를 전체적으로 살펴보는 것이 주목적.
          ex)모자이크플롯, 박스플롯, 평행좌표, 산점도

데이터 확인

작업공간으로 데이터를 불러오는 '읽기', 새로운 데이터를 만드는 것은 '쓰기'
데이터를 읽는 것은 절대 경로와 상대 경로 나뉨
절대 경로는 파일이 저장된 위치로 접근하는 모든 경로를 포함
상대 경로는 현재 작업중인 디렉터리 기준으로 표현되는 상대적인 경로 - 같은 파일 내에서 작업할 때 쓰면 좋지 않을까?

파일 형식 - csv와 excel <- 둘 다 pandas를 통해 dataframe 형식으로 읽을 수 있음
csv는 쉼표로 데이터를 구분하는 텍스트파일, 크기가 작고 압축이 용이하고 대중적으로 사용됨
excel은 행과 열이 데이터프레임으로 일대일 대응 + 특정 시트만을 불러오거나, 여러 시트를 불러올 수 있음

데이터를 불러왔으면, 파일을 파악해야함
1) 데이터프레임을 head(n)을 통해 상위 n개의 데이터를 확인
2) 데이터 변수를 확인, 어떤 정보를 가지는 변수인지, 변수의 유형을 확인, 변수의 자료형을 확인
       - 변수의 정의를 확인 : 해당 변수가 어떤 의미를 담고 있는지(예/고객이름, 판매량 등)
       - 변수 유형 확인 : 범주형/수치형, 범주형은 순서 없이 분류된 명목형 자료와 순서 관계가 존재하는 순서형 자료가, 수치형은 정수 단위로 떨어지는 이산형 데이터와 연속적인 값을 갖는 연속형 자료로 나뉨
       ** 정수면 이산형, 소수면 연속형인 것이 아니라, 구간을 잘랐을 때 셀 수 있으면 이산형임, 숫자가 의미하는 바가 특정 포인트를 의미하는지, 아니면 연속적인 범위로서 의미를 갖는 것인지 파악할 것
3) 단변수 분석을 통해 원시 데이터의 분포를 확인하고, 전처리 아이디어를 얻는다

파일을 간단히 파악한 후에는, 데이터를 사용하기 쉽게 적절한 방식으로 손질하는 과정을 거쳐야 함

데이터를 보고 누락이나 중복 데이터 등 오류를 날리고, 분석 목적에 맞게 변형하는 과정을 거쳐야 품질 좋은 데이터를 얻는다

전처리를 위한 기술에는 데이터 정제, 변환, 필터링, 통합 등이 있다.

결측값이 있는 상태로 데이터분석 모델을 만들면 변수 간의 관계가 왜곡될 수 있어 모델의 정확성이 떨어지고, 시각적 표현에도 문제가 생길 수 있음 -> 결측치를 제거해줘야함



결측치 제거를 위해서는 NaN, ?, 0 등 결측값을 확인하고, 0의 경우는 진짜 0의 값을 가진 것인지, 결측되어 0으로 표기된 것인지 확인
+ 사용하는 function : info(), value_counts(dropna=False), isnull(), notnull(), replace() 등

결측값 처리

1. 이터가 있는 행 또는 열을 삭제 : dropna(axis=0/1, how='any'/'all', subset=[col1, col2, ...], inplace=True/False)
+ 삭제 기준, 삭제 조건, 기준열, 원본 데이터 변경 여부를 작성해 반영함
+ 무작위로 발생한 결측치를 삭제해야지, 다른걸 날리면 모델이 왜곡될 수 있음

2. 결측값을 다른 값으로 대체 : 주로 평균이나 최빈값을 활용, 일괄적으로 같은 값으로 대체하는 일괄 대체와 유사한 유형에 따라 다른 값으로 대체하는 유사유형 대체가 있음 fillna()를 사용

이상치 처리 

관측된 데이터의 범위에서 많이 벗어난 값을 의미하는 이상치를 처리해야 함

1. describe()를 통해 통계를 바탕으로 이상치의 존재를 확인
2. BoxPlot을 사용해 시각화를 해서 확인
   + for i in df.describe().columns:
    df[[i]].boxplot()
    plt.show()
3. Z-score을 통해 특정 기준값을 넘어가는 데이터를 이상치로 확인
4. Tukey Fences를 바탕으로 이상치를 확인

이상치를 위의 방법을 통해 확인한 후에는, 이상치를 제거해야 함
1. 휴먼 에러에 의해 발생한 이상치는 전삭제
2. 절대적인 관측치가 작으면, 단순 삭제로 손실되는 데이터 양이 문제가 될 수 있음, 이럴 땐 이상치를 평균 등으로 대체하거나 이상값을 예측해서 대체해서 적을 수 있음
3. 이상값이 자연발생하면, 변수화를 통해 모델에 포함시켜, 모델이 설명하고자 하는 현상을 더 잘 표현시킬 수 있음
4. 자연 발생한 이상값을 분리해 모델을 만드는 리샘플링을 할 수 있음 - 이상값을 포함한 모델과 제외한 모델을 모두 만들고, 각각의 모델에 대해 설명을 다는 것이오

   + 원핫인코딩은 이진벡터(0,1)로 변환함
2. 구간화

3. 변환
   + 기존의 피처를 다른 피처로 변환해 변수를 추가, 분석가의 데이터 특성에 대한 이해도에 기반해 다양한 데이터 생성 가능
