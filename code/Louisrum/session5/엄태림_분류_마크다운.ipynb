<h1>분류 기초</h1>

<h2>분류란?</h2>

<h3>(1) 머신러닝: 지도 학습과 비지도 학습</h3>

머신러닝 : 인공지능의 한 분야로, 컴퓨터가 스스로 학습할 수 있도록 도와주는 알고리즘이나 기술을 개발하는 분야

컴퓨터는 1) 알고리즘을 이용하여 데이터를 분석하고, 2) 분석 결과를 스스로 학습한 후, 3) 이를 기반으로 판단이나 예측을 실시함

지도 학습은 문제와 정답을 다 아는 자료를 제공해 학습시키는 것 -> 모델을 통해 예측하는 값을 이미 만들어둔 정답과 같아지도록 기계를 학습시키는 것(회귀, 분류 등)

비지도 학습은 답을 알려주지 않거나, 없는 자료를 주고 학습시키는 방법 -> 데이터 속의 패턴 또는 유사도를 학습(군집화)

강화 학습은 보상을 통해 상을 최대화하고 벌을 최소화하는 방식 -> 시행착오를 반복해 정답을 찾는 과정(알파고)

> 지도 학습과 비지도 학습 예시 : (과제) 사진을 보고 기린과 고양이를 구별하기
> - (지도학습) 자 이게 기린이야 외워! 반복하기.
> - (비지도학습) 오 얘는 목이 기네? 오 얘는 얼룩무늬가 있구나! 오 얘는 뿔도 있다 이러면서 데이터의 패턴을 흡수해 기린을 찾아냄

비지도학습의 특징
- 지도학습과는 다르게 정답이 있지 않음, 주어진 데이터로만 학습
- 인간의 개입 없이 스스로 데이터를 학습하여 그 속의 패턴 또는 각 데이터 간의 유사도를 학습해 구조나 패턴을 찾음
- 군집화(비슷한 데이 묶어 큰 단위 만들기), 밀도 추정(데이터 분포를 예측), 차원 축소(데이터 차원을 간추림)

군집화 : anomaly detection(변칙감지), 고객세분화(잠재고객을 여러 segment로 나눔) / 얼마나 묶을지 지정해 제어 가능



<h3>(2)지도 학습: 회귀와 분류</h3>

문제랑 정답을 같이 주고 학습시켜 예측값이 정답과 최대한 같아지도록 학습시킴

구체적인 과정
1. labeled data(독립변수와 종속변수가 모두 존재하는 데이터)를 training set과 test set으로 나눔
2. training set을 활용해 독립변수를 통해 추정한 예측값이 종속변수와 같아지도록 모델을 훈련시킴(기계의 예측이 의도한 정답이 되도록 지도함)

반복 학습을 통해 오류를 줄여가면서 정답에 가까워지는 방법

회귀와 분류 (회귀는 계속 했으니 생략)

분류(Classification) : 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤에 새롭게 관측한 데이터에 대한 레이블을 판별하는 것

회귀와 분류의 차이 : 회귀는 연속형 변수를 예측하기 위해, 분류는 범주형 변수를 예측하기 위해 사용


분류에 쓰이는 대표적인 머신러닝 알고리즘
1. 로지스틱 회귀(Logistic Regression) : 독립변수와 종속변수의 선형 관계성에 기반해 분류
2. 결정 트리(Decision Tree) : 데이터 균일도에 따른 규칙 기반의 분류
3. 서포트 벡터 머신(SVM) : 개별 클래스 간의 최대 마진을 효과적으로 찾아 분류
4. 최소 근접(Nearest Neighbor) 알고리즘 : 근접 거리 기준으로 분류


<h3>(3)이진 분류와 다중 분류</h3>

- 이진 분류 : 예측하고자 하는 변수가 어떤 기준에 대하여 참 또는 거짓의 값만 가짐
- 다중 분류 : 예측하고자 하는 변수가 가질 수 있는 값이 3개 이상



<h2>분류 모델</h2>
<h3>(1)로지스틱 회귀</h3>

- 이진 분류 문제를 풀 때 사용, 특정 클래스에 속할 확률을 추정하는 것이 목표
- 독립 변수의 선형 조합(선형 회귀)에 로지스틱 함수를 적용하여 출력값을 0과 1 사이로 변환

일반적인 선형 회귀는 연속적인 값을 예측하는 데 사용되지만, 로지스틱 회귀는 이진 분류 문제를 해결하는 데 적합합니다.
- 사실 선형 회귀는 분류 작업에 유리하지 않다(예측 값은 0과 1 사이여야 하는데, 선형회귀로 하면 이보다 훨씬 큰 수를 가지게 됨)
-> 출력이 0과 1 사이의 값을 가지면서 s자 형태를 그리는 함수를 이용 -> 시그모이드 함수 사용!

시그모이드 함수(Sigmoid Function) : 출력이 0과 1 사이의 값을 가지면서 S자 형태로 그려지는 함수



- 입력값이 커지면 1에 수렴하고, 입력값이 작아지면 0에 수렴
- 출력값이 0부터 1까지의 값을 가짐
- 출력값이 특정값 이상이면 1, 이하면 0으로 정하면 이진 분류 문제를 풀기 위해서 사용할 수 있음!

<h3>(2)결정 나무(Decision Tree)</h3>

결정 나무 : 조건에 따라 데이터를 분류하여 초종적으로 순수한 라벨의 집합으로 구성될 때까지 분류를 반복하는 모델

자료구조의 트리의 구조와 용어가 거의 일치 -> 머신러닝에서 이걸 의사결정용으로 사용하는 것



CART(Classification And Regression Tree) 알고리즘

데이터를 임계값 기준으로 두 child로 나누는 알고리즘

- 임계값 불순도(지니 계수)가 낮아지는 방향으로!

불순도란? 분류하려는 데이터 집합에서 서로 다른 클래스가 섞여 있는 정도를 의미

- 지니 계수란? : 불순도를 나타내며, 통계적 분산 정도를 정량화해서 표현한 값(0과 1 사이) -> 얼마나 많은 값이 섞여 있는지 알 수 있음, 클수록(1에 가까울수록) 안좋고, 작을수록(0에 가까울수록) 좋음


<b>CART 알고리즘의 주요 단계</b>
1. 임계값 설정 : 임계값을 설정하고, 임계값을 기준으로 하여 데이터를 그룹으로 나눈다
2. 불순도 감소 알고리즘 : 불순도가 낮은 쪽으로 가지를 형성해나감(하지만, 당장 더 낮은 쪽으로 가기만 하기 때문에, 가장 효율적인 방향은 아닐 수 있다)

<b>실제 학습 시 고려해야 할 것들: 모수 설정, 차이점 시각화, Prunning</b>
1. Parameter 설정
2. 시각화 : 시각화를 통해 과적합 여부 등 확인해 분류가 잘 이뤄졌는지 확인
3. Prunning(가지치기): 불필요한 노드 지우기 -> 노드가 너무 많으면 과적합될 수 있음 -> 노드 제거로 일반화 성능을 높일 수 있음

<h3>(3)서포트 벡터 머신(SVM)</h3>

다양한 경계선 중 최적의 라인을 찾아내는 알고리즘

-> 명확하게 분류할 수 있는 데이터 집단에서 뛰어난 성능을 보이며, 고차원 공간(다수의 feature)에서도 효과적으로 사용 가능


SVM의 구성
- support vectoer : 구분하는 선과 가장 가까운 포인트
- Decision Boundaty(결정 경계) : 집단을 구분하는 선
- Margin : 선과 각 점의 거리

최적의 선을 찾는 방법
- 결정 경계는 데이터로부터 가장 멀리 떨어져 있는게 좋다! SVM은 거리가 가장 큰 경우를 선택함으로서 최적의 선을 찾는 것

<h3>(4)KNN(K-Mearest Neighbor)</h3>

KNN의 가정 : 비슷한 특성을 가진 데이터끼리 서로 가까이 있는다는 점을 이용한 분류 알고리즘 - 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘 : k개의 이웃을 설정하고, 가까운 k개의 데이터

KNN 계산 순서
1. 데이터 준비 - KNN은 사전 학습이 필요 없음, 특징 벡터와 레이블로 이뤄진 데이터를 준비
2. K값 설정 - 원하는 K값을 설정, K는 가장 가까운 이웃의 개수를 나타내며, 보통 홀수개로 설정(짝수로 설정하면 동점이 발생하는 문제가 생길 수 있음)
3. 거리 계산 - 새로운 데이터가 주어지면, 이 값과 기존 모든 데이터 간의 거리를 계산합니다.(유클리드 거리, 맨해튼 거리 등을 사용)
4. 가장 가까운 K개의 이웃 선택 - 계산된 거리 중에서 가장 작은 거리 값을 가진 K개의 데이터를 선택
5. 분류하기 - K개의 이웃 중에서 가장 많이 등장하는 클래스가 예측 결과가 됩니다

KNN의 중요한 특징 <b>어떠한 학습이 필요하지 않다!</b> 별도의 모델 없이 데이터만을 이용해 즉석에서 분류를 하는 것

- 장점 : 훈련이 필요치 않음, 정보 손실 없음
- 단점 : 쿼리를 처리하는 데에 시간이 오래 소사전학습을 한 모델을 사용하는 것이 아니라, 요청이 들어오면 그때그때 계산을 통해 가장 가까운 점을 구하는 것이기 때문
    - Not Robust : 이상치에 큰 영향을 받음(각 데이터와의 거리를 기반으로 하는 분류 모델이기 때문!)


<h3>(참고)앙상블</h3>
여러 분류 모델을 결함해 단독으로 했을 때 보다 더 좋은 성능을 내는 기법, 여러 모델을 학습시키고, 각 모델의 예측 결과를 종합헤서 최종 예측

종류
1. 보팅(voting) : 다른 알고리즘의 모델을 병렬로 사용
2. 배깅(bagging) : 동일 알고리즘의 모델을 병렬로 사용
3. 부스팅(boasting) : 동일 알고리즘의 모델을 직렬로 사용





<h2>분류 평가 지표</h2>
<h3>(1)혼동 행렬</h3>

분류 모델의 예측 결과를 정확한 예측과 잘못된 예측으로 구분하여 나타낸 표


총 4가지 경우의 수가 있음
- True Positive : 실제로 참, 결과도 참
- True Negative : 실제로 거짓, 결과도 거짓
- False Positive : 실제는 거짓, 결과는 참(위양성)
- False Negative : 실제는 참, 결과는 거짓(위음성)

TP, FP, FN, TN

T_ : 결과를 맞췄다
F_ : 결과를 틀렸다
_P : 참으로 예측했다
_N : 거짓으로 예측했다



이걸 바탕으로 아래와 같은 평가지표를 도출할 수 있음(전부 0에서 1 사이의 값)
- 정확도 : 모든 가능한 예측 중 참인 비율
- 정밀도 : 참이라고 예측한 경우 실제 참의 비율
- 재현도 : 실제로 참인 경우 중 참으로 예측하는 비율
  
Accuracy(정확도) : 모델이 입력된 데이터에 대해 얼마나 정확하게 예측하는지를 나타냄
- 전체 중에 결과를 정확히 예측한 비율(T로 시작): (TP+TN)/(TP+TN+FP+FN)
- 단점 : 정답 레이블의 비율이 불균형하면 모델의 정확도를 신뢰할 수 없다!

Precision(정밀도) : TP/TP+FP, FP의 값이 어느정도인지를 확인(위양성의 정도를 확인)
- 예측을 참이라고 한 경우 중 실제로도 참인 비율

Recall(재현도) : TP/TP+FN, 참을 거짓으로 판단한 정도
- 실제 참 중에서 예측으로 참을 맞춘 비율

정밀도와 재현도는 하나가 줄면 하나가 늘음

재현도를 높이기 위해선 Threshold(모델의 참 거짓을 가르는 분류 기준)를 낮춰야함 -> 참으로 진단하는 비율을 늘리기
- 암 환자의 경우 암이 아닌데 암으로 진단하는 것 보다, 암인데 암이 아닌 것으로 진단하는게 더 위험함 -> 재현도 올리기

스팸 메일 분류기
- 스펨메일이 아닌데 스펨메일로 분류하는 현상이 더 치명적일 수 있음(위양성이 더 안좋음) -> 정밀도를 높여야 함

Threshold를 잘 정하려면, 정밀도&재현도를 확인하는 그래프를 그리고, 두 그래프의 접점을 임계값으로 정하면 오류를 최소화할 수 있음


<h3>(2)F1-Score</h3>
F1-Score : 정밀도와 재현도의 조화평균, 머신러닝 모델의 성능을 평가하는 주요지표, 정밀도와 재현도의 균형을 평가함



<h3>(3)ROC/AUC Curve</h3>

1. ROC Curve : 얼마나 분류가 잘 되었는지를 보여주는 그래프
True Positive Rate(TPR) : 참인 것들 중에 참이라고 예측한 비율(=Recall)
False Positive Rate(FPR) : 거짓인 것들 중 참이라고 잘못 예측한 비율

2. AUC Curve : Roc와 x축 사이의 면적(적분값) - 모델의 성능을 숫자로 표현, 0에서 1사이의 값을 가지며, 1에 가까울수록 좋음

<h3>(4)다중 분류 평가 지표</h3>

여태까지는 모두 이진 분류 평가 지표, 다중 분류 평가 지표는 위의 방법들로 클래스별로 점수를 구한 뒤, 이를 적절히 평균을 내리는 것

다중 분류 평가 지표
- Macro average: 클래스별로 구한 평가 지표 평균, 모든 라벨이 유사한 중요도를 가질 때
- Weighted average: 클래스별로 구한 평가 지표 가중 평균 - 샘플 수로 가중평균, 샘플이 많은 라벨에 중요도
- Micro average: 모든 클래스의 예측 결과를 더하여 전체적인 성능을 평가하는 지표 - 동등한 가중치, 라벨과 상관없이 전체적인 성능을 평가



<h2>하이퍼파라미터 최적화</h2>
<h3>(1)하이퍼파라미터 최적화</h3>

하이퍼파라미터: 학습 시작 전에 사용자가 직접 설정하는 변수 -> 모델 학습 과정에 반영됨
하이퍼파라미터 최적화: 적절한 하이퍼파라미터를 찾아 모델 성능을 향상시키는 것, tuning을 거쳐 적절한 하이퍼파라미터를 찾아 모델 성능 향상을 꾀함

최적화 과정:
1. 하이퍼파라미터 탐색범위를 설정(최적값을 찾고 싶은 범위를 설정)
2. 평가 지표 계산 함수(성능 평가 함수) 정의(탐색하려는 하이퍼파라미터를 인수로 받아 평가지표 값을 계산해주는 함수 정의)
3. 1단계에서 샘플링한 하이퍼파라미터값을 사용하여 검증데이터로 정확도 평가
4. 이걸 특정 횟수 반복해 정확도결과를 보고, 범위를 좁혀나감

<h3>(2)하이퍼파라미터 최적화 방법</h3>

Grid Search: 정해진 범위에서 Hyperparameter를 모두 순회 - > 정해진 범위에서 모두 순회하며 가장 좋은 성능을 내는 값을 찾음
- 장점: 범위가 넓고 step이 작을수록 꼼꼼하게 전 범위를 탐색하니 최적해를 정확히 찾을 수 있다.
- 단점: 시간이 너무 오래 걸림
- 적용: 넓은 범위, 큰 스템을 활용해 범위를 좁힘



Random Search: 정해진 범위에서 Hyperparameter를 무작위로 탐색 -> 정해진 범위 내에서 무작위로 탐색해 가장 좋은 성능 찾음
- 장점: 범위가 넓고 step이 작을수록 꼼꼼하게 전 범위를 탐색하니 최적해를 정확히 찾을 수 있다.
- 단점: 시간이 너무 오래걸린다
- 적용: 넓은 범위, 큰 스템을 활용해 범위를 좁힘

Bayesian Optimization: 사전 정보를 바탕으로 Hyperparameter 값을 확률적으로 추정하며 탐색
- 특징: Gausain Process라는 통계학을 기반으로 만들어진 모델
- 여러 개의 하이퍼파라미터들에 대해서, Aquisition Fucntion을 적용했을 때, 가장 큰 값이 나올 확률이 높은 지점을 찾아냅니다.
