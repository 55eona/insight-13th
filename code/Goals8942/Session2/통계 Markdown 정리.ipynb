# 통계란?

 확률과 통계의 정의 
 - 확률: 데이터의 생성 과정을 통한 실제치 예측\
   [모형을 통한 결과 예측]
 - 통계: 관측치를 통한 확률 모형의 모수 추정\
    [결과를 통한 모형 추정]

통계학(Statistics)의 종류
1. 기술통계
   - 수치와 시각화를 통해 샘플을 요약하여 표현하는 방법론
   - 내가 갖고 있는 샘플 자체에 대한 관심
   - EDA는 기술통계 범주에 해당 
2. 추론통계
   - 추정과 검정을 통해 샘플로부터 모집단을 추론하는 방법론
   - 샘플의 모집단에 대한 관심
   - 회귀, 분류 등의 머신러닝은 추론통계 범주에 해당

통계를 공부하는 이유
- 통계학은 [데이터 수집 및 분석 -> 데이터의 패턴/ 관계 파악 및 결론 도출을 위한 학문이자 방법론]
- 통계학의 본질: 데이터로부터 의미를 찾는 것
- 데이터 분석의 최종 목적: 과거 데이터를 통해 미래를 예측하는 것
- 데이터 자체만으로는 의미 있는 정보 습득 한계 -> 통계학적 지식의 중요성 높음

데이터의 척도
- 데이터: 관찰과 조사를 통해 얻은 수치적/비수치적 자료
- 양적 자료(quantitative, numerical data): 숫자로 표현할 수 있는 데이터
   1. 구간(등간)척도(interval scale)
      - 연속적인 수로 수량화 가능
      - 숫자 간 간격이 동일하기 때문에, 크고 작은 정도 표현 가능
      - 절대적 원점(0)은 존재하지 않음
         - 온도 0도는 "온도가 없다"는 뜻이 아님
      - 덧셈, 뺄셈 가능 / 곱셈, 나눗셈 불가능
      - ex. 온도, 연도
   2. 비율척도(ratio scale)
      -  연속적인 수로 수량화 가능
      -  절대적 원점(0)이 존재함
      -  모든 사칙연산 가능
      -  ex. 몸무게, 인구수, 길이
- 질적 자료(qualitative, categorical data): 숫자로 표현할 수 없는 데이터 - 사칙연산 불가능
    1. 명목척도(nominal scale):
       - 단순 범주(이름)만을 나타내는 척도
       - 숫자는 분류의 기능만 할 뿐, 양적인 의미는 전혀 가지지 않음
       - ex. 성별, 색깔, 취미
    2. 서열척도(ordinal scale)
       - 순위 또는 서열이 있는 척도
       - 수량화 가능
       - 범주의 순서는 비교 가능하지만, 차이의 정도는 표현 불가
       - ex. 달리기 순위, 학년, 만족도


# (수치) 기술통계

기술통계학(Descriptive Statistics)
- 데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야
- 평균, 중앙값, 표준편차 등의 개념
- 데이터의 기본적 특성 파악 가능
- 측정/실험을 통해 수집한 자료를 정보로 전환하기 위해 요약하고 정리해 그 특성을 기술하는 절차에 관련한 개념

기술통계의 종류
1. 수치 기술 통계
   - 기술통계량으로 데이터를 요약 & 설명 (중심경향성, 산포도 등)
2. 시각화 기술 통계
   - 표와 그래프 형태로 데이터 시각화 (EDA)

중심경향성
- 데이터의 중심적 경향을 나타내는 기술통계량

  1. 평균(mean)
     - 데이터 값의 총합을 데이터의 개수로 나눈 값
     - 데이터의 중심 경향성을 나타내는 가장 대표적 지표
     - 이상치에 민감
     - series.mean()
  2. 중앙값(median)
     - 데이터 값들을 크기 순으로 나열했을 때, 가운데 위치한 값
     - 데이터의 중심 경향성을 나타내는 대안적 지표 
     - 이상치에 덜 민감 (Robust)
     - series.median()
   3. 최빈값(mode)
      - 데이터에서 가장 자주 나타나는 값
      - 이산형 데이터(Discrete Data)에서 주로 사용됨
      - 데이터의 분포를 이해하는 데 도움
      - series.mode()

* 왜도(skewness)
   - 평균, 중앙값, 최빈값을 통해 자료의 개형(왜도)를 확인 가능
   - 데이터 분포의 비대칭성을 나타내는 통계적 지표
   1. Positive Skewness (Right-skewed)
     - 오른쪽 꼬리가 왼쪽보다 길 때
     - 평균과 중앙값이 최빈값보다 큼
   2. Negative Skewness (Left-skewed)
      - 왼쪽 꼬리가 오른쪽보다 길 때
      - 평균과 중앙값이 최빈값보다 작음
    3. Normal Distribution (정규분포)
       - 대칭적인 분포
       - 평균=중앙값=최빈값
    - series.skew()

산포도
- 데이터가 퍼져있는 정도를 설명하는 기술 통계량

1. 분산(Variance)
   - 데이터의 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 통계적 측정 지표
   - 분산이 높으면, 데이터가 많이 퍼져있음
   - 각 데이터 값과 평균 간의 차이(편차)를 제곱한 값들의 평균
   - np.var(arr)
2. 표준편차(Standard Deviation)
   - 분산에 제곱근을 취한 값
   - 분산의 단위는 관측값 단위의 제곱(대상 데이터와 단위 불일치)\
     -> 제곱근을 씌워줌으로써, 단위를 일치시켜줄 수 있음
   - 단위 일치성 덕분에, 데이터를 해석하고 비교하는 데 더 직관적
   - np.std(arr)
3. 범위(Range)
   - 데이터의 최대값 - 최소값
   - 데이터 전체의 변동 범위
   - np.max(arr)-np.min(arr)
4. 사분위수 (Quartile)
   - 데이터의 표본을 4 개의 동일한 크기로 나눈 값
     1. Q1: [하위 25% / 상위 75%]에 위치한 값 (하위 사분위수)
     2. Q2: [하위 50% / 상위 50%]에 위치한 값 (중앙값)
     3. Q3: [하위 75% / 상위 25%]에 위치한 값 (상위 사분위 수)
    - np.percentile(arr,[25/50/75])
5. 사분위간 범위 (IQR)
   - Q3에서 Q1 값을 제외한 범위, 즉 데이터의 중앙 50%가 퍼져 있는 범위
   - 이상치 탐지에 활용 (Tukey's Fences)
      - 아래쪽 이상치: Q1-1.5*IQR
      - 위쪽 이상치: Q3+1.5*IQR
6. 변동계수 (Coefficient of Variation)
   - 데이터의 상대적 변동성을 나타내는 통계적 지표 (절대적 X)
   - 표준편차를 해당 데이터의 평균으로 나눈 값 [(표준편차/평균) * 100]
   - 표준편차는 단위가 존재
     - 평균이 다른 두 표본집단 간 비교/ 단위가 다른 두 데이터 비교에 부적합
   - 표준편차를 평균으로 나눠주면 단위가 사라짐
     - 집단의 퍼진 정도를 동일한 기준으로 비교 가능 (비율 비교)
   - np.std(arr)/np.mean(arr) * 100

# 추론통계 - 추정

추론통계학 (Inferential Statistics)
- 수집한 데이터 기반으로 미래/목표모집단의 정보를 추론하고 결론을 도출하는 방법을 다루는 분야
- 통계적 가설검정, 신뢰구간 등
- 모집단에서 일부를 뽑아 만든 표본 집단을 통해 모집단에 대한 정보를 추측하는 과정
- 데이터 -> 표본 -> 연구모집단 -> 목표모집단

확률분포 (Probability Distribution)
- 확률변수가 특정한 값을 가질 확률을 나타내는 함수

확률변수 (Probability Variable)
- random한 세계에서 어떤 특정 사건이 발생할 때, 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수

확률분포의 유형
 1. 이산확률분포
    - 확률 변수가 이산적인 값을 가지는 경우
    - 이산적(Discrete): 유한한, 셀 수 있는, 특정 값을 선택할 수 있는
    - 종류: 베르누이분포, 이항분포, 기하분포, 음이항분포, 초기하분포, 포아송분포
 2. 연속확률분포
     - 확률 변수가 연속적인 값을 가지는 경우
     - 연속적(Continuous): 무한한 범위를 가지는, 특정 값을 선택할 수 없는
     - 종류: 정규분포, T분포/F분포, 감마분포, 지수분포, 카이제곱분포, 균등분포

### 대표적인 분포 모형 

정규 분포(Normal Distribution)
- 형태
  - 종 모양
  - 평균을 기준으로 대칭
  - 표준편차가 클수록 곡선이 평평해짐 (데이터가 넓게 퍼짐)
- 매개변수(모수)
  - 평균(μ), 표준편차(σ)
- 특징
  - 가장 잘 알려진 확률분포
  - 많은 자연/사회 현상이 정규분포를 따르는 것으로 알려져 있음

중심극한정리
- 표본의 크기가 크면, '표본 평균의 분포'가 모집단의 분포와 상관없이 정규분포에 가까워짐
- 일반적으로 n(표본의 크기)>= 30 일 때
- 표본평균의 분포: N(m,σ²/n)

대수의 법칙
- 표본의 크기가 커질수록, 표본 평균이 모평균에 근사하는 현상
- 취합하는 표본의 수가 많을수록 통계적 정확도는 올라감
  - 데이터는 많으면 많을수록 좋다 

T분포 (T-Distributiion)
- 형태
  - 종 모양
    - 정규분포보다 꼬리가 두꺼움 (불확실성의 반영)
  - 매개변수(모수)
    - 자유도(df)
  - 특징
    - 표본으로 모평균(혹은 모평균 간의 차이)를 추정할 때 사용되는 분포
    - '표본 크기가 작을 때' 정규분포보다 더 정확한 결과를 제공함

자유도
- 얼마나 많은 정보가 독립적으로 변할 수 있는가를 나타내는 숫자
- 특정 상황에서는 일부 정보가 이미 결정됨 = 다른 정보들이 그만큼 덜 자유롭게 변할 수 있음
- "독립적인 정보의 수"
- 표본의 분산 계산에서 n이 아닌 n-1로 나눠주는 이유
  - 표본의 분산을 계산하는 과정에서 표본의 평균값이 사용되니,\
   n-1개의 값이 결정되면 1개의 값은 자동으로 결정됨\
   => 자유도 = n-1

오버피팅 (Overfitting, 과적합)
- 모델이 학습 데이터에 너무 맞추어져서 새로운 데이터에 대한 일반화 성능이 저하되는 현상
- T분포의 자유도는 표본 크기에 비례
  - 자유도가 높을수록, T분포는 정규분포에 가까워지고, 통계적 분석의 정확도가 높아질 가능성 있음
  - 하지만, 너무 많은 자유도는 오버피팅과 같은 문제 발생 가능
  - 적절한 자유도 선택이 중요함

표준정규분포 vs T분포

- 표준정규분포보다 T분포가 훨씬 중심으로부터 많이 퍼져있음 (꼬리가 두꺼움)
- 표준정규분포
   - 평균=0, 분산/표준편차=1인 정규분포
   - 모집단의 평균과 분산을 정확히 알고 있을 때 사용됨
   - 주어진 모집단에서 직접 계산되므로, 불확실성 적음
- T분포
  - 모집단의 분산을 모를 때, 표본의 분산을 사용해 모집단의 분산을 추정해 만들어짐
  - 표본의 크기가 작을수록 불확실성이 커짐
  - 이 불확실성 때문에 T분포가 표준정규분포보다 중심으로부터 많이 퍼져있음
    - 극단적 값이 나올 가능성 반영
   
- 모집단의 분산이 알려져 있지 않은 경우 표본분포를 결정할 때 
  - n(표본의 크기)이 30 미만 -> T분포 이용
  - n(표본의 크기)이 30 이상 -> 표준정규분포 이용

카이제곱분포 (Chi-Square Distribution)
- 형태
  - 0에서 시작해 오른쪽으로 긴 꼬리를 가진 비대칭 분포
  - 자유도가 증가함에 따라 중심이 오른쪽으로 이동, 모양이 대칭적으로 변화, 정규분포화
- 매개변수(모수)
  - 자유도(df)
- 특징
  - 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합
  - 주로 독립성 검정, 적합도 검정 등에 사용됨

# 추론통계 - 검정

가설 검정 
- 모집단의 실제 값이 얼마라는 주장과 관련해, 표본의 정보를 사용해 가설의 합당성 여부 판정 과정

가설 검정의 절차 
1. 가설 설정
  -  귀무가설(H0): 기각하고자 하는 명제
     - 보통 "~는 차이가 없다"라는 명제
     - 현재 통용되고 있는 가설
     - 지지될 때 위험성이 비교적 약한 가설 (무죄추정의 원칙)
  - 대립가설(H1): 내가 주장하고자 하는 명제
     - 보통 "~는 차이가 존재한다"라는 명제
     - 새롭게 주장하고 싶은 가설
     - 지지될 때 위험성이 비교적 강한 가설

2. 유의 수준 설정
- 유의수준
  - 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치
  - 주로 0.05, 5%로 설정
  - 연구자가 어떤 결정이 얼마나 신뢰할 만한지를 판단하는 기준 (귀무가설 기각의 기준)
- 1종, 2종 오류
  - 1종 오류(𝛼) : 귀무가설이 실제로 참이었는데 기각할 확률 (위양성)
  - 2종 오류(𝛽) : 귀무가설이 실제로 거짓인데 기각하지 못할 확률 (위음성)
  - 가설검정은 1종 오류를 더 치명적인 것으로 파악
    - 유의수준 = 1종 오류의 확률(𝛼)

3. 검정통계량 산출
- 검정통계량: 가설 검정의 결과를 판단하는 데 사용되는 값
  1. Z값 (Z-Score)
    - 표준 정규 분포를 사용해 얻은 값
    - 특정 데이터가 평균으로부터 얼마나 떨어져 있는지를 표준편차 단위로 나타냄
    - Z=(X−μ)/σ [X=관측값, μ=평균, σ=표준편차]
    - 큰 표본 크기(일반적으로 30 이상)에 대한 가설 검정에 주로 사용
      - 모집단의 평균에 대한 가설 검정
    - 수치 해석
      - Z값의 절댓값이 크면, 해당 데이터가 평균으로부터 멀리 떨어져 있음을 의미함
        - 비정상적 수치 or 이상치일 가능성 있음
      - 임계점 찾기
        - 양측검정 (양쪽 꼬리에서 임계점 찾기) 시의 기준
          - Z=1.96일 때 유의수준 0.05 (=신뢰수준 95%)
          - Z=2.58일 때 유의수준 0.01 (=신뢰수준 99%)
          - Z값이 임계점보다 크면, 귀무가설 기각
          - 신뢰구간 = 1-유의 수준(𝛼:1종 오류의 확률)
       
   2. T값 (T-Score)
     - T분포를 사용해 얻은 값
     - 작은 표본 크기에 대한 가설 검정(T검정)에 주로 사용됨
     - T분포는 정규분포와 유사하지만, 꼬리 부분이 더 두꺼워 작은 표본에서 발생할 수 있는 불확실성을 보다 잘 반영함
     - T=(X-μ)/(s/√n) [X=표본평균, μ=연구자가 설정하는 값, s=표본 표준편차, n=표본 크기]
     - 모집단의 평균에 대한 가설 검정, 두 표본 평균 간의 차이 검정에 주로 사용
     - 수치 해석
       - T값의 절댓값이 크면, 해당 데이터가 모집단 평균으로부터 멀리 떨어져 있음을 의미함
         - 관측된 표본 평균이 모집단 평균과 유의미한 차이가 있을 가능성이 높음
       - 임계점 찾기
         - T분포는 자유도(df)에 따라 모양이 달라지므로 자주 쓰는 자유도와 유의수준에 따른 T분포표를 사용
         - 같은 유의수준에 대해 T값은 Z값보다 커지며, 자유도가 높아질수록 Z값에 수렴
           - 자유도가 높아질수록 T분포는 표준정규분포로 수렴

   3. F값 (F-Score)

     - F분포를 사용해 얻은 값
     - 두 표본 분산 사이의 차이를 비교하는 데 주로 사용됨
     - F 분포는 두 개의 독립적인 카이제곱 분포된 변수들의 비율
     - F=분산_1 / 분산_2 [분산_1과 분산_2는 두 독립적 표본의 분산]
     - 분산 분석(ANOVA)에서 두 개 이상의 그룹 간 평균의 차이를 검정하는 데 사용됨
     - 수치 해석:
       - F값은 두 표본의 분산의 비
       - 1에 가까우면 두 표본의 분산이 서로 비슷하다는 것을 의미함
          - 두 그룹이 퍼진 정도가 비슷함

4. 가설 기각/채택 판단
- 유의확률(p-value)
  - 검정통계량에서 유도할 수 있음
    - 검정통계량의 분포에서 실제 관측치의 위치를 확인했을 때,\
      해당 관측치보다 더 극단적인 값이 나올 확률의 합 
  - 유의 확룰(p)가 유의 수준(α)보다 작을 시에 귀무가설 기각
  - 유의 확룰(p)가 유의 수준(α)보다 클 시에 귀무가설 기각 불가능

피어슨 상관 분석
- 두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정하는 방법
- 귀무가설 (H0): 두 변수 간 선형 상관관계가 존재하지 않는다.
- 대립가설 (H1): 두 변수 간 선형 상관관계가 존재한다.
- 코드
  - import scipy.stats as spst\
	spst.pearsonr(df['col1'], df['col2'])
  - 결과
     - (상관계수, p-value)의 튜플 형태

- 공분산 (Covariance)
  - 두 변수가 함께 변화하는 방향과 크기 표현
    - Cov(x,y)가 양수 -> 양의 관계
    - Cov(x,y)가 음수 -> 음의 관계
    - Cov(x,y)가 0 -> 양도 음도 아닌 관계
     - 선형적인 관계는 없으나, 비선형적 관계가 존재할 수 있음
     - 선형관계: 두 변수간에 직선 형태로 나타나는 관계
     - 비선형관계: 직선이 아닌 다른 형태의 관계 (꺾인 직선, 곡선, 파형 등)
   - np.cov(df['col1'],df['col2'])

- 상관계수(Correlation)
  - 두 변수 간 선형적 관계를 나타내는 통계적 지표
  - 상관계수는 -1과1 사이의 값을 가짐
  - 공분산을 각 변수의 표준편차로 나누어 표준화한 것
  - 방향
    - Corr(x,y)가 양수 -> 양의 상관관계 (1이면 완전한 양의 상관관계)
    - Corr(x,y)가 음수 -> 음의 상관관계 (-1이면 완전한 음의 상관관계)
   - 정도
     - |Corr(x,y)| < 0.3 → 두 변수간 상관관계가 약함
     - |Corr(x,y)| < 0.5 → 두 변수간 상관관계가 중간
     - |Corr(x,y)| > 0.5 → 두 변수간 상관관계가 강함
  - df[['col1','col2']].corr()
  - 주의점
    - 위의 상관관계 정도는 참고용 지표, 절대적 기준 X
    - 상관계수는 선형 관계만 파악 가능 + 직선의 기울기 고려 X
  - 스피어만 상관계수
    - 데이터를 순위화해서 피어슨 상관계수를 구한 것
    - 비선형 관계도 상대적으로 더 잘 파악 가능

[범주형-수치형]
1. T-test (T 검정)
   - 범주가 2개
   - T = 두 평균 간 차이 / 표준오차
   - 귀무가설 (H0): 두 그룹의 평균에는 차이가 없다.
   - 대립가설 (H1): 두 그룹의 평균에는 차이가 있다.
     - **'두 그룹의 평균에 차이가 없음'을 주장하고 싶다
       - T-test 검정법 자체가 두 변수 간 차이를 기준으로 '얼마나 멀리 떨어져 있는지'를 체크
       - '차이가 없음'이라는 가설을 대립가설로 설정할 수는 없음
       - 대신, '귀무가설을 기각하지 못했다'를 통한 간접적 주장은 가능
   - 코드
     - mincho = df.loc[df['민초여부'] == 'Yes', '수명']\
       no_mincho = df.loc[df['민초여부'] == 'No', '수명' ]\
       spst.ttest_ind(mincho, no_mincho)
       - **ttest_ind - ind: independent - 두 범주가 서로 독립적인 경우
         -  ttest_rel - rel: related - 두 범주가 독립적이지 않을 경우
     - 결과
       - TtestResult(Statistic=T값, pvalue=P값, df=자유도)의 형태
   - T값은 보통 절대값 2를 기준으로 함
     - |T|<2 -> 차이가 작음
     - |T|>2 -> 차이가 큼

2. F-test(ANOVA)
   - 범주가 3개 이상
   - Analysis of Variance (ANOVA)
   - **ANOVA에도 매우 다양한 종류 존재
   - 단일 요인 ANOVA(One-Way ANOVA)
     - 가장 기본적 버전
   - F값 = 분모 그룹과 분자 그룹의 분산 비
   - ANOVA 검정에서의 F값 = 집단 간 분산 / 집단 내 분산
       - 전체 평균과 각 집단의 평균을 가지고 만든 분산 / 각 집단의 평균과 개별 값으로 만든 분산
   - H0: 각 MBTI 그룹 별 수명 평균은 차이가 없다.
   - H1: 각 MBTI 그룹 별 수명 평균은 차이가 있다.
   - 코드
     - ISTP = MBTI.loc[MBTI['mbti'] == 'ISTP', '수명']\
       ESFP = MBTI.loc[MBTI['mbti'] == 'ESFP', '수명']\
       INFP = MBTI.loc[MBTI['mbti'] == 'INFP', '수명']\
       spst.f_oneway(ISTP, ESFP, INFP)
     - 결과
       - F_onewayResult(statistic=F값, p-value=P값)
    - F >= 2~3 -> 차이가 존재함 (클수록 차이가 큼)
    - 주의점
      - ANOVA는 전체 평균 대비 각 그룹간 차이가 있는지만 알 수 있음
      - 모두가 서로 다르다가 아니라, 하나가 튀고 나머지가 비슷하더라도\
        귀무가설을 기각하는 결과가 나올 수 있음
        - 보통 사후분석 (post-hoc tests)을 진행
          - 차이가 있다는 것을 안 뒤에 어떤 범주가 튀는지 파악하는 작업
          - Tukey의 HSD(Honest Significant Difference) 테스트 사용
            - 코드
              - from statsmodels.stats.multicomp import pairwise_tukeyhsd\
                tukey = pairwise_tukeyhsd(endog=MBTI['수명'], groups=MBTI['mbti종류'], alpha=0.05)\
                print(tukey)



 - 양측검정과 단측검정
   - 양측검정: 두 그룹 간 '차이'가 존재하는가
   - 단측검정: 두 그룹 간 차이의 '한쪽'만 보고 싶을 때 사용 (방향이 존재할 때)

     - H1: 반민초단의 수명이 민초단의 수명보다 길 것이다. 

[범주형-범주형]

Chi2-test (카이제곱 검정)
- 인사이트 전체 학회원 100명 대상 '경영학과 여부'와 '갈색 머리 여부' 조사
- H0: 경영학과 여부와 갈색머리 여부는 관련이 없다
- H1: 경영학과 여부와 갈색머리 여부는 관련이 있다
- 데이터
  - 경영학과: 20 / 경영학과X: 80
  - 갈색머리: 25 / 갈색머리X: 75
- 두 요소가 서로 독립이라고 가정했을 때의 기댓값
  - 경영, 갈색: 100*20%*25% = 5
  - 경영X, 갈색: 20
  - 경영, 갈색X: 15
  - 경영X, 갈색X: 60
  - 해당 값들이 '기대빈도'
- 실제 관측값
  - 경영, 갈색: 15
  - 경영X, 갈색: 10
  - 경영, 갈색X: 5
  - 경영X, 갈색X: 70
  - 해당 값들이 '관측빈도'
- 카이제곱 통계량 = [(관측빈도-기대빈도)^2/기대빈도]의 총합
- 카이제곱 분석은 범주 갯수 제한 X
  - 범주가 많아질수록 카이제곱 통계량 커짐
  - 카이제곱 통계량 해석은 특정 숫자가 아닌, 자유도의 약 2배 가량을 기준으로 잡음
    - 2배 이상이면 차이가 존재
    - 범주형 변수의 자유도 = [범주의 수 -1]
    - 카이제곱 변수에서의 자유도 = (x변수의 자유도)*(y변수의 자유도)
      - (x변수 범주의 수 -1)*(y변수 범주의 수 -1)
- 코드
  - table = pd.crosstab(df['col1'], df['col2'])\
    spst.chi2_contingency(table)
  - 결과
    - (카이제곱 통계량, p값, 자유도, 기대빈도)
