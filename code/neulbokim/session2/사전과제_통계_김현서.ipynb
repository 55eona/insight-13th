## 1-1) 도입
* 원하는 수치를 얻기 위해 **모집단** 전체를 조사하는 것은 **시간적, 비용적 측면**에서 **비현실적**임.
* 대신, 일부 샘플(**표본집단**)만 추출하여 분석할 수 있음.

### 샘플(표본집단)을 구했을 때 우리의 관심사?
1. 내가 갖고 있는 **샘플 자체**에 대한 관심 
    * ➡️ **기술통계**의 영역
    * **수치**와 **시각화**를 통해 **샘플을 요약하여 표현**하는 방법론
    * EDA는 기술통계의 범주에 해당함.
<br><br>

2. 이 **샘플의 모집단**에 대한 관심
    * ➡️ **추론통계**의 영역
    * **추정**과 **검정**을 통해 **샘플로부터 모집단을 추론**하는 방법론
    * 회귀, 분류 등의 머신러닝은 추론통계의 연장선임.

## 1-2) 통계를 공부해야 하는 이유
* **통계학(Statistics)**: 
    * 데이터를 수집, 분석하여
    * 그로부터 패턴과 관계를 찾아내고
    * 결론을 도출해내는 학문이자 방법론
* **통계학의 본질**: **데이터로부터 의미를 찾는 것**
* 데이터 분석의 최종 목적: **미래를 예측**하는 것
* **데이터 자체**만으로는 **의미 있는 정보**를 얻을 수 없으므로, **통계학적 지식을 활용**해야 함.


## 1-3) 데이터의 척도
### 1️⃣ 데이터란?
* 통계학에서 데이터란: **관찰**과 **조사**를 통해 얻은 **수치적** 또는 **비수치적 자료**
### 2️⃣ 데이터의 구분
* **양적 자료(quantitive, numerical data)**: 숫자로 표현 가능
    * **구간(등간) 척도 (interval scale)**
        * 연속적인 수로 수량화 가능
        * 숫자간의 간격이 동일함 ⇒ 크고 작은 정도를 표현할 수 있음
        * 절대적인 원점(0)은 존재하지 않음
        * 덧셈, 뺄셈 가능하지만 곱셉, 나눗셈은 불가능
            * ex) 온도(0도: 온도가 없음을 의미하는 게 아님), IQ 지수 등
    * **비율 척도 (ratio scale)**
        * 연속적인 수로 수량화 가능
        * 절대적인 원점이 존재함
        * 모든 사칙연산이 가능함
            * ex) 몸무게(0이 실제로 무게가 존재하지 않음을 의미함), 인구수, 길이 등
* **질적 자료(qualitative, categorical data)**: 숫자로 표현 불가능
    * **명목 척도 (nominal scale)**
        * 다른 뜻 없이 단순 범주(이름)만 나타내는 척도
        * 숫자로 표현할 수는 있으나, 분류의 기능만 할 뿐이지 양적인 의미는 전혀 가지지 않음
        * 사칙연산 수행 불가능
            * ex) 색깔(빨: 0, 주: 1, 노: 2, 초: 3), 성별(남: 0, 여: 1) 등
    * **순위(서열) 척도 (ordinal scale)**
        * 순위 또는 서열이 있는 척도
        * 수량화 가능
        * 범주의 순서를 비교할 수 있으나, 얼마만큼 차이가 난다고는 해당 자료를 통해 말할 수 없음
        * 사칙연산 수행 불가능
            * ex) 달리기 순위(1등, 2등, 3등), 학년(1학년, 2학년), 만족도(낮음: 0, 보통:1, 높음: 2)

## 2-1) 기술통계란?
### 📊 기술 통계학 (Descriptive Statistics)
* 데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야
* **평균, 중앙값, 표준편차** 등의 개념이 이에 해당
* 이를 통해 데이터의 기본적인 특성을 파악할 수 있음
#### ➡️ 측정이나 실험을 통해 수집한 자료를 정보로 전환하기 위하여 요약/정리하여 그 특성을 기술하는 절차에 관한 개념
#### 1️⃣ 수치 기술 통계
* 하나의 수치, 즉 기술통계량으로 데이터를 요약&설명하는 방법
    * 중심 경향성, 산포도 등
#### 2️⃣ 시각화 기술 통계
* 표와 그래프의 형태로 데이터를 시각화하는 방법 (EDA에서 배웠던 내용)

## 2-2) 중심 경향성 (수치 기술 통계)
* 데이터의 중심적인 경향을 나타내는 기술 통계량
### 1️⃣ 평균(mean)
* **(데이터 값의 총합)/(데이터의 개수)**
* 데이터의 **중심 경향성**을 나타내는 **가장 대표적인 지표**
* 평균은 **총합치를 사용**하므로 **이상치(outlier)에 민감**할 수 있음.
* `series.mean()`

### 2️⃣ 중앙값(median)
* 데이터 값들을 크기 순으로 나열했을 때 **가운데 위치한 값**
* 평균과 달리 **이상치에 덜 민감**함.
* 데이터의 **중심 경향성**을 나타내는 **대안적인 지표**
* `series.median()`

### 3️⃣ 최빈값(mode)
* 데이터에서 **가장 자주 나타나는 값**
* 주로 **이산형 데이터(discrete data)에서 사용**됨
* 데이터의 분포를 이해하는 데에 도움을 줌
* `series.mode()`

### ➕) 왜도(skewness)
* **평균, 중앙값, 최빈값**을 통해 **자료의 개형(왜도: Skewness 등)** 을 확인할 수 있음
* 중심 경향성을 보여주지는 않지만, **데이터 분포의 비대칭성**을 나타내는 통계적 지표임
* 데이터 분포의 특징을 설명할 때 주로 사용함
    1. **Positive Skewness (Right-skewed)**
        * 오른쪽 꼬리가 왼쪽보다 더 길 때
        * 평균, 중앙값 > 최빈값
        * ex) 소득 분포: 소수의 고소득자가 평균을 올림.
    2. **Negative Skewness (Left-Skewed)**
        * 왼쪽 꼬리가 오른쪽보다 더 길 때
        * 평균, 중앙값 < 최빈값
        * ex) 사망 연령 분포: 소수가 일찍 사망함으로써 평균을 내림
    3. **Normal Distribution**
        * 왼쪽과 오른쪽이 대칭인 정규 분포
        * 평균 = 중앙값 = 최빈값
* `series.skew()`

## 2-3) 산포도 (수치 기술 통계)
* 데이터가 **퍼져있는 정도**를 설명하는 기술 통계량
### 1️⃣ 분산(variance)
* 데이터 값들이 **평균으로부터 얼마나 멀리 퍼져있는지**를 나타내는 중요한 통계적 측정 지표
* 분산 ↑ = 데이터가 많이 퍼져있음
* 각 데이터 값과 평균 간의 차이(편차)를 제곱한 값들의 평균을 통해 구할 수 있음
    * (편차의 제곱의 평균)
    * ${\displaystyle\frac{\sum_{k=1}^N (x_i-\mu)^2}{N}}$
        * $N$: 전체 데이터의 수(=도수)
        * $x_i$: i번째 데이터의 값
        * $i$: 1~N까지의 자연수
        * $\mu$: 전체 데이터의 평균
* `np.var(arr)`
### 2️⃣ 표준편차(standard deviation)
* 분산에 제곱근을 취한 값
    * ${\sqrt{\displaystyle\frac{\sum_{k=1}^N (x_i-\mu)^2}{N}}}$
        * $N$: 전체 데이터의 수(=도수)
        * $x_i$: i번째 데이터의 값
        * $i$: 1~N까지의 자연수
        * $\mu$: 전체 데이터의 평균
* **표준편차를 사용하는 이유?**
    * 분산의 단위: **관측값 단위의 제곱**이 되므로, 대상 데이터의 단위와 일치하지 않음.
    * 제곱근을 씌워 **원래 데이터의 단위와 일치**시켜주기 위함.
* `np.std(arr)`
### 3️⃣ 범위(range)
* 데이터의 **최대값과 최소값의 차이**를 나타내는 지표
* 데이터의 **전체 변동 범위**를 알려줌
* `np.max(arr) - np.min(arr)`
### 4️⃣ 사분위수(quartile)
* 데이터의 **표본**을 **네 개의 동일한 크기로 나눈 값**
    1. Q1 = 하위 25%에 위치한 값. **하위 사분위수**
    2. Q2 = 하위 50%에 위치한 값. **중앙값**
    3. Q3 = 하위 75%(상위 25%)에 위치한 값. **상위 사분위수**
* `np.percentile(arr, 25)`
* `np.percentile(arr, 50)`
* `np.percentile(arr, 75)`
### 5️⃣ 사분위간 범위(IQR)
* Q3에서 Q1 값을 제외한 범위.
* 데이터의 중앙 50%가 퍼져 있는 범위.
* 이상치를 탐지할 때 유용함
    * 아래쪽 이상치: $Q1 - 1.5*IQR$보다 작은 값
    * 위쪽 이상치: $Q3 + 1.5*IQR$보다 큰 값

* 네모 박스가 사분위 수, 화살표의 범위가 사분위간 범위를 나타냄
### 6️⃣ 변동계수(coefficient of variation)
* 데이터의 절대적인 변동성이 아닌, **상대적인 변동성**을 나타내는 통계적 지표
* 표준편차를 해당 데이터의 평균으로 나눈 값으로 계산됨
    * $변동계수 = (표준편차/평균)*100$
* **변동계수를 사용하는 이유?**
    * 표준편차는 **단위가 존재** => 평균값이 다른 두 표본집단 비교 / 단위가 다른 두 데이터 비교 시에 적절하지 않음
    * 이때 표준편차를 평균으로 나눠주면 **단위가 사라짐**
    * 집단의 **퍼진 정도**를 **동일한 기준**으로 비교할 수 있게 됨
* `np.std(arr) / np.mean(arr)*100`

## 3-1) 추론통계란?
### 📊 추론 통계학 (Inferential Statistics)
* 수집한 데이터를 기반으로 **미래의, 혹은 조사하지 못한 전체 정보를 추론**하고 결론을 도출하는 방법을 다루는 분야
* **통계적 가설검정, 신뢰구간** 등의 개념이 이에 해당
#### ➡️ 모집단에서 일부를 뽑아 만든 표본집단을 통해 모집단에 대한 정보를 추측하는 과정

## 3-2) 확률분포
### 〽️ 확률 변수
* random한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수
    * ex) 동전을 던지는 실험: '앞면: 1', '뒷면: 0'으로 대응
### 🔆 확률 분포(probability distribution)
* 확률변수가 특정한 값을 가질 확률을 나타내는 함수
### 📉 이산확률분포
#### 1️⃣ 베르누이분포 (Bernoulli Distribution)
#### 2️⃣ 이항분포 (Bionomial Distribution)
#### 3️⃣ 기하분포 (Geometric Distribution)
#### 4️⃣ 음이항분포 (Negative Bionomial Distribution)
#### 5️⃣ 초기하분포 (Hypergeometric Distribution)
#### 6️⃣ 포아송분포 (Poisson Distribution)

### 📈 연속확률분포
#### 1️⃣ 정규분포 (Normal Distribution)
* **형태**
    * 종(bell) 모양
    * 평균을 기준으로 대칭
    * 표준편차가 클수록 곡선이 평평해짐 (데이터가 넓게 퍼짐
* **매개변수**
    * 평균($\mu$), 표준편차($\sigma$)
* **특징**
    * 가장 잘 알려진 확률분포
    * 많은 자연 현상과 사회 현상이 정규분포를 따르는 것으로 알려져 있음
    <img src="https://img.notionusercontent.com/s3/prod-files-secure%2F8d7ae215-ed29-46ba-88cb-25d0d6cd52f9%2F9bb7d1d7-cbbf-46f3-8270-f0d246ace12e%2FUntitled.png/size/w=1700?exp=1742373997&sig=eaa45Gthu-eqx6oxCRcHod1aJp6EPMJdKOGseRQwP28" width="500" height="500"/>
    * **중심극한정리**에 의해 모집단의 분포와 상관없이 **표본의 크기가 커질수록 표본평균의 분포는 정규분포에 근사하게 됨** (일반적으로 `n(표본의 크기)>=30`일 때)
    <img src="https://img.notionusercontent.com/s3/prod-files-secure%2F8d7ae215-ed29-46ba-88cb-25d0d6cd52f9%2F9aeea19f-d8fb-4c6f-acc5-f9a41a189d2c%2FUntitled.png/size/w=1700?exp=1742374148&sig=Pg89GbYgAWBdDH2iTztjcjOwgOCx_Tlxr5NcfpMbp4E" width="500" height="500"/> 
    * **대수의 법칙**
        * 표본의 크기가 커질수록 **표본평균이 모평균에 근사**하는 현상
        * 취합하는 표본의 수가 많을수록 통계적 정확도가 올라감
        * 데이터가 많을수록 좋음

#### 2️⃣ T분포(T-Distribution)
* **형태**
    * 종(bell) 모양
    * 정규분포보다 꼬리가 두꺼움
    * 평균을 기준으로 대칭
    * **자유도가 커짐**에 따라 **표준정규분포 $N(0,1)$에 수렴**
* **매개변수**
    * 자유도(df)
* **특징**
    * 표본으로 모평균(혹은 모평균 간의 차이)을 추정할 때 사용되는 분포
    * **표본 크기가 작을 때** 정규분포보다 더 정확한 결과를 제공함.
    <img src="https://img.notionusercontent.com/s3/prod-files-secure%2F8d7ae215-ed29-46ba-88cb-25d0d6cd52f9%2Fea1ed642-63c6-493e-8ae3-3e1636ba1c6e%2Fimage.png/size/w=1700?exp=1742374363&sig=oukPFv2y1T8q7G-7wghaLlVkCfDaqbSBao3Vh79VMPA" width="500" height="500"/> 
* **T분포의 자유도**는 **표본 크기에 비례**함.
    * **자유도가 높을수록** (더 많은 독립적인 정보를 사용할수록) **T분포는 정규분포에 가까워**지고, **통계적 분석의 정확도가 높아짐**.
    * 그러나 자유도가 너무 높아지면 **과적합(Overfitting)** 과 같은 문제가 발생할 수 있으므로, 적절한 자유도를 선택하는 것이 중요함.
* **표준정규분포 vs T분포**
    * 표준정규분포보다는 **T분포가 훨씬 더 중심으로부터 많이 퍼져있음**.

    * **표준정규분포**
        * 평균이 0이고 표준편차가 1인 정규분포
        * 모집단의 평균과 분산을 정확히 알고 있을 때 사용됨
        * 주어진 모집단에서 직접 계산됨 => 불확실성 ↓
    * **T분포**
        * 모집단의 분산을 모를 때, 표본의 분산을 사용해 모집단 분산을 추정하여 만들어짐
        * 표본의 크기가 작을수록 이 추정은 불확실성이 커짐
        * 이러한 불확실성 때문에 T분포는 표준정규분포보다 중심으로부터 더 퍼져 있음.
        * 극단적인 값이 나올 가능성을 반영한 것임.
    * 통상적으로 모집단의 분산이 알려져 있지 않은 경우, <br> 
        표본의 크기가 **30 미만**이면 **t분포**를, <br>
        표본의 크기가 **30 이상**이면 **표준정규분포**를 이용함. (feat 중심극한정리)

## 4-1) 가설 검정의 절차
### 1️⃣ 가설 설정 - 2️⃣  유의 수준 설정 - 3️⃣ 검정통계량 산출 - 4️⃣ 가설 기각/채택 판단
### 1️⃣ 가설 설정
* **귀무가설 $H_0$** 과 **대립가설 $H_1$** 를 설정함
* **귀무가설**은 주로 **기각하고자 하는 명제**로 설정함.
    * 보통 "~는 차이가 없다"라는 명제임
    * 현재 통용되고 있는 가설
    * 지지될 때 위험성이 비교적 약한 가설 (무죄추정의 원칙)
* **대립가설**은 주로 **내가 주장하고자 하는 명제**로 설정함.
    * 보통 "~는 차이가 존재한다"라는 명제임
    * 새롭게 주장하고 싶은 가설
    * 지지될 때 위험성이 비교적 강한 가설
### 2️⃣ 유의 수준 설정
* **유의수준**이란?
    * 연구자가 결과를 판단할 때 **허용할 수 있는 오류의 최대치**
    * 연구자가 어떤 결정이 얼마나 신뢰할 만한지를 판단하는 기준이 됨.
    * 주로 0.05, 즉 5%로 설정함.
        * 귀무가설이 실제로 참인데도 불구하고, 5%의 확률로 귀무가설을 잘못 기각할 수 있음을 의미함
        * 5%로 발생할 수 있는 오류까지는 감수하면서, 귀무가설을 기각할 준비가 되어 있다는 뜻임.
        * ex) 약물 효과 검증시 유의 수준 0.05: 이 약물이 실제로는 효과가 없더라도, 5%의 확률로 약물이 효과가 있다고 잘못 결론 내릴 가능성을 수용하는 것
* **1종, 2종 오류**
    * **1종 오류($\alpha$)** : 귀무가설이 실제로 참이었는데 기각할 확률
    * **2종 오류($\beta$)** : 귀무가설이 실제로 거짓인데 기각하지 못할 확률
    * 가설검정은 **1종 오류를 더 치명적인 것**으로 파악함.
    * 따라서 **유의수준은 1종 오류의 확률**을 의미하게 됨.
      <img src="https://img.notionusercontent.com/s3/prod-files-secure%2F8d7ae215-ed29-46ba-88cb-25d0d6cd52f9%2Fa1f29b64-4728-4ff0-95b3-3217db6e85e7%2FUntitled.png/size/w=1700?exp=1742387162&sig=qFtYfKTgxjDKMZiR6tD4DKLIZ_2ziA-q_VC_UwXaZwk" width="500" height="500"/>
### 3️⃣ 검정통계량 산출
* **검정통계량**이란?
    * 가설 검정의 결과를 판단하는 데 사용되는 값
    * 데이터의 분포나 특징에 따라 **T값, F값** 등 어떤 통계량을 사용할지 정해야 함. <br> <br>
    
* **Z값(Z-Score)**
    1. 정의
        * Z값(Z점수)는 **표준 정규 분포(평균이 0이고 표준편차가 1인 분포)** 를 사용하여 얻은 값
        * 특정 데이터가 **평균으로부터 얼마나 떨어져 있는지**를 **표준편차 단위**로 나타냄
    2. 계산
        * $Z={\displaystyle\frac{(X-\mu)}{\sigma}}$
        * $X$: 관측값, $\mu$: 평균, $\sigma$: 표준편차
    3. 사용
        * 큰 표본 크기(일반적으로 30 이상)에 대한 가설 검정에 주로 사용됨
        * 예를 들어, 모집단의 평균에 대한 가설을 검정할 때 사용됨
    4. 수치 해석
        * **|Z값|이 크면?** <br>
            => 해당 데이터가 **평균으로부터 멀리 떨어져 있**으며, 때로는 비정상적이거나 이상치일 가능성이 있음을 뜻함
        * **임계점 찾기** <br>
            일반적으로 시행하는 양측검정(양쪽 꼬리에서 임계점을 찾을 때)에서는 <br>
            $Z=1.96$일 때 유의수준 0.05 (=신뢰수준 95%) <br>
            $Z=2.58$일 때 유의수준 0.01 (=신뢰수준 99%) <br>
            을 기준으로 판단함. <br>
            ➡️ Z값이 임계점보다 크다면 귀무가설 기각 <BR>
            (신뢰구간 = 1 - 유의 수준($\alpha$))
    
* **T값(T-Score)**
    1. 정의
        * T값은 **T분포**를 사용하여 얻은 값
        * 작은 표본 크기에 대한 가설 검정(T검정)에 주로 사용됨
        * T분포는 정규분포와 유사하지만, 꼬리 부분이 더 두꺼워 **작은 표본에서 발생할 수 있는 불확실성**을 보다 잘 반영함.
    2. 계산
        * $T={\displaystyle\frac{(X-\mu)}{s/\sqrt{n}}}$
        * $X$: 표본 평균, $\mu$: 연구자가 설정하는 값, $s$: 표본 표준편차, $n$: 표본 크기
    3. 사용
        * 주로 **모집단의 평균**에 대한 **가설을 검정**하거나,
        * **두 표본 평균 간의 차이를 검정**할 때 사용됨
    4. 수치 해석
        * **|T값|이 크면?** <br>
            => 해당 데이터가 **모집단 평균으로부터 멀리 떨어져 있**으며, 관측된 표본 평균이 모집단 평균과 유의미한 차이가 있을 가능성이 높다는 것을 나타냄.
        * **임계점 찾기** <br>
            T분포는 자유도(df)에 따라 모양이 달라지므로 <br>
            자주 쓰는 자유도와 유의수준에 따른 t분포표 라는 것이 사용되고 있음. <br>
            ➡️ 같은 유의수준에 대해서 **t값은 z값보다 커**지며 **자유도가 높아질수록 z값에 수렴**함
    
* **F값(F-Score)**
    1. 정의
        * F값은 F분포를 사용하여 얻은 값
        * 주로 두 표본 분산의 차이를 비교하는 데 사용됨
        * F분포는 두 개의 독립적인 카이제곱 분포의 비율로 정의됨
    2. 계산
        * $F={\displaystyle\frac{\chi^2(d_1)/d_1}{\chi^2(d_2)/d_2}}$
        * 자유도 $d_1$과 $d_2$를 가지는 F분포는 $F(d_1, d_2)$로 표기함
    3. 사용
        * 주로 분산분석(ANOVA)에서 두 개 이상의 그룹 간 평균의 차이를 검정하는 데 사용됨
    4. 수치 해석
        * F값은 두 표본의 분산의 비이므로, <br>
            이 값이 **1에 가까우면 두 표본의 분산이 서로 비슷하다**는 것을 의미함. <br>
            즉, **두 그룹이 퍼진 정도가 비슷하다**고 볼 수 있음.
### 4️⃣ 가설 기각/채택
* **유의확률(p-value)**
    * 가설 검정에 쓰일 수 있는 또 다른 척도
    * 검정통계량에서 유도할 수 있음
    * **유의확률(p)가 유의 수준($\alpha$)보다 작을 시에 귀무가설을 기각**
    * **유의확률(p)가 유의 수준($\alpha$)보다 클 시에 귀무가설을 기각하지 못함**

## 4-2) 데이터의 자료형에 따른 검정 방법
### 0️⃣ 공분산과 상관계수
#### ☑️ 공분산(covariance)
* 두 변수가 **함께 어떤 방향으로 변화하는지**와 **그 크기**를 표현하는 데에 사용됨
* `np.cov(df['col1'], df['col2'])`
* $Cov(X,Y)>0$이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
* $Cov(X,Y)<0$이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
* $Cov(X,Y)=0$이면 X와 Y는 양도 음도 아닌 관계
    * 선형적인 관계는 없으나, **비선형적인 관계가 존재할 수 있음**
        * ex) X가 Y의 제곱에 비례하는 경우
        * **선형 관계**
            * 두 변수 간에 직선 형태로 나타나는 관계.
            * 하나가 증가할 때 다른 하나가 같이 증가하거나, 감소하는 것
            * ex) $y=x+c$
        * **비선형 관계**
            * 두 변수 간에 직선이 아닌 다른 형태로 나타나는 관계.
            * 꺾인 직선, 곡선, 파형 등.
            * ex) $x^2+y^2=r^2$
#### ☑️ 상관계수(correlation)
* 두 변수 간의 **선형적인 관계**를 나타내는 통계적 지표
* 상관계수는 **-1과 1 사이의 값**을 가짐
* `df[['col1', 'col2']].corr()`
* **상관관계와 인과관계는 다르다**는 점을 주의해야 함
    * **상관관계**
        * 두 변수가 수치적으로 연관되어 있는 관계
    * **인과관계**
        * 한 변수가 다른 변수를 발생시키는 관계
* 공분산을 각 변수의 표준편차로 나누어 -1과 1 사이로 표준화한 것이므로 다음과 같이 해석함
    * $Corr(X,Y)>0$이면 양의 상관관계 (1이면 완전한 양의 상관관계)
    * $Corr(X,Y)<0$이면 음의 상관관계 (-1이면 완전한 음의 상관관계)

    * 일반적으로 우리가 말하는 상관계수는 **피어슨 상관계수**를 뜻함.

#### ☑️ [범주형-수치형] 시각화로 알아보는 방법: 막대그래프 그리기
---
```python
sns.barplot(x='col1', y='col2',data=df)
plt.show()
```
---
* 타이타닉 데이터의 생존여부(범주형)별 나이(수치형) 평균
<img src="https://img.notionusercontent.com/s3/prod-files-secure%2F8d7ae215-ed29-46ba-88cb-25d0d6cd52f9%2Fecf800db-4e66-4d52-b78e-638b98885c59%2FUntitled.png/size/w=1230?exp=1742390664&sig=s_0kjR2ZiUd1hot6rRxijFMi1q2xM4TlHYhBEaLhaSA" width="500" height="500"/>
* 검은 선은 각 평균의 95% 신뢰구간(CI)을 나타냄.
    * 95% 확률로 신뢰구간 안에 모평균이 포함됨.
    * 표본을 100번 뽑으면 95번 정도 해당 신뢰구간 안에 모평균을 포함함.
* **결과 해석**
    * 눈으로 봤을 떄 두 평균의 차이가 크고, <br>
        신뢰구간이 거의 겹치치 않으면 <br>
        ➡️ 대립가설을 지지할 가능성이 높다.
        
#### ☑️ T-test: 범주가 2개일 때
* 범주가 2개일 때 => **T-test 사용!!**
    * T-test 수행 시 t값: 두 평균 간 차이를 표준 오차로 나눈 값.
* 가설
    * 귀무가설 $H_0$: 두 그룹의 평균에는 차이가 없다.
    * 대립가설 $H_1$: 두 그룹의 평균에는 차이가 있다.
---
```python
mincho = df.loc[df['민초여부'] == 'Yes', '수명']
no_mincho = df.loc[df['민초여부'] == 'No', '수명']

spst.ttest_ind(mincho, no_mincho) # ind: independent(두 범주가 독립적인 집단)
#spst.ttest_rel(before, after) # rel: related(같은 그룹이 다른 상태로 변화)
```
---
* 결과는 `TtestResult(statistic=t값, p-value=p값, df=자유도)`의 형태로 나옴.
* t값은 보통 |2|를 기준으로, 그보다 크면 차이가 있으며 클수록 차이가 크다고 생각함.
* p값은 이 결과가 얼마나 믿음직한지를 나타내며, 보통 0.05를 기준으로 함.

#### ☑️ F-test(ANOVA): 범주가 3개 이상일 때
* 범주가 3개 이상일 때 => **F-test 사용!!**
    * ANOVA(Analysis Of Variance, 분산분석): F-test의 일종
        1. 단일 요인 ANOVA (One-Way ANOVA)
        2. 이중 요인 ANOVA (Two-Way ANOVA)
        3. 반복 측정 ANOVA (Repeated Measures ANOVA)
        4. 혼합 설계 ANOVA (Mixed-Design ANOVA)
        5. 공분산분석 (ANCOVA)
        6. 다변량 분산분석 (MANOVA)
        7. 등등.. 다양한 종류가 있음
    * F-통계량: 분모 그룹과 분자 그룹의 분산 비.
    * ANOVA 검정 시 F값: **집단 간 분산 / 집단 내 분산**
        * (전체 평균과 각 집단의 평균을 가지고 만든 분산) / (각 집단의 평균과 개별 값으로 만든 분산) 으로 계산함.
* 가설
    * 귀무가설 $H_0$: 각 MBTI 그룹별 수명 평균은 차이가 없다.
    * 대립가설 $H_1$: 각 MBTI 그룹별 수명 평균은 차이가 있다.
---
```python
ENFP = MBTI.loc[MBTI['mbti']=='ENFP', '수명']
INFP = MBTI.loc[MBTI['mbti']=='INFP', '수명']
INTJ = MBTI.loc[MBTI['mbti']=='INTJ', '수명']

spst.f_oneway(ENFP, INFP, INTJ)
```
---
* 결과는 `F_onewayResult(statistic=f값, p-value=p값)`의 형태로 나옴.
* f값은 보통 2~3 이상이면 차이가 있으며 f값이 클수록 차이가 크다고 생각함.
* p값은 f값 결과가 얼마나 유의미한지를 의미함.


### 3️⃣ [범주형 - 범주형] $\chi^2$-test
#### ☑️ [범주형-범주형] 시각화로 알아보는 방법: 모자이크플롯 그리기
---

---
* 타이타닉 데이터의 pclass-객실(범주형)별 survived-생존여부(범주형) 모자이크 플롯
<img src="https://img.notionusercontent.com/s3/prod-files-secure%2F8d7ae215-ed29-46ba-88cb-25d0d6cd52f9%2F61932acb-b4ec-4214-915c-4bf336a81aeb%2FUntitled.png/size/w=1700?exp=1742391731&sig=GGoxgmZnNxBjCTbpt8JBYtvm_2y9-FKcBXowmACdPSs" width="500" height="500"/>
* 눈으로 봤을 때
    * 범주별 차이가 명확히 보인다면 => **유의미한 관계 있을 가능성 큼**
    * 범주별 차이가 적다면 => 범주가 달라져도 y값이 달라지지 않음, 즉 **독립적인 관계일 가능성 큼**

#### ☑️ $\chi^2$-test
* 가설 (예시)
    * 귀무가설 $H_0$: 경영학과 여부와 갈색머리 여부는 관련이 없다.
    * 대립가설 $H_1$: 경영학과 여부와 갈색머리 여부는 관련이 있다.
* **기대빈도**
    * 기대할 수 있는 값
* **관측빈도**
    * 실제로 관측된 값
* 같은 위치에 있는 **기대빈도**와 **관측빈도**의 차이가 서로 크다면, 경영학과 여부와 갈색머리 여부는 
* $\chi^2 = {\displaystyle\frac{\sum(관측빈도 - 기대빈도)^2}{기대빈도}}$
* 카이제곱 분석은 **범주 개수에 제한이 없음**
    * 따라서 범주가 많아질수록 카이제곱 통계량이 커질 수밖에 없음
    * 따라서 카이제곱 통계량 해석은 어떠한 특정 숫자가 아닌, **자유도의 약 2배 가량**을 기준으로 잡음. (2배 이상이면 차이가 존재)
* **범주형 변수의 자유도**: `범주의 수 - 1`
* 카이제곱 변수에서의 자유도: `(x 변수의 자유도)*(y 변수의 자유도)`, 즉 <br>
                       `(x 변수의 범주의 수 - 1) * (y 변수의 범주의 수 - 1)`
* 코드
---
```python
table = pd.crosstab(df['col1'], df['col2']) #교차표 작성
spst.chi2_contingency(table)
```
---
* 결과는 순서대로 `(카이제곱 통계량, p값, 자유도, 기대빈도)`가 됨.
