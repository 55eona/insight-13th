# <span style="background-color:#E0F7FA; color:#1565C0; padding:4px; border-radius:5px;"> regplot</span>

▶ 회귀선이 추가된 산점도 그래프 <br>
▶ 두 개의 연속 변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선을 함께 나타냄

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x='Age',y='Fare',data=data,ax=ax1) ## 회귀선 있는 버전
sns.regplot(x='Age',y='Fare',data=data, ax=ax2, fit_reg=False) ## 회귀선 없는 버전

plt.show()
```

▶ f는 도화지, ax는 만화를 그릴 때의 하나의 컷!

### (8) 1행 1열에 승선항(Embarked)별 승객의 수를 알 수 있는 시각화를 진행해보세요.

Q. 승객이 많은 승선항 순으로 나열하면? <br>
A. S ⇒ C ⇒ Q

```python
# 위 마크다운에 '?' 채우고 코드 작성
f, ax = plt.subplots(2, 2, figsize=(12, 4))
f.subplots_adjust(hspace=1)

sns.countplot(x='Embarked', data=data, ax=ax[0][0])
ax[0][0].set_title('Embarked')
plt.show()
```

### (9) 2행 2열에 객실 등급(Pclass)에 따른 생존율을 알아 볼 수 있는 시각화를 진행해보세요.

Q. 생존률이 가장 높은 등급은? <BR>
A. 1

```python
# 위 마크다운에 '?' 채우고 코드 작성
f, ax = plt.subplots(2, 2, figsize=(12, 4))
f.subplots_adjust(hspace=1)

sns.countplot(x='Embarked', data=data, ax=ax[0][0])
ax[0][0].set_title('Embarked')

sns.countplot(x='Pclass', hue='Survived', data=data, ax=ax[1][1])
ax[1][1].set_title('Survived vs Dead by Pclass')
plt.show()
```

### (10) 나머지 칸에 본인이 파악하고 싶은 정보를 자유롭게 시각화하고, 인사이트를 작성해보세요.

Q. 알아낸 것은? <BR>
A1. 함께 탑승한 형제자매의 수(SibSp)가 1~2명인 경우, 10살 이하의 생존률이 높다. 이는 형제자매 중 나이가 많은 탑승자가 나이가 어린 탑승자를 먼저 구조하고자 했던 것이라고 추정할 수 있다.  <BR> <BR>
A2. 그러나 함께 탑승한 형제자매의 수(SibSp)가 3명 이상일 경우, 10살 이하의 생존률이 높지 않고 오히려 낮다. 이는 동나이대의 어린 형제자매였을 경우 구조되기 어려웠음을 추정하게 한다. 또는 나이가 많은 형제자매와 탑승한 어린 아이였어도, 함께 탑승한 형제자매가 많았을 때 오히려 생존에 불리했다는 점을 보여준다. 예를 들면 20살 형, 10살 동생, 7살 동생, 5살 동생이 있었을 경우 20살 형이 3명의 동생을 모두 구조되게 챙기기는 어려웠을 것이다. <BR>  
A3. 함께 탑승한 부모, 자식의 수(Parch)가 4명인 경우, 즉 늙은 부모와 어린 자식 모두를 챙겨야 하는 경우이거나 자식의 수가 매우 많은 경우 사망률이 전 연령에서 매우 높다는 점을 주목할 만하다. 그래프 상으로는 모두 사망한 것으로 보인다. 이는 부양 가족이 매우 많은 피부양자가 자신을 희생하고 다른 가족들을 구조하고자 했던 것으로 보인다. <BR>


```python
f, ax = plt.subplots(2, 2, figsize=(12, 8))
f.subplots_adjust(hspace=1)

sns.countplot(x='Embarked', data=data, ax=ax[0][0])
ax[0][0].set_title('Embarked')

sns.countplot(x='Pclass', hue='Survived', data=data, ax=ax[1][1])
ax[1][1].set_title('Survived vs Dead by Pclass')

sns.violinplot(x='SibSp', y='Age', hue='Survived', data=data, split=True, ax=ax[0][1])
ax[0][1].set_title('SibSp and Age vs Survived')
ax[0][1].set_yticks(range(0,110,10))


sns.violinplot(x='Parch', y='Age', hue='Survived', data=data, split=True, ax=ax[1][0])
ax[1][0].set_title('Parch and SibSp vs Survived')
ax[1][0].set_yticks(range(0,110,10))


plt.show()
```
