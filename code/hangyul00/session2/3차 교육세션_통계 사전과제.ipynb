1. 통계란?
모집단: 전체 / 표본집단: 전체에서 추출한 일부
기술통계: 수치와 시각화를 통해 샘플을 요약하여 표현하는 방법론
(ex. 내가 갖고 있는 샘플 자체에 대한 관심)
추론통계: 추정과 검정을 통해 샘플로부터 모집단을 추론하는 방법론
(ex. 이 샘플의 모집단에 대한 관심)

통계학: 데이터를 수집하고 분석하여 그로부터 패턴이나 관계를 찾고, 결론을 도출하는 학문
(=데이터에서 의미를 찾는 것)
-> 최종 목적은 "데이터에서 의미를 찾는 것"

* 데이터란?
통계학에서 데이터란 관찰과 조사를 통해 얻은 수치적, 비수치적 자료
양적 자료: 숫자로 표현할 수 있는 데이터
- 구간 척도: "연속적인 수"로 수량화할 수 있으며, 숫자간의 간격이 동일하여 크고 작음을 표현할 수 있음
- Ex. 온도 (0도가 온도가 "없음"을 의미하지 않음)
- 비율척도: 연속적인 수로 수량화할 수 있으며, 절대적인 원점이 존재함
- Ex. 몸무게 (0은 없음을 의미함)
질적 자료: 숫자로 표현할 수 없는 데이터
- 명목 척도: 단순히 단순 범주만 나타내는 척도 (숫자로 표현할 수는 있지만 분류의 기능만 함)
- Ex. 색깔 (빨강: 0, 파랑: 1, 노랑: 2)
- 서열척도: 순위 또는 서열이 있는 척도로 수량호할 수 있는 자료
- Ex. 달리기 순위, 학년, 만족도

```python
2. 기술통계란
기술통계학: 데잉터의 주요 특성을 요약하고 설명하는 분야
    1) 수치기술통계: 기술통계량으로 데이터를 요약하고 설명하는 방법
    2) 시각화 기술통계: 표와 그래프의 형태로 데이터를 시각화하는 방법
중심경향성: 데이터의 중심적인 경향을 나타내는 기술 통계량
    1) 평균: 데이터의 값의 총합/데이터의 개수 (이상치에 민감할 수 있음) / series.mean()
    2) 중앙값: 데이터 값들을 크기 순으로 나열했을 때 가운데 위치한 값 (이상치에 덜 민감) / series.median()
    3) 최빈값: 가장 자주 나타나는 값 (주로 이산형 데이터에서 사용되고 분포이해에 도움) / series.mode()

* 왜도? 
왜도는 데이터 분포의 비대칭성을 나타내는 통계적 지표 (series.skew())
right-skewed: 오른쪽 꼬리가 왼쪽보다 긴 상태 (평균,중앙값>최빈값)
left-skewed: 왼쪽 꼬리가 오른쪽보다 긴 상태 (평균,중앙값<최빈값)
normal distribution: 정규분포 (평균=중앙값=최빈값)

산포도?
산포도: 데이터가 퍼져있는 정도를 설명하는 기술 통계량
    1) 분산 (np.var(arr)): 데이터 값들이 평균으로부터 얼마나 멀리 떨어져있는지를 나타내는 지표 / 분산이 높으면 데이터가 많이 퍼져있음
    2) 표준편차 (np.std(arr)): 분산에 제곱근을 취한 값 / 분산은 관측값 단위의 제곱이 되므로 제곱근을 씌워 단위를 일치시켜야 함
    3) 범위 (np.max(arr)-np.min(arr)): 최대값과 최소값의 차이 / 전체 변동 범위를 알려주는 지표
    4) 사분위수: 데이터의 표본을 네 개의 동일한 크기로 나눈 값
    - Q1: 하위 25% / Q2: 중앙값 / Q3: 상위 25%
    - np.percentile(arr, 25) / np.percentile(arr,50) / np.percentile(arr,75)
    5) IQR: Q3-Q1의 값 / 즉, 데이터의 중앙 50%가 퍼져있는 범위
    - 아래쪽 이상치: Q1 - (1.5*IQR)
    - 위쪽 이상치: Q3 + (1.5*IQR)
    6) 변동계수: 데이터의 상대적인 변동성을 나타내는 통계적 지표
    - 변동계쑤 = (표준편차/평균)*100
    - 표준편차는 단위가 존재하기 때문에 평균으로 맞춰주어 단위를 사라지게 만들고, 동일한 기준으로 비교할 수 있게 되므로 사용함
    - np.std(arr) / np.mean(arr) *100
```

```python
3. 추론통계란
추론통계학: 수집한 데이터를 기반으로 전체 정보를 추론하고 결론을 도출하는 방법을 다루는 학문
통계적 가설검정과 신뢰구간 등이 여기에 속함

확률분포: 확률변수가 특정한 값을 가질 확률을 나타내는 함수
이산확률분포: 확률 변수가 이산적인 값을 가지는 경우 (유한하거나 셀 수 있는 경우, 특정값 선택이 가능)
    1) 베르누이분포
    2) 이항분포
    3) 기하분포
    4) 음이항분포
    5) 초기하분포
    6) 포아송분포
연속확률분포: 확률 변수가 연속적인 값을 가지는 경우 (무한한 범위를 갖고, 특정값 선택이 불가능)
    1) 정규분포
    2) T분포/F분포
    3) 감마분포
    4) 지수분포
    5) 카이제곱분포
    6) 균등분포

* 정규분포 (표본크기가 30 이상인 경우 보통 사용)
종 모양이며, 평균을 기준으로 대칭, 표준편차가 클수록 곡선이 평평해짐 (즉, 넓게 퍼져있음)
모집단의 평균과 분산을 정확히 알고 있을 때 사용됨
중심극한정리에 의해 n(표본의크기)>=30이면 표본평균의 분포는 정규분포에 근사하게 됨
대수의법칙: 표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상

* T 분포 (표본크기가 30 미만인 경우 보통 사용)
종 모양이며, 정규분포보다 꼬리가 두꺼움, 평균을 기준으로 대칭이며 자유도가 커지면 표준정규분포(0,1)에 수렴
표본으로 모평균을 추정할 때 사용되는 분포이며, 표본 크기가 작을 때 정규분포보다 정확한 결과를 제공
표본의 분산을 사용하여 모집단 분산을 추정하여 만들기 때문에 표본의 크기가 작을수록 불확실성이 커짐
-> 그리고 T 분포는 이런 극단적인 값이 나올 가능성을 반영한 것
    * 자유도: 통계에서 얼마나 많은 정보가 독립적으로 변할 수 있는가를 나타내는 숫자
    Ex. 4, 6, 8이 있을 때, 이 중 2개의 숫자를 알고, 평균을 안다면 나머지 숫자는 자동으로 결정됨.
    즉, 1개의 숫자는 자유롭게 변할 수 없지만, 2개의 숫자가 자유롭게 변할 수 있다는 의미
    결국, 자유도가 클수록 (더 많은 독립적 정보를 사용할수록) 정규분포에 가까워지게 됨
    그러나, 너무 많은 자유도를 가지면 overfitting 문제가 발생할 수 있음

* 카이제곱분포
0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포
자유도가 증가하면 중심이 오른쪽으로 이동하고 대칭적으로 변화하여 정규분포에 가까워짐
독립적인 표준 정규 분포를 따르는 변수들의 제곱합을 의미함
주로 독립성을 검정하거나, 적합도를 검정할 때 사용됨
```

```python
4. 추론통계 - 검정
가설검정: 모집단의 실제 값이 얼마가 된다는 주장과 관련하여 표본의 정보를 사용하여 가설의 합당성 여부를 판단하는 과정
가설 설정 -> 유의 수준 설정 -> 검정통계량 산출 -> 가설 기각/채택
1) 가설 설정
귀무가설 H0과 대립가설 H1을 설정함
귀무가설은 주로 기각하고자 하는 명제
대립가설은 내가 주장하고자 하는 명제
2) 유의수준 설정
결과를 판단할 때 허용할 수 있는 오류의 최대치를 의미. 보통 0.05 (5%)로 설정함
즉, 귀무가설이 실제로 참인데도 5%의 확률로 귀무가설을 잘못 기각할 수 있다를 의미함
-> 1종, 2종오류
    1종 오류: 귀무가설이 참이었는데 기각할 확률
    2종 오류: 귀무가설이 거짓인데 기각하지 못할 확률
가설검정은 1종 오류를 더 치명적인 것으로 파악하여 유의수준은 1종 오류의 확률을 의미함
3) 검정통계량 산출
데이터의 분포나 특징에 따라 T값, F값 등 어떤 통계량을 사용할지 선정함
    i) Z값 
    표준 정규분포를 사용하여 평균으로부터 얼마나 떨어져있는지를 나타내는 값
    Z = (X - 평균)/표준편차
    크기가 30이상인 가설 검정에 주로 사용됨
    Z = 1.96일때 유의수준 0.05의 임계점
    Z = 2.58일때 유의수준 0.01의 임계

    ii) T값
    T 분포를 사용하여 얻은 값으로 작은 표본 크기에 대한 가설 검정에 주로 사용
    꼬리 부분이 더 두꺼워 작은 표본에서 발생하는 불확실성을 보다 잘 반영함
    (표본 평균-연구자가 설정하는 값)/(표본 표준편차*표본 크기의 제곱근)
    주로 모집단의 평균에 대한 가설을 검정하거나, 표본 평균 간의 차이를 검정할 때 사용함
    T 값의 절댓값이 크면 모집단 평균으로부터 멀리 떨어져있고, 모집단 평ㅇ균과 유의미한 차이가 있을 가능성이 높다는 것을 나타냄
    T 분포표 사용함

    iii) F값
    F분포를 사용하여 얻은 값으로, 두 표본 분산 사이의 차이를 비교하는 데 사용함
    두 개의 독립적인 카이제곱 분포된 변수들의 비율로 정의함
    F = 분산1/분산2 (분산1과 분산2는 두 독립적인 표본의 분산)
    주로 분산 분석에서 두 개 이상의 그룹 간 평균의 차이를 검정하는데 사용
    값이 1에 가까우면 두 표본의 분산이 서로 비슷하여 퍼진 정도가 비슷함을 의미
4) 가설 기각/채택
유의확률이 유의수준보다 작을 시에 귀무가설을 기각
유의 확률이 유의수준보다 클 시에는 귀무 가설을 기각하지 못함
```

```python
4-2. 데이터의 자료형에 따른 검정 방법
1) [수치형-수치형] (피어슨)상관분석
점으로 데이터를 찍었을 때 바로 상관관계가 나오는 경우는 매우 드물고, 이 경우에 상관분석을 이용
가설설정
H0: 두 변수간 선형 상관관계가 존재하지 않는다
Ha: 두 변수간 선형 상관관계가 존재한다
코드예
import scipy.stats as spst
spst.pearsonr(df['col1'], df['col2'])
상관계수를 r 이라고 할 때,
|r|<0.3 -> 두 변수간 상관관계가 약함
|r|<0.5 -> 두 변수간 상관관계가 중간
|r|>0.5 -> 두 변수간 상관관계가 강

선형 관계만 파악할 수 있으며, 직선의 기울기 또한 고려하지 않음
스피어만 상관계수라는 것이 존재하며, 이는 데이터를 순위화해서 피어슨 상관계수를 구하는 것을 의미

2) [범주형-수치형] F-test(ANOVA), T-test
- T-test: 범주가 2개일 때 (t값은 두 평균 간 차이를 표준 오차로 나눈 값)
가설설정
H0: 두 그룹의 평균에는 차이가 없다.
Ha: 두 그룹의 평균에는 차이가 있다.
( ** 만약 두 그룹의 평균엥는 차이가 없다를 주장하고 싶다면, 위처럼 가설을 설정하고
    기각하지 못한다는 식으로 주장해야 함)
코드예시
mincho = df.loc[df['민초여부'] == 'Yes', '수명']
no_mincho = df.loc[df['민초여부'] == 'No', '수명' ]

spst.ttest_ind(mincho, no_mincho)
결과는 t값, pvalue=p값, df=자유도의 형태로 나타남
t값은 절댓값 2를 기준으로, 그보다 크면 차이가 있으며, 클수록 차이가 크다는 것을 의미
p값은 이 결과가 얼마나 믿음직한지를 나타내며 보통 0.05를 기준
** ttest_ind는 두 범주가 독립적인 경우에 나타냄 (서로 독립적이지 않은 경우에는 ttest_rel을 사용)
** 단측검정은 두 그룹 간 차이의 '한 쪽'만 보고 싶을 때 사용하는 방법
    H0: 치킨을 매일 먹는 사람이 아닌 사람의 수명과 작거나 같을 것이다.
    Ha: 치킨을 매일 먹는 사람의 수명이 아닌 사람의 수명보다 길 것이다.
    -> t분포는 좌우대칭형이므로, 코드는 같게 하고, p값을 2로 나누면 단측 검정의 결과가 나타남

- F-test(ANOVA): 범주가 3개 이상일 때 (단일요인, 이중요인, 반복측정, 혼합설계, 공분산분석, 다변량분산분석 등 여러 분석이 있음)
F 값은 집단 간 분산/ 집단 내 분산으로 계산함
즉, (전체 평균과 각 집단의 평균을 가지고 만든 분산) / (각 집단의 평균과 개별값으로 만든 분산)으로 계산
가설설정
H0: 각 MBTI 그룹 별 수명 평균은 차이가 없다.
Ha: 각 MBTI 그룹 별 수명 평균은 차이가 있다.

코드예시
ISTP = MBTI.loc[MBTI['mbti'] == 'ISTP', '수명']
ESFP = MBTI.loc[MBTI['mbti'] == 'ESFP', '수명']
INFP = MBTI.loc[MBTI['mbti'] == 'INFP', '수명']

spst.f_oneway(ISTP, ESFP, INFP)
결과는 f값, p값으로 나타남
f값은 보통 2~3이상이면 차이가 있으며 클수록 차이고 크다는 것을 의미
p값은 f값 결과가 얼마나 유의미한지를 나타냄
** anova 검정의 경우, 전체 평균 대비 각 그룹간 차이가 있는지만 확인하기 때문에 어느 집단에서 한 값이
튀고 나머지가 비슷해도 귀무가설을 기각하는 결과가 나올 수 없다
-> 그래서 사후분석을 진행함

3) [범주형-범주형] Chi2-test
가설설정
H0: 경영학과 여부와 갈색머리 여부는 관련이 없다
Ha: 경영학과 여부와 갈색머리 여부는 관련이 있다
기대빈도 -> 위의 예시를 바탕으로, 경영학과 비율*갈색머리 비율 처럼 각 그룹의 확률을 곱하는 것
관측빈도 -> 실제 관측된 값
** 같은 위치에 있는 기대빈도와 관측빈도의 차이가 크다면 독립적이지 않을 가능성이 높음

계산 - 카이제곱분석을 사용함!
각 위치 별로 (관측빈도-기대빈도)^2/기대빈도를 하고 모두 더하면 됨!
카이제곱 통계의 경우 범주 개수에 제한이 없기 때문에 범주가 많을수록 카이제곱 통계량이 커질 것
따라서 카이제곱의 통계량 해석은 '자유도의 약 2배'가량을 기준으로 잡고, 2배 이상이면 차이가 존재한다고 해석
(이때, 범주형 변수의 자유도는 범주의수-1이 됨
따라서, 경영학과와 갈색머리 예시에서는 (2-1)*(2-1)=1이 됨)
(만약, 학과 10개, 머리색 7가지의 예시였다면 (10-1)*(7-1)=54가 되고,
이때 카이제곱 통계량이 약 108을 넘어야 의미있는 차이가 있다고 판단할 수 있음)

코드예시
table = pd.crosstab(df['col_행'], df['col_열']) #일단 교차표를 작성해야 하므로 crosstab 사용
spst.chi2_contingency(table)
카이제곱통계량, p값, 자유도, 기대빈도 순서대로 값들이 나타남

4) [수치형-범주형]
수치형이 변화할 때 범주형의 비율이 어떻게 변화하는지 sns.kdeplot을 통해 파악할 수 있음
sns.kdeplot(x='col_수치',data-df, hue='col_범주')
그러나, 수치형을 구간별로 잘라 범주화하지 않는 이상 검정 방법이 존재하지 않는 문제
이때, 범주형을 target, 수치형을 feature로 두고 모델을 만든 후에, 모델 계수를 유의성 검정하고, p값을 보고 판단하는 방법을 사용할 수 있음
```
