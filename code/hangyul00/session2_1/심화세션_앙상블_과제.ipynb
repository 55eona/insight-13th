### 1. Voting

`VotingClassifier`

[Docs](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.VotingClassifier.html)

```python
# estimators 인자로 들어갈 분류기 리스트를 지정합니다.
models = [
    ('lr', LogisticRegression(random_state=SEED)), 
    ('dt', DecisionTreeClassifier(random_state=SEED)),
    ('rf', RandomForestClassifier(random_state=SEED))
]
```

### 4. Stacking

`StackingClassifier`

- **estimators**: 기본 모델 목록 (이름, 모델) 쌍의 리스트
- **final_estimator**: 최종 분류기 (기본값: LogisticRegression)
- **cv**: 교차 검증 폴드 수 (기본값: 5)
- **passthrough**: 원본 특성을 메타 모델에 전달 여부 (기본값: False)
- **n_jobs**: 병렬 처리 수

[Docs](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.StackingClassifier.html)

```python
base_learners = [
    ('lr', LogisticRegression(random_state=SEED)),
    ('dt', DecisionTreeClassifier(random_state=SEED)),
    ('rf', RandomForestClassifier(random_state=SEED))
]

stack_clf = StackingClassifier(
    estimators=base_learners,
    final_estimator=LogisticRegression(random_state=SEED), # 최종 분류기를 LogisticRegression(random_state=SEED) 로 지정해주세요
    cv=5
)

# 모델 훈련
stack_clf.fit(X_train, y_train)

# 테스트 데이터에 대한 예측 수행
y_stk_pred = stack_clf.predict(X_test)

# 정확도 출력
print(f"Stacking Accuracy: {accuracy_score(y_test, y_stk_pred):.4f}")
```

## 2. 피처엔지니어링

- 범주형 변수인 Education, City, Gender, EverBenched에 대한 Label 인코딩이나 원핫인코딩을 진행할 수 있음.
- 현재 연도에서 입사 년도를 뺀 '재직 기간'에 대한 파생 변수를 생성할 수 있음
- 급여가 높은 경우인 PaymentTier의 값이 1인 경우에 대해 고급여 여부를 통해 급여 여부에 따른 이직율을 분석할 수 있음
- 또한, EverBenched는 Yes/No로 구분되어 있으므로, 이를 1/0으로 변환할 수 있음
  

```python
from sklearn.preprocessing import LabelEncoder

# ✅ 1. 범주형 변수 인코딩 (Label Encoding)
label_cols = ['Education', 'City', 'Gender']
for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))

# ✅ 2. 파생 변수 생성
current_year = 2025
df['YearsAtCompany'] = current_year - df['JoiningYear']                   # 조직 내 재직 기간
df['IsHighTier'] = (df['PaymentTier'] == 1).astype(int)                  # 고급여 여부

# ✅ 3. 이진 변수 처리
df['EverBenched'] = df['EverBenched'].map({'Yes': 1, 'No': 0})
```

## 3. 앙상블 기법 활용

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import (
    BaggingClassifier, VotingClassifier, RandomForestClassifier,
    AdaBoostClassifier, GradientBoostingClassifier
)
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import pandas as pd

# 🎯 Feature/Target 설정
X = df.drop(columns=['LeaveOrNot'])
y = df['LeaveOrNot']

# ✅ Stratified K-Fold 설정
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {}

# ✅ 1. Bagging
bagging = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42)
results['Bagging'] = cross_val_score(bagging, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 2. Hard Voting
hard_voting = VotingClassifier(estimators=[
    ('lr', LogisticRegression(max_iter=1000)),
    ('dt', DecisionTreeClassifier()),
    ('nb', GaussianNB())
], voting='hard')
results['Hard Voting'] = cross_val_score(hard_voting, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 3. Soft Voting
soft_voting = VotingClassifier(estimators=[
    ('lr', LogisticRegression(max_iter=1000, solver='liblinear')),
    ('rf', RandomForestClassifier(n_estimators=100)),
    ('gb', GradientBoostingClassifier())
], voting='soft')
results['Soft Voting'] = cross_val_score(soft_voting, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 4. Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
results['Random Forest'] = cross_val_score(rf, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 5. AdaBoost
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
results['AdaBoost'] = cross_val_score(ada, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 6. Gradient Boosting (GBM)
gbm = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
results['Gradient Boosting'] = cross_val_score(gbm, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 7. XGBoost
xgb = XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss', random_state=42)
results['XGBoost'] = cross_val_score(xgb, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 8. CatBoost
catboost = CatBoostClassifier(iterations=100, verbose=0, random_state=42)
results['CatBoost'] = cross_val_score(catboost, X, y, cv=cv, scoring='accuracy').mean()

# ✅ 결과 정리
results_df = pd.DataFrame(list(results.items()), columns=['Model', 'Mean CV Accuracy'])
results_df.sort_values(by='Mean CV Accuracy', ascending=False, inplace=True)
print(results_df)
```
