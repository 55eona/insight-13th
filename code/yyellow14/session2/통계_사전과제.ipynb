# __통계__
---

### __통계란__
- 양적 자료 vs 질적 자료
- 명목척도: 단순 범주만 나타내는 척도
- 서열척도: 순위 또는 서열이 있는 척도로 수량화할 수 있는 자료
- 구간(등간)척도: 연속적인 수로 수량화 할 수 있으며, 크고 작은 정도를 표현할 ㅜ 있음음
- 비율척도: 연속적인 수로 수량화 가능, 절대적인 원점 존재재
---



### __(수치) 기술 통계__
##### 기술 통계란
- 기술 통계학: 측정이나 실험을 통해 수집한 자료를 정보로 전환하기 위하여 요약하고 정리하여 그 특성을 기술하는 절차에 관한 개념
- 수치 기술 통계: 하나의 수치, 즉 기술통계량으로 데이터를 요약 & 설명하는 방법
- 시각화 기술 통계: 표와 그래프의 형태로 데이터를 시각화하는 방법
##### 중심 경향성
- 평균: `series.mean()`
- 중앙값: `series.median()`
- 최빈값: `series.mode()`
- 왜도: `series.skew()`
##### 산포도
- 분산: `np.var(arr)`
- 표준편차: `np.std(arr)`
- 범위: `np.max(arr) - np.min(arr)`
- 사분위수: `np.percentile(arr, 25); np.percentile(arr, 50); np.percentile(arr, 75)`
- 사분위간 범위: Q3에서 Q1 값을 제외한 범위, 즉 데이터의 중앙 50%가 퍼져 있는 범위
- 변동계수: `np.std(arr) / np.mean(arr) * 100`

### __추론 통계 - 추정__
##### 추론 통계란?
- 추론 통계학: 모집단에서 일부를 뽑아 만든 표본 집단을 통해 모집단에 대한 정보를 추측하는 과정
##### 확률분포
- 확률분포, 확률변수
- 이산확률분포의 종류: 베르누이분포, 이항분포, 기하분포, 음이항분포, 초기하분포, 포아송분포
- 연속확률분포의 종류: 정규분포, T분포/F분포, 감마분포, 지수분포, 카이제곱분포, 균등분포포
---

### __추론통계 - 검정__
##### 가설 검정의 절차
- 가설 설정: 귀무가설 H0와 대립가설 H1를 설정
- 유의 수준 설정: 유의수준이란 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치
- 검정통계량 산출: 검정통계량은 가설 검정의 결과를 판단하는 데 사용되는 값. 데이터의 분포나 특징에 따라 T값, F값 등 어떤 통계량을 사용할지 정해야한다.
- 가설 기각/채택 판단: 유의확률(p-value)은 검정통계량에서 유도할 수 있으며 유의 확률(p)가 유의 수준(α)보다 작을 시에 귀무가설을 기각, 클 시에는 귀무 가설을 기각하지 못한다.
##### 데이터의 자료형에 따른 검정 방법법
- [수치형 - 수치형] (피어슨) 상관분석
- [범주형 - 수치형] F-test(ANOVA), T-test
- [범주형 - 범주형] Chi2-test
