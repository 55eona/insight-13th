### __회귀와 모델링__
##### 회귀란
- 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것. 연속적인 숫자를 다룸.
##### 회귀의 종류
- 선형회귀, 비선형회귀, 릿지회귀, 라쏘회귀, 다항회귀 등
##### 선형회귀
- 선형회귀: 데이터를 가장 잘 설명하는 회귀선을 찾아 예측하는 과정
- 단순선형회귀: 데이터를 하나의 독립 변수로 표현하는 경우. y=ax+b
- 최소제곱법: 잔차의 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식
- 경사하강법: Loss function, learning rate, local minima 문제, momentum
- 다중선형회귀분석: 독립 변수가 2개 이상인 경우의 선형회귀
- 다중공선성: 회귀 분석에서 독립 변수들 간에 상관관계가 큰 경우 발생. 회귀분석의 정확도가 낮아짐.
##### 규제선형모델
- 규제선형모델: 모델이 학습 데이터에 과적합 되지 않도록 규제를 가하고자 등장한 모델.
- 릿지 회귀(L2 규제): W(회귀계수)의 제곱에 대해 페널티를 부여하는 방식
- 라쏘 회귀(L1 규제): W의 절댓값에 대해 페널티를 부여하는 방식
- 엘라스틱넷(L1 + L2): 릿지의 계수 축소 기능과 라쏘의 변수 선택 기능 모두 가짐.
##### 모델평가방법
- 성능 평가 지표
- 평균 제곱 오차: Mean Squared Error
- 평균 절대 오차: Mean Abolute Error
- 변수 유의성 평가: t검정
