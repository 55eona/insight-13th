# 머신러닝과 모델링
## 1. 머신러닝
머신러닝 = 새로운 데이터를 **예측**하거나 **결정**을 내릴 수 있도록 하는 기술. 패턴을 찾아내고, 패턴을 바탕으로 새로운 데이터에 대한 예측을 가능하게 함. 
- 비지도학습 : 정답이 없는 데이터로 학습하는 방식
- 지도학습 : 정답이 있는 데이터로 학습하는 방식

## 2. 모델링
전체 데이터 -> 훈련/테스트 데이터로 분리 -> 훈련 데이터로 모델 만들기, 테스트셋으로 반복 성능을 평가 (반복) -> 최종 모델 생성
- 훈련 데이터 : 모델을 학습/훈련시킬 떄 쓰이는 데이터
- 테스트 데이터 : 학습된 모델을 검증하기 위해 사용되는 데이터. unseen data에 대해 예측하는 모델을 가지고 모델의 성능을 평가할 수 있다. 

# 회귀와 모델링
## 1. 회귀란?
- 회귀 = **패턴**을 찾아내어 미래 값을 **예측**하는 것
- 연속적인 숫자를 다룰 때 회귀를 이용한다. (<-> 목표 변수가 불연속적일 때는 분류를 이용)

## 2. 회귀의 종류
선형회귀, 비선형 회귀, 릿지회귀, 라쏘회귀, 다항회귀 등
다양한 회귀 기법이 존재한다. 

## 3. 선형회귀
= 데이터를 가장 잘 설명하는 회귀선 (y = ax + b 등)을 찾아 예측한다. 
회귀식에서 y = 종속변수, x = 독립변수
- 선형 회귀 기법을 사용하려면 특정 조건을 만족해야 함 : 선형성, 독립성, 등분산성, 정규성

### (1) 단순선형회귀분석
- 독립변수가 <하나>만 있음.
- 데이터를 가장 잘 설명하는 직선 y = ax + b를 찾는다.
- 즉, 데이터를 통해 적절한 회귀계수 a와 절편 b를 찾는 것이 목표이다.

#### 최소제곱법 
- a, b를 잘 찾는 법: 오차를 정의해나가기 = 최소제곱볍
- 최소제곱법 = 최적의 회귀선을 찾는 방식. 잔차의 제곱의 합 (RSS)이 최소가 되는 지점으로 최적의 회귀선을 구한다. (*잔차 = y_hat - y. 유도된 회귀식의 y값과 실제 데이터 포인트가 가진 y값의 차.)
#### 경사하강법
- 최소제곱법을 구하기 어려울 때에는 경사하강법(GD)를 활용한다.
- 목적함수 : 달성하고자 하는 목표. 오차를 최소화한, 최적화된 함수. 모델의 예측이 얼마나 잘 맞는지 평가하는 기준
- 손실함수 : **개별 데이터 샘플**에 대한 오차를 측정하는 함수. ex. 회귀 분석에서의 mse, mae / 분류 문제에서의 교차 엔트로피
- 비용함수 : **전체 데이터셋**에서 평균적인 손실을 측정하는 함수. 모든 데이터 포인트에 대한 손실함수 계산 -> 평균 or 합산
- 모델 학습 시에는 이 <비용함수>를 최소화하는것을 목표로 한다.

  - 미분계수가 양수인 경우 W는 왼쪽으로 감소 / 음수인 경우 W가 오른쪽으로 증가. => 미분계수가 0인 점에 수렴/도달하도록 한다.
  - **학습률** : 미분계수가 음/양수인걸 확인한 뒤, "얼마나 이동하는가?", 보폭의 크기.
  - 학습률이 너무 작다면 모델이 수렴할 때까지 너무 오랜 시간이 걸림. 너무 크다면 모델이 수렴하지 못하고 발산하여 손실이 커짐.
  - 학습률은 개발자가 직접 설정. 포기에 작은 값을 시도하다가, 실험을 통해 조정한다. 혹은, learning rate scheduler/ grid search, random search를 활용해 자동으로 찾을 수도 있다.
  - local minima 문제 : 함수가 진짜 최솟값을 가지는 global minima에 도달하지 못하고, 중간의 local minima에서 빠져나오지 못하는 문제.
  - 모멘텀 : 이전의 기울기/방향을 고려하려 "관성"을 부여. 작은 언덕은 빠르게 넘어 global minima를 찾을 수 있도록 한다. 

### (2) 다중선형회귀분석
= 독립 변수가 2개 이상일 경우의 선형회귀 -> N가지의 독립변수를 가짐. 
- 3차원 공간 + 평면 형태의 회귀선을 가진다. 최소제곱법/GD 사용 똑같이 가능.
- (1) 다중공선성 : 두 독립 변수가 강한 상관관계를 가질 때, **어떤 IV가 DV에 얼마나 영향을 미치는지 정확하게 구분하기 어려워진다**. 어떤 변수의 영향을 반영해야 할지 분확실해지므로 회귀분석의 정확도가 떨어진다.
    - 다중공선성 확인 방법 : 상관계수 확인, vif 지수
    - 다중공선성 대처 방법 : 변수 제거, 변수 변환 (아내수입, 남편수입 -> 가족수입으로 병향), 규제 선형 모델 활용 (릿지, 라쏘, 엘라스틱넷 등), PCA 주성분분석 (정보를 보존하면서 차원을 축소, 복잡성 줄이기)

## 4. 규제선형모델
= 모델이 학습데이터에 과적합 (overfitting, train set을 외워 새로운 데이터에서 정확성 떨어짐) 되지 않도록 규제를 가하고자 등장한 모델
- 특히 다중회귀에서는 독립 변수가 많으므로 과적합 가능성이 높다.
- 기존 선형 모델에서는 RSS를 이용해 최소화함. but 회귀 계수가 커지게 되어(특정 IV에 과하게 의존하게 됨) 과적합이 발생할 수 있음.

### 규제선형모델의 종류
- L2규제 (릿지회귀)
    - 회귀계수의 **제곱**에 대해 페널티를 부여함
    - 기둥 크기에 따라 다른 힘으로 누르는 망치 - 회귀계수가 크면 세게, 작으면 살살 눌러서 전체적으로 줄임
- L1규제 (라쏘회귀)
    - 회귀계수의 **절댓값**에 페널티를 부여
    - 모든 기둥을 같은 힘으로 누르는 망치 - 회귀계수가 작으면 0이 되기도 함 = 변수가 사라질 수 있음.
    - 작은 변수가 사라지면 중요한 변수만 남게 된다. 
- L1 + L2 규제 (엘라스틱넷 회귀)
    - 불필요한 변수 제거 + 남은 변수들은 적절한 크기로 유지
    - 변수 제거를 통한 희소성 + 변수가 균형있게 줄어들어 안정적인 예측 동시에 가능하다.

## 5. 모델 평가방법
내가 만든 회귀 모델이 좋은 성능을 내고 있는가? 올바르게 만든 회귀 모델인가? 에 대한 평가
### (1) 성능 평가 지표
- 평균 제곱 오차 (mse) : 모델 예측값-실제 관측값 사이의 오차 제곱의 평균. (제곱이 있어 큰 오차, 이상치에 민감하게 반응)
- 평균 절대 오차 (mae) : 절대값 오차의 평균 (이상치가 존재해도 적절한 예측 성능 유지 가능)
- 데이터와 모델의 특성, 평가 목적에 따라 지표를 선택할 수 있다.
### (2) 변수 유의성 평가
- t검정 : 독립 변수가 종속 변수에 유의미한 영향을 주는지 판단
- H_0 : 회귀계수(a)가 0이다 = 독립변수가 종속변수에 영향을 주지 않는다. p-value가 0.05보다 이상일 때 귀무가설 채택.
- H_1 : 회귀계수(a)가 0이 아니다 = 독립변수가 종속변수에 영향을 준다. p-value가 0.05보다 작을 때 대립가설 채택.
