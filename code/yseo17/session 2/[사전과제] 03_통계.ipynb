# 1. 통계란
- 정확한 통계를 구하는 것 = 모집단 조사이지만 시간과 비용이 많이 들어감. 따라서 모집단의 샘플 (=표본)을 추출하여 분석한다.
- 기술통계 : 수치, 시각화를 통해 샘플을 요약
- 추론통계 : 추정, 검정을 통해 샘플로부터 모집단을 추론 (회귀, 분류, 머신러닝을 통한 예측)
- 양적 자료 - 구간척도, 비율척도
- 질적 자료 - 명목척도, 서열척도(순위)

# 2. (수치) 기술 통계
## 2.1. 기술통계란
- 데이터의 주요 특성을 요약하고 설명한다
- 수치 기술 통계 : 하나의 수치 (기술통계량) 으로 데이터를 요약 및 설명
- 시각화 기술 통계 : 표/그래프 형태로 데이터를 시각화
## 2.2. 중심경향성
- 평균 : 데이터 값의 총합/데이터 개수, .mean()
- 중앙값 : 데이터를 순서대로 나열했을 때 가운데에 위치한 값, .median()
- 최빈값 : 가장 자주 나타나는 값, .mode()
- 중심경향성을 통해 왜도를 알 수 있음. series.skew()
## 2.3. 산포도
- 분산 : 데이터 값들이 평균으로부터 멀리 퍼진 정도.
- 표준편차 : 분산에 제곱급을 취한 값. (제곱이 된)단위를 일치시키기 위해 표준편차를 사용한다.
- 범위 : 데이터의 전체 변동범위. 최대-최소
- 사분위수 : 데이터의 표본을 네 개의 동일한 크기로 나눈 값. np.percentile(arr, 25)
- 사분위간 범위 : Q3 - Q1, 중앙의 50%가 퍼져 있는 범위. Q1 - 1.5*IQR, Q3 + 1.5*IQR 바깥의 범위를 이상치로 판단한다.
- 변동계수 : 데이터의 상대적인 변동성. 표준편차를 평균으로 나누어 100을 곱한 값. 

# 3. 추론 통계 - 추정
## 3.1. 추론통계란
- 수집된 데이터를 기반으로 미래의/조사하지 못한 전체 정보를 추론하고 결론을 도출하는 것.
## 3.2. 확률분포
- 확률분포 : 확률변수가 특정한 값을 가질 확률
- 확률변수 : 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수
- 이산확률분포 : 확률 변수가 이산적인 값을 가지는 경우. 베르누이분포, 이항분포 등
- 연속확률분포 : 확률 변수가 연속적인 값을 가지는 경우. 정규분포, T/F분포 등
## 3.3. 대표적인 분포 모형
- 정규분포 : 종모양. 매개변수 = 평균, 표준편차.
- T분포 : 종모양. 표본 크기가 작을 때 정규분포보다 더 정확하다. 매개변수 = 자유도(df)
- 카이제곱분포 : 독립성 검정, 적합도 검정 등에 사용된다. 매개변수 = 자유도 (자유도가 증가할수록 중심이 오른쪽으로 이동, 대칭적으로 변화한다)

# 4. 추론통계
## 4.1. 가설 검정의 절차
[가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단]
1. 가설 설정 : 귀무가설 (기각하고자 하는 명제, "-는 차이가 없다")과 대립가설 (내가 주장하고자 하는 명체 "-는 차이가 존재한다")
2. 유의 수준 설정 : 결과를 판단할 때 허용할 수 있는 오류의 최대치. 유의수준이 0.05 = 귀무가설이 참인데도 불구하고 잘못 기각할 확률이 5%. 1종 오류 (참인데 기각)와 2종 오류 (거짓인데 기각 x)
3. 검정통계량 산출 : z값 (표준 정규분포에서 평균으로부터 떨어진 정도), T값 (T분포로 얻은 값), F값 (F분포로 얻은 값. 두 표본 분산 사이의 차이를 비교)
4. 가설 기각/채택 판단 : 유의확률이 유의 수준보다 작으면 귀무가설을 기각할 수 있다.
## 4.2. 데이터의 자료형에 따른 검정 방법
- [수치-수치형] 피어슨 상관분석 : 공분산, 상관계수. -1.0 ~ 1.0 사이. scipy.stats.pearsonr()
- [범주-수치형] F-test(ANOVA, 범주 3개 이상), T-test (범주가 두개). scipy.stats.ttest_ind()
- [범주-범주형] Chi2-test : 기대빈도와 실제 관측값을 비교. 차이가 크다면 독립적이지 않음. pd.crosstab()
- [수치-범주형] : 수치형을 구간화하여 범주화하지 않는 이상 마땅히 없음.
