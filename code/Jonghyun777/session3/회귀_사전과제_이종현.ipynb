### **회귀와 모델링**

회귀란?

> - 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것
> - 집값 예측, 주식 가격 예측 등과 같이 연속적인 숫자를 다룸

선형회귀

!

> - 선형 회귀는 데이터를 가장 잘 설명하는 회귀선(ex. 직선 y=ax+b)을 찾아 예측하는 과정
<BR>선형 회귀는 기존 관측된 데이터포인트를 기준으로 **그 데이터를 가장 잘 설명하는 선**을 찾아 **미래를 예측**.
<BR>이러한 회귀 선은 **현재 데이터를 요약**하는 동시에, 아직 관측되지 않은 **미래 값을 예측하는 도구**가 됨
> - 예측하고자 하는 것=종속변수
<BR> 예측을 위해 사용하는 변수=독립변수

선형 회귀의 종류
> - 단순 선형 회귀 : 데이터를 직선 형태(하나의 독립 변수)로 표현하는 경우
> - 다중 선형 회귀 : 데이터를 두 개 이상의 독립 변수를 사용하여 표현하는 경우 

단순선형회귀분석

!\

> - **하나의 독립변수만 존재하는 선형 회귀이다.**
> -  데이터를 가장 잘 설명하는 **직선 y=aX+b** 을 찾는 과정에 해당.
> -  y=aX+b에서 **a는 회귀계수(기울기)** **b는 y절편**을 의미.
> - 적절한 **회귀계수(a)**와 **절편(b)**를 찾는 과정

**결론적으로 우리는 데이터를 잘 설명할 수 있는 최적의 직선 y=ax+b를 찾는 것임**
<BR> How? 최소제곱법으로

### 최소제곱법

> - 잔차의 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식
> - 잔차: 실과 예측값의 차이를 의미
>  - 핵심: 회귀선과 실제 데이터와의 차이를 최소화하기 위한 시도!

: 최소제곱법과 같은 목적! 다만 계산 과정이 복잡할 때 최소로 가는 과정에서 경사하강법 사용

✔︎ 목적함수

`우리가 달성하고자 하는 목표`

> - 최적화를 위해 오차를 최소화하는 것이 목적인 함수
> - 최적화하려는 대상을 수학적으로 표현한 함수

✔︎ 손실함수

`개별 데이터 샘플에 대한 오차를 측정하는 함수`

> - 한 개의 데이터 포인트에 대해 모델이 얼마나 틀렸는지를 평가
> - 예측값 $\hat{y}$와 실제값 $y$의 차이를 계산
<BR>**회귀 분석** → 평균제곱오차(MSE), 평균절대오차(MAE)
    $\text{(MSE for one sample)} = (y - \hat{y})^2$
<BR>**분류 문제** → 교차 엔트로피(Cross-Entropy Loss)

✔︎ 비용함수

`전체 데이터셋에서 평균적인 손실을 측정하는 함수`

> - 손실함수를 모든 데이터 포인트에 대해 계산한 후 평균 or 합산한 값
> - 머신러닝에서는 대부분 모델을 훈련할 때 비용함수를 최적화(최소화)하는 것이 목표

`머신러닝 과정에서는 모델을 평가하고 최적화하는 함수 찾는게 핵심! `

`<BR>그런 의미에서 엄밀하게는 손실함수와 비용함수는 개별, 합계로서 차이가 있으나 혼용하는 경우가 많음`

**손실함수 사용**

손실 함수 그래프 위에 **파란 점**을 각각 찍으면 그래프의 최솟값은 **검정 점**이고, **파란 점**에서의 미분값은 그래프의 접선의 기울기가 됨

- 만약 현재 값이 최솟값의 왼쪽에 있다면 기울기(미분 값)가 음수이기 때문에 "현재 값 - 미분 값(음수)"를 하면 현재 값이 커지. 즉, **오른쪽으로 점이 이동**
- 만약 현재 값이 최솟값의 오른쪽에 있다면 기울기(미분 값)가 양수이기 때문에 "*현재 값 - 미분 값(양수)"를 하면 현재 값이 작아짐. 즉, **왼쪽으로 점이 이동**
- 그러다가 최솟값에 도달하면 미분값이 **0**이 되어서 "현재 값 - 0"으로 더이상 변화하지 않고 멈추게 됨.

최솟값 왼쪽의 값은 점점 오른쪽으로 이동하고, 최솟값 오른쪽의 값은 왼쪽으로 이동하며 최종적으로 최솟값에 도달하게 되는 원리!

### **다중회귀분석**

: 독립변수가 2개 이상인 경우 선형회귀 -> N가지의 독립변수

> - 독립 변수가 2개인 경우 3차원 공간에 표현되고 회귀선은 평명이 됨
> - 단순회귀와 마찬가지로 최소제곱법 사용 가능.
> - 관측치와 평명간의 차이가 잔차가 됨.(단순회귀랑 똑같이 파악이 힘들 때 경사하강법 사용)
> - 변수가 2개보다 많은 경우에는 비슷한 느낌으로 확장

1. 다중공선성

`Ex) 주택 가격을 예측하는 모델에서 '주택의 크기'와 '방안 개수'라는 두 독립 변수가 강한 상관관계를 가질 때, 
<BR> 이 두 변수 간에 다중공선성이 발생 `

So, 다중공선성은 회귀 분석에서 독립 변수들 간에 상관관계가 큰 경우 발생
<BR>->  다중공선성이 높은 경우 어떤 독립 변수가 종속 변수에 얼마나 영향을 미치는지를 
<BR>정확하게 구분하기 어렵기 때문에 회귀 모델은 어떤 변수의 영향을 반영해야 할지 불확실해지고, 
<BR>그 결과 회귀분석의 정확도가 낮아짐.

***다중공선성 확인 방법***

1. 상관계수

- 상관계수는 변수간의  통계적 관계 즉. **상관관계의 정도를 수치**로 나타낸 것.
    - 상관계수 값 r은 **-1과 1**사이에 존재
    - r이 -1과 1에 가까울수록 두 변수의 상관성 높음
    - r이 음수라면 음의 상관관계(한쪽이 커지면 한쪽은 작아짐)양수라면 양의 상관관계(한쪽이 커지면 같이 커짐)를 뜻한다.
    - 0은 선형 상관 관계없음

+Pairplot과 같은 시각화로 산점도를 찍어 상관성 확인 가능

2. VIF 지수(분산 팽창 인수)

> - 회귀 모델의 결정계수 R square를 사용하여 계산.
> - VIF가 높으면 다중공선성이 존재한다고 판단
 - **VIF = 1**: 해당 독립 변수는 다른 변수와 상관관계가 전혀 없음을 의미
- **VIF < 5**: 일반적으로 다중공선성 문제가 없다고 간주
- **VIF > 5**: 다중공선성의 징후가 있을 수 있으며, 주의를 요함
- **VIF > 10**: 다중공선성이 심각하다고 간주되며, 이 변수를 포함한 회귀 모델은 불안정할 가능성이 큼. 이 경우 해당 변수를 제거하거나, 변수 간의 상관관계를 줄이기 위한 조치가 필요할수도


***다중공선성 대처 방법***

2. 변수 변환
3. 구제 선형 모델 활용

### **규제선형모델**

> - 모델이 학습 데이터에 과적합 되지 않도록 규제를 가하고자 등장한 모델
> - 여러 독립 변수를 사용하여 예측하므로 모델이 복잡해지는 다중회귀에서 과적합 확률 증가
> - 기존 선형 모델에서는 RSS(잔차 제곱합)를 최소화(최소제곱법)하는 방식으로 비용함수설정

규제선형모델의 종류

1. L2규제
2. L1규제
3. L2규제 + L1규제 결합

1. L2 규제(릿지 회귀)


: 기둥 크기에 따라 다른 힘으로 누르는 망치로 설명 가능
<BR> 큰 기둥은 세게 누르고 작은 기둥은 살살 눌러서, 비율이 균등하지 않을 수 있지만 전체적으로 작아짐
> - 회귀계수의 제곱을 패널티로 주기 때문에, 큰 회귀계수일 수록 강한 패널티 받음
> - 모든 기둥(계수)의 크기는 작아지지만, 완전히 사라지지는 않음(회귀계수가 0이 되지 않음) -> 변수 제거x
> - 계수 크기 조절 기능: 모든 변수를 유지하면서도, 크기를 적절히 줄여 과적합 방지

2. L1 규제( 라쏘 회귀)

:모든 기둥을 같은 힘으로 누르는 망치
<BR> 작은 기둥들은 힘을 버티지 못하고 완전히 사라질수도

> - 회귀계수의 절댓값을 패널티로 주기 때문에, 회귀계수의 크기와 상관없이 같은 패널티 받음
> - 작은 기둥(작은 계수)은 힘을 이기지 못해 0이 되기도 함 → 일부 기둥이 완전히 사라짐(변수 선택 기능)
> - 변수 선택 기능: 비중이 낮은 변수(작은계수)는 아예 사라지고, 중요한 변수만 남음

3. L1 + L2 엘라스틱넷

: L1,L2 두개의 망치를 동시에 사용
<BR> 작은 기둥을 아예 부수는 L1 망치와 남은 기둥을 균형 잡히게 조절하는 L2 망치를 동시에 사용하여
<BR> 불필요한 변수는 제거하고, 남은 변수들은 적절한 크기로 유지

> - L1의 변수 선택 기능(일부 계수를 0으로 만듦)과 L2의 규제 기능(큰 계수를 줄임)을 동시에 활용하는 모델
> - L1은 상관관계가 높은 변수 중 일부만 선택하는 경향이 있음
> - 안정적인 예측과 희소성을 동시에 달성 가능

### **모델 평가 방법**

`내가 만든 회귀 모델의 좋서능을 평가하는 법`

1. 성능 평가 지표

> - 평균 제곱 오차 MSE:
<BR> - MSE는 Mean Squared Error의 약어
<BR> - 회귀 분석에서 모델의 예측값과 실제 관측값 사이의 오차 제곱의 평균을 의미
<BR> - 모델의 예측 성능을 평가하는 지표 중 하나입니다.
> - 평균 절대 오차 MAE:
<BR> - MAE는 Mean Absolute Error의 약어
<BR> - 회귀 분석에서 모델의 예측값과 실제 관측값 사이의 절대값 오차의 평균을 의미
<BR> - 모델의 예측 성능을 평가하는 지표 중 하나로 사용


2. 변수 유의성 평가

**t 검정**

: 독립 변순의 회귀계수가 유의미한지 검정하는데 사용
<BR>귀무 가설:회귀계수(a)가 0이다 → 독립변수가 종속변수에 영향을 주지 않는다.
<BR>대립가설: 회귀계수가 0이 아니다 -> 독립변수가 종속변수에 영향을 준다. 



**2. 검정 과정**

1️) t-값 계산

$$
t = \frac{\hat{\beta} - 0}{SE(\hat{\beta})}
$$

- $\hat{\beta}$ : 추정된 회귀계수
- $SE(\hat{\beta})$ : 표준오차
- t-분포를 따름.



2️) **p-value 확인 및 판단**

- **p-value < 0.05** → 유의 수준 5%에서 $H_0$ 기각 → **해당 변수는 유의미하다.**
- **p-value ≥ 0.05** → $H_0$ 기각 불가 → **해당 변수는 유의미하지 않다.**




3. 해석

- **회귀계수가 0이면 해당 독립변수는 의미 없음.**
- **p-value가 작을수록** 해당 변수는 종속 변수에 **더 큰 영향을 준다고 볼 수 있음.**
- **t-값이 크면** 귀무가설을 기각할 가능성이 높아짐.
