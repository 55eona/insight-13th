**1.지도학습과 비지도 학습의 차이에 대해 설명해주세요.**

지도학습은 문제와 정답을 모두 알려주는 학습 방식이다. 따서서 예측값을 이미 만들어둔 정답과 같아지도록 기계를 학습시킨다. 대표적인 예로는 회귀와 분류가 있다.

비지도 학습은 지도학습과는 반대로 답을 가르쳐주지 않고 학습시키는데, 데이터 속의 패턴이나 데이터 간의 유사도를 기계가 직접 학습하도록 한다. 군집화가 비지도 학습의 대표적인 예시이다. 

**2.회귀와 분류의 차이에 대해 설명해주세요**

회귀는 데이터가 연속형 변수를 예측할 때 사용하며 분류는 데이터가 범주형 변수를 예측할 때 사용된다.
따라서 예측하고자 하는 값에 따라 회귀와 분류를 적절히 사용하면 된다. 예를 들어 음식 배달에 대한 데이터 세트로 음식 배달시간을 예측할 때는 연속형 변수를 다뤄야 하기 떄문에 회귀를 사용하고, 반대로 건강 지표 데이터 세트로 당뇨병 여부 에측을 할 때에는 분류를 사용한다. 

**3.분류 모델의 네 가지 종류와, 각 모델이 무엇인지 간단하게 정리해주세요.**

> - 결정 나무 : 조건에 따라 데이터를 분류하며, 최종적으로 데이터가 순수한 label의 집합으로 구성될 때까지 분류를 반복하는 모델. 스무고개 놀이 방식으로 이해하면 쉬움!
> - 서포트 벡터 머신:클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘.고차원 공간에서도 효과적으로 사용 가능
> - KNN: 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘. 비슷한 특성을 가진 데이터끼리 서로 가까이 있다는 점을 이용함

#### 2. 분류 모델

 [로지스틱 회귀]

(2) 결정 나무

: 조건에 따라 데이터를 분류하며, 최종적으로 데이터가 순수한 label의 집합으로 구성될 때까지 분류를 반복하는 모델.(스무고개 놀이와 비슷한 방식으로 이해)

- **Root Node**: Decision Tree의 시작이 되는 노드
- **Edge**: 노드와 노드를 연결하는 길목
- **Leaf Nodes**: Tree의 가장 마지막 노드로, 모델에서 label에 해당
- **Height(depth)**: Tree의 깊이로, 클수록 tree의 구조는 복잡해짐
- **Level**: 노드의 절대적 위계, Root node의 level = 0, leaf node의 level = height - 1
- **Parent**: 상대적으로 높은 위계의 노드
- **Child**: 상대적으로 낮은 위계의 노드
    - **Binary Tree**(이진 트리): Tree 중에서 children이 최대 2개인 tree

[CART(classification and regression tree) 알고리즘 ]

가장 대표적인 결정 나무 알고리즘, 데이터셋을 임계값을 기준으로 두 child로 나누는 알고리즘

임계값은 분순도(지니계수)가 낮아지는 방향으로 나눠야 됨!

불순도?

: 분류하는 데이터 집합에서 서로 다른 클래스(범주)가 섞여있는 정도를 의미. CART 알고리즘에서는 불순도를 확인하기 위해 "지니 계수"를 사용

CART 알고리즘의 주요단계
> - 1. 임계값 설정
>   2. 불순도 감소 알고리즘: 동일한 객체들끼리 모아지게 하기 위해 불순도 알고리즘을 사용하여 다른 객체들이 섞여있는지 확인하고 불순도가 낮은쪽으로 가지를 형성

[실제 학습 시 고려해야 할 것들: 모수 설정, 차이점 시각화, Prunning]

1. **Parameter 설정**
    
    
    | `min_samples_split` | 분할되기 위해 노드가 가져야 하는 최소 샘플(데이터) 수 |
    | --- | --- |
    | `min_samples_leaf` | 리프 노드가 가지고 있어야 하는 최소 샘플 수 |
    | `min_weight_fraction_leaf` | `min_samples_leaf`와 같지만 가중치가 부여된 전체 샘플 수에서의 비율 |
    | `max_leaf_nodes` | 리프 노드의 최대 개수 |
    | `max_features` | 각 노드에서 분할에 사용할 특성의 최대 수 |

2. **시각화**

시각화를 수행함으로써 분류가 잘 이루어졌는지 확인 가능

3. **Prunning(가지치기)**
   

불필요한 노드 지우기

(3) 서포트 백터 머신 (SVM)

: 클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘

명확하게 분류할 수 있는 데이터 집단에서 성능이 뛰어나며, 고차원 공간(다수의 feature)에서도 효과적

[SVM의 구성]

> - support vector:구분하는 선과 가장 가까운 포인트
> - decision boundary(결정 경계): 집단을 구분하는 선
> - Margin: 선과 각 점의 거리

최적의 선을 찾는 방법은 Margin거리가 가장 큰 경우를 선택하는 것임. 

(4)KNN (K-Nearest neighbor)

: 비슷한 특성을 가진 데이터끼리 서로 가까이 있는다는 점을 이용한 분류 알고리즘.유유상종
<br> 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류함

계산순서
> - 1. 데이터준비: knn은 미리 학습하는 과정 x, 각 데이터는 특징 벡터와 레이블로 이루어짐
>   2. K값설정: 원하는 K 값을 설정합니다. K는 가장 가까운 이웃의 개수를 나타내며, 보통 홀수개로 설정
>   3. 거리계산: 기존 모든 데이터 간의 거리를 계산 
>   4. 가장 가까운 K개의 이웃 선택: 가장 작은 거리 값을 가진 k개의 데이터 선택
>   5. 분류하기: k개의 이웃 중 가장 많이 등장하는 클래스가 예측 결과가 됨
>

    - 정보 손실 없음
