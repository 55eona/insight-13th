### **통계**

##### **통계란?**

***왜 통계를 공부해야 할까***


> - 통계학의 본질은 데이터로부터 의미를 찾는 것
> - 데이터 분석의 최종 목적은 미래 예측
> - 통계학적 지식을 활용하여 데이터 분석

***데이터의 척도***

> - 통계학에서 데이터란 관찰과 조사를 통해 얻는 수치적 또는 비수치적 자료
> - 숫자로 표현 가능 데이터=양적자료  ,    숫자로 표현 불가=질적
> - 질적자료: 명목척도,서열 척도        양적자료: 구간척도,비율척도

명목척도: **단순 범주만** 나타내느 척도. 사칙연산 불가능  ex) 색깔 , 성별

서열척도: **순위 또는 서열**이 있는 척도. 사칙연산 불가능

구간척도: 연속적인 수로 수량화 가능. 숫자 간의 간격이 동일. 덧셈과 뺄셈은 가능 but 곱셈,나눗셈 불가능 
ex) 온도(0도가 온도가 없음을 의미하지 않음), IQ

비율척도: 연속적인 수로 수량화할 수 있으며, **절대적인 원점**이 존재. 모든 사친연산 가능 
ex) 몸무게,인구수,길이

##### **기술통계**

***기술통계란?***

`데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야. 평균,중앙값,표준편차 등의 개념이 이에 해당`

> - 수치 기술 통계: 하나의 수치, 즉 기술 통계량으로 데이터를 요약&설명하는 방법
> - 시각화 기술 통계: 표와 그래프의 형태로 데이터를 시각화하는 방법

***중심 경향성***

: 데이터의 중심적인 경향을 나타내느 기술 통계량

> - 평균
> - 중앙값
> - 최빈값
> - 왜도

***산포도***

: 데이터가 퍼져있는 정도를 설명하는 기술 통계량
> - 분산
> - 표준편차
> - 범위
> - 사분위수
> - 사분위간 범위
> - 변동계수

##### **추론 통계-추정**

***추론통계란?***

추론 통계학: 수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전체 정보를 추론하고 결론을 도출하는방법을 
다루는 분야로, **통계적 가설검정**과 **신뢰구간** 등이 여기에 속함

***확률분포***

> - **확률분포**는 확률변수가 특정한 값을 가질 확률을 나타내느 함수를 의미
> - **확률변수**는 random한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수

` 확률분포는 확률 변수의 속성에 따라 이산확률분포와 연속확률분포로 범주를 나눌 수 있음
이산확률분포는 확률 변수가 이산적인 값을 가지는 경우에 해당 
연속확률분포는 확률 변수가 연속적인 값을 가지는 경우 해당`

***대표적인 분포 모형***

> - 정규분포
> - T분포
> - 카이제곱분포

##### **추론통계-검정**

**가설검정**은 모집단의 실제 값이 얼마나 된다는 주장과 관련해 표본의 정보를 사용해서 가설의 합당성 여부를 판정하는 과정을 의미

***가설 검정의 절차***


➡️ **가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단**



1. 가설 설정 

> - 귀무 가설 H0와 대립가설 H1을 설정
> - 주로 기각하고자 하는 명제로 설정
> - 대립가설은 내가 주장하고자 하는 명제로 설정


2. 유의 수준 설정 

`유의수준이란 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치로, 주로 0.05 (5%)로 설정. 연구자가 어떤 결정이
얼마나 쇤뢰할 만한지를 판단하는 기준이 됩니다.`


***1종,2종 오류***

##### 1종 오류: 귀무가설이 실제로 참이었는데 기각할 확률
##### 2종 오류: 귀무가설이 실제로 거짓인데 기각하지 못할 확률
##### **가설검정은 1종오류를 더 치명적으로 파악. So, 유의수준은 1종 오류의 확률을 의미**

***검정통계량 산출***

검정통계량은 가설 검정의 결과를 판단하는 데 사용되는 값. 데이터의 분포나 특징에 따라 T값,F값 등 어떤 통계량을 사용할 지 정해야 함.

**1. Z-score**

> - 표준 정규 분포(평균이 0이고 표준편찬가 1인 분포)를 사용하여 얻은 값
> - 특정 데이터가 평균으로부터 얼마나 떨어져 있는 지를 표준편차 단위로 나타냄

**2. T-score**

> - T분포를 사용하여 얻은 값으로, 작은 표본 크기에 대한 가설 검정(T검정)에 주로 사용
> - T분포는 **정규 분포와 유사**하지만, 꼬리 부분이 더 두꺼워 작은표본에서 발생할 수 있는 분확실성을 보다 잘 반영

**3. F-Score**

> - F분포를 사용하여 얻은 값으로, 주로 두 표본 분산 사이의 차이를 비교하는 데 사용됨
> - 두 개의 독립적인 카이제곱 분포된 변수들의 비율로 정의

***가설 기각/채택 판단***

**유의 확률(p-value)**

: 가설 검정에 쓰일 수 있는 또 다른 척도. 유의확률은 검정 통계량에서 유도할 수 있으며, 유의확률이 유의수준보다 작을 시 에 귀무가설을 기각, 클 시에는 귀무 가설을 기각하지 못합니다. 

***데이터의 자료형에 따른 검정 방법***

**피어슨 상관 분석**

: 두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정하는 방법. 

가설 
H0: 두 변수간 선형 상관관계가 존재하지 않는다.
Ha: 두 변수간 선형 상관관계가 존재한다.

**공분산**

: 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현하는데 사용

> - Cov (X,Y) > 0이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
> - Cov (X,Y) < 0이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
> - Cov (X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계

**상관계수**

 ##### 두 변수 간의 선형적인 관계를 나타내는 통계적 지표. 
 상관계수는 -1과 1 사이의 값을 가지며, 상관관계와 인과관계는 다르다는 점을 주의해야함

> - Corr (X,Y)이 양수면 양의 상관관계 (1이면 완전한 양의 상관관계)
> - Corr (X,Y)이 음수면 음의 상관관계 (-1이면 완전한 음의 상관관계)
