# 통계학 요약 정리

## 1. 통계의 개념

- **통계학**: 데이터를 수집, 분석하여 패턴과 관계를 찾아내고 결론을 도출하는 학문
- **목적**: 데이터로부터 의미를 찾아 미래를 예측하는 것
- **기본 분류**:
  - **기술통계**: 샘플 자체를 수치와 시각화로 요약하는 방법
  - **추론통계**: 샘플로부터 모집단을 추론하는 방법

### 데이터의 척도

- **양적 자료**: 숫자로 표현 가능한 데이터
  - **구간척도**: 간격이 동일하고 크기 비교 가능 (덧셈, 뺄셈만 가능) - 예: 온도, IQ
  - **비율척도**: 절대적 영점이 존재하고 모든 사칙연산 가능 - 예: 몸무게, 길이

- **질적 자료**: 숫자로 표현할 수 없는 데이터
  - **명목척도**: 단순 범주만 나타냄 - 예: 색깔, 성별
  - **서열척도**: 순위나 서열이 있음 - 예: 순위, 만족도

## 2. 기술 통계

### 중심 경향성
- **평균(mean)**: 데이터 총합 ÷ 개수 (이상치에 민감)
- **중앙값(median)**: 데이터를 정렬했을 때 가운데 값 (이상치에 덜 민감)
- **최빈값(mode)**: 가장 빈번하게 나타나는 값
- **왜도(skewness)**: 분포의 비대칭성을 나타내는 지표

### 산포도
- **분산(variance)**: 데이터가 평균으로부터 퍼진 정도
- **표준편차(standard deviation)**: 분산의 제곱근 (데이터와 같은 단위)
- **범위(range)**: 최대값 - 최소값
- **사분위수(quartile)**: 데이터를 4등분하는 값들
  - Q1: 하위 25% 지점
  - Q2: 중앙값
  - Q3: 하위 75% 지점
- **사분위간 범위(IQR)**: Q3 - Q1 (이상치 탐지에 유용)
- **변동계수**: (표준편차/평균) × 100 (상대적 변동성 비교)

## 3. 추론 통계 - 추정

### 확률분포 유형
- **이산확률분포**: 확률변수가 셀 수 있는 값을 가질 때
- **연속확률분포**: 확률변수가 연속적인 값을 가질 때

### 주요 분포
- **정규분포**:
  - 종 모양의 대칭 분포
  - 매개변수: 평균(μ), 표준편차(σ)
  - 중심극한정리: 표본 크기가 커지면 표본평균의 분포는 정규분포에 근사

- **T분포**:
  - 정규분포보다 꼬리가 두꺼운 종 모양
  - 매개변수: 자유도(df)
  - 표본 크기가 작을 때(n<30) 모평균 추정에 사용

- **카이제곱분포**:
  - 비대칭 분포, 자유도가 증가할수록 대칭에 가까워짐
  - 독립성 검정, 적합도 검정 등에 사용

### 중요 원리
- **대수의 법칙**: 표본 크기가 커질수록 표본평균이 모평균에 근사
- **중심극한정리**: 표본 크기가 클 때 표본평균의 분포는 정규분포에 근사

## 4. 추론 통계 - 검정

### 가설 검정의 정의
가설 검정은 모집단의 실제 값에 대한 주장을 표본 정보를 사용하여 판정하는 과정입니다.

### 가설 검정의 절차

#### 1. 가설 설정
- **귀무가설(H0)**: 기각하고자 하는 명제, 보통 "차이가 없다"는 주장
- **대립가설(H1)**: 주장하고자 하는 명제, 보통 "차이가 존재한다"는 주장

#### 2. 유의 수준 설정
- 허용할 수 있는 오류의 최대치, 주로 0.05(5%)로 설정
- **1종 오류(α)**: 귀무가설이 참인데 기각할 확률
- **2종 오류(β)**: 귀무가설이 거짓인데 기각하지 못할 확률

#### 3. 검정통계량 산출
- **Z값**: 대표본(n≥30)에서 사용, Z = (X-μ)/σ
  - 유의수준 0.05에서 임계값 = 1.96
  - 유의수준 0.01에서 임계값 = 2.58
- **T값**: 소표본에서 사용, T = (X-μ)/(s/√n)
  - 자유도에 따라 임계값이 달라짐
- **F값**: 분산 비교에 사용, F = 분산1/분산2
  - 주로 분산분석(ANOVA)에 활용

#### 4. 가설 기각/채택 판단
- **유의확률(p-value)**: p-value가 유의수준(α)보다 작으면 귀무가설 기각
- 검정통계량이 임계값보다 크면 귀무가설 기각

### 검정 예시
새로운 수학 교육 프로그램 효과 검증:
1. **가설 설정**
   - H0: 새 프로그램과 기존 프로그램의 평균 점수 차이 없음
   - H1: 새 프로그램의 평균 점수가 더 높음
2. **유의수준**: α=0.05 설정
3. **데이터 수집 및 검정통계량 계산**: t=2.05
4. **p-값 계산**: p=0.02
5. **결론**: p(0.02) < α(0.05)이므로 귀무가설 기각, 새 프로그램이 더 효과적이라 결론


### 데이터 자료형에 따른 검정 방법

#### 1. 수치형-수치형: 피어슨 상관분석

- **목적**: 두 변수 간 선형 상관관계 검정
- **가설**:
  - H0: 두 변수 간 선형 상관관계가 존재하지 않는다
  - Ha: 두 변수 간 선형 상관관계가 존재한다
- **공분산과 상관계수**:
  - 공분산: 두 변수의 함께 변화하는 방향과 크기 표현
  - 상관계수: -1과 1 사이의 값으로 표준화된 공분산
- **해석 가이드**:
  - |r| < 0.3: 약한 상관관계
  - |r| < 0.5: 중간 상관관계
  - |r| > 0.5: 강한 상관관계
- **코드**: `scipy.stats.pearsonr(df['col1'], df['col2'])`

### 2. 범주형-수치형: T-test와 ANOVA(F-test)

#### T-test (범주가 2개일 때)
- **목적**: 두 그룹의 평균 차이 검정
- **가설**:
  - H0: 두 그룹의 평균에는 차이가 없다
  - Ha: 두 그룹의 평균에는 차이가 있다
- **코드**: `scipy.stats.ttest_ind(group1, group2)`
- **해석**: t값이 절대값 2 이상이면 차이가 유의미함

#### F-test/ANOVA (범주가 3개 이상일 때)
- **목적**: 여러 그룹 간 평균 차이 검정
- **가설**:
  - H0: 각 그룹별 평균은 차이가 없다
  - Ha: 각 그룹별 평균은 차이가 있다
- **코드**: `scipy.stats.f_oneway(group1, group2, group3, ...)`
- **해석**: F값이 2-3 이상이면 차이가 유의미함
- **주의**: 어떤 그룹이 다른지는 사후분석(Tukey HSD 등) 필요

### 3. 범주형-범주형: Chi2-test

- **목적**: 두 범주형 변수 간 관련성 검정
- **가설**:
  - H0: 두 범주형 변수는 관련이 없다(독립적)
  - Ha: 두 범주형 변수는 관련이 있다
- **방법**: 기대빈도와 관측빈도의 차이를 분석
- **카이제곱 통계량**: Σ((관측빈도-기대빈도)²/기대빈도)
- **해석**: 카이제곱 통계량이 자유도의 약 2배 이상이면 유의미
- **코드**:
  ```python
  table = pd.crosstab(df['col1'], df['col2'])
  scipy.stats.chi2_contingency(table)
  ```

### 4. 수치형-범주형

- 명확한 검정 방법이 없음
- 범주형을 target, 수치형을 feature로 모델링하여 간접적으로 관계 파악 가능
- 수치형을 범주화한 후 chi2-test 적용 가능
