### 기술통계 - 평균, 중앙값, 표준편차
수치 기술 통계: 하나의 수치. 기술통계량 ~ 중심 경향성, 산포도
시각화 기술 통계: 표와 그래프의 형태로 데이터를 시각화 ~ EDA

* 중심 경향성 - 데이터의 중심적인 경향

평균 : series.mean()
  - 총합치 사용, 이상치에 민감할 수 있음 
중앙값 : series.median()
  - 평균과 달리 이상치에 덜 민감
최빈값series.mode()
  - 주로 이산형 데이터에서 사용. 데이터의 분포를 이해하는데 도움
왜도(skewness): 자료의 개형 확인 시 중심 경향성 X, 데이터 분포의 비대칭성을 나타내는 통계적 지표
  - positive Skewness(Right-skewed), negative skewness(Left-skewed), Normal distribution
  - series.skew
  
* 산포도 - 데이터가 퍼져 있는 정도
분산: np.var(arr) ~ 데이터가 평균으로부터 얼마나 퍼져있는가?
표준편차: np.std(arr) ~ 분산의 제곱근. 분산 단위: 관측값 단위의 제곱 -> 표준편차 단위 일치
범위: np.max(arr) - np.min(arr) ~ 데이터의 최대값과 최소값의 차이를 나타냄
사분위수(quartile): np.percentile(arr, 25/50/75) ~ 데이터 표본을 4개로 나눔
IQR(사분위간 범위)~ Q1 ~ Q3 중앙 50%가 퍼져있는 범위 
   - 아래쪽 이상치: Q1 - 1.5 * IQR보다 작은 값
   - 위쪽 이상치: Q3 + 1.5 * IQR 보다 큰 값
변동계수(coefficient of variation): np.std(arr) / np.mean(arr) * 100 
~ 데이터의 상대적인 변동성. 표준편차 / 평균... 표준편차(단위) / 평균(단위) ~> 집단의 퍼진 정도를 동일한 기준으로 비교할 수 있다.

### 정규분포
종 모양, 대칭,매개 변수(평균(μ), 표준편차(σ)) ~ 많은 자연현상과 사회 현상
표본이 커질수록 정규분포에 근사 n>= 30. 대수의 법칙 & 중심극한정리

### T 분포
정규분포보다 꼬리가 두꺼운 대칭, 종 모양. 매개변수(자유도(df))
표본 크기가 작을 때, 정규분포보다 더 정확한 결과 제시 ~ 표본으로 모평균 추정 시 사용

*** 자유도: 얼마나 많은 정보가 독립적으로 변할 수 있는가?
=> if. 4, 6, 8이 있다. 두 개의 숫자와 평균을 알면 자동으로 하나의 수가 결정 => 자유도는 2. 
너무 많은 자유도 => 오버피팅: 모델이 학습 데이터에 너무 맞춰져 새로운 데이터에 대한 일반화 성능이 저하되는 현상

### 카이제곱분포
0에서 시작해서 오른쪽으로 긴 꼬리를 가진 비대칭 분포. 매개변수(자유도(df))
자유도가 증가함에 따라 중심이 오른쪽으로 이동. 더 대칭. 정규분포에 가까워짐 ~ 독립적인 표준 정규 분포를 따르는 변수들의 제곱합. 독립성/적합도 검정 드에 사용

## 추론 통계 - 검정 
가설 검정: 모집단의 실제 값이 얼마다를 표본으로 추정하는데 가설의 합당성 여부 판정
가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택

* 가설 설정: 귀무가설 H0(기각하고자 하는 명제, 현재 통용) vs 대립 가설 H1(내가 주장하고자 하는 명제)

* 유의 수준: 연구자가 결과를 판단할 대 허용할 수 있는 오류의 최대치 ~ 주로 5%. 
" 귀무 가설이 실제로 참인데도 불구하고, 5%의 확률로 귀무가설을 잘못 기각할 수 있다 "
~ 1종 오류 / 2종 오류... 1종 오류: 실제로 참인데 기각 / 2종: 잘못인데 기각 X. => 1종 오류 [유의수준]

* 검정 통계량
$Z=(X−μ)/σ$ ~ Z = 1.96시 유의수준 0.05, 신뢰 수준 95% / Z = 2.58일 때 유의 수준 0.01, 신뢰 99%

$T = \frac{X - \mu}{s / \sqrt{n}}$ ~ 모집단의 평균에 대한 가설. 두 표본 평균 간의 차이 검정 시
T값의 절대값이 크면, 모집단 평균으로부터 멀리 떨어짐. 관측 표본 평균이 모집단 평균과 유의미한 차이가 있을 가능성이 높다.... 
~ T 분포표 사용 ~ 자유도가 높아질수록 z값에 수렴

F값=분산_1/분산_2 ~ 두 표본 분산 사이의 차이 비교, 두 독립적인 카이제곱 분포 변수들의 비율로 정의
ANOVA 분석(분산 분석)에서 두 개 이상의 그룹 간 평균의 차이 검정 시 사용
1에 가까우면 두 표본의 분산이 서로 비슷. 

* p value: 가설 검정 시 사용. 검정통계량으로 유도된 유의 확률이 유의수준보다 작으면 귀무가설 기각. 

### 데이터의 자료형에 따른 검정 방법.
* 피어슨 상관분석: 수치형 - 수치형
음... 이게 무슨 관계지? 싶은 것들 -> 상관분석 이용
두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정. 
H0: 두 변수간 선형 상관관계가 존재하지 않는다.
Ha: 두 변수간 선형 상관관계가 존재한다.

* 공분산: 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기(정도) ~ y = x + c
np.cov(df['col1'], df['col2'])

* 상관계수: 두 변수 간의 선형적인 관계를 나타내는 지표. -1 ~ 1 사이. 수치적으로 연관된 관계
공분산 / (각 변수의 표준편차) 
df[['col1', 'col2']].corr()

import scipy.stats as spst
	spst.pearsonr(df['col1'], df['col2'])

* 범주형 - 수치형
1) 시각화 ~ barplot 이용 시 신뢰구간도 볼 수 있음.
신뢰구간: 95% 확률로 신뢰구간 안에 모평균이 포함됨

2) T-test: 범주가 2개일때 ~ 두 그룹의 평균에는 차이 있음을 검증...
t값: 두 평균 간 차이 / 표준편차 => 절댓값 2를 기준으로 클수록 두 그룹의 평균 차이가 크다... 

3) F-test(ANOVA): 범주가 3개 이상. ~ n 그룹의 평균에 차이가 있음을 검증 Ex) MBTI별 수명
F값: 집단 간 분산 / 집단 내 분산 ~ 전체 평균 - 집단 전체 분산 / 집단 평균 - 개별 분산
ANOVA 종류 많음.
