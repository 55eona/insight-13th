회귀 / 단순, 다중 선형회귀분석 / 최소제곱법
선형 회귀
- 프로세스: 사전 검증(데이터 탐색, 처리) -> 모델 생성 및 모델 fit  / 유의성 검정(F-검정, T-검정) -> 모델 평가 (R-Squared) -> 모델 성능 개선(필요 없는 변수 제거, 비선형 모델 활용)
- 다중선형회귀: 관심있는 y(종속변수)에 대해 영향을 줄 것 같은 여러 설명변수(X)를 채택하여 선형적으로 식을 작성
- 종속 변수 y에 대한 설명변수 X의 가중평균
- 가중치: y를 얼마나 잘 설명하는가?

* 선형회귀 기본 가정
    - 선형성: 종속/설명 변수 간의 관계가 선형적
      $E(y|X) = \beta_1+\beta_2X_2+...+\beta_nX_n$
      scattor plot 찍어보기
      
    - 독립성: 각각의 설명변수가 서로 선형독립 ~ 다중공선성 X
      VIF 지수, 상관계수
  
    - 오차항의 평균은 0이다.(0에 가까워야 한다) $E(\varepsilon_i|X_i) = 0$
      백색잡음(특정 패턴 없이 랜덤하게 변하는 신호)
      np.mean(residuals)
  
    - 등분산성: 오차항의 분산이 일정해야 한다.
      일부 구간에서 오차 작고 일부 구간에서는 오차가 큰 것을 방지한다.
      잔차의 도표화
      
     - 오차항은 자기상관되어 있지 않다.
      $Cov(\varepsilon_i, \varepsilon_j|X_i) =0$ ~ 공분상이 0
      Durbin-Watson 검정

    - 정규성: 오차항이 정규분포를 따른다. ~ p-value에 대해 유의미한 해석을 하기 위해서... 다중선형회귀분석의 결과에 큰 영향 X
      Shapiro-Wilk 검증, Q-Q plot

      회귀 분석 전 검증 가능: 선형성 / 독립성 / 오차항 자기상관 없다.
      회기 분석 후 검증 가능: 6가지 가정 모두
 
* 회귀 분석 평가 방법.
    - 시각화 / 통계지표
    - 모델의 유의성 검정(F-검정, T-검정)
      F-검정: 전체 모델을 검정 대상으로 회귀 모델이 유의미한지 검정한다. p < 0.05이면 회귀모델이 유의미하다.
      * F statistics, Prob(F-statistic): 회귀 모델이 유의미한지 검정하는 지표
      * F-statistic이 클수록 통계적으로 유의미함을 의미. Prob(F-sta~): 0.05보다 작다면, 귀무가설을 기각할 수 있다.
      * 귀무가설 기각: 적어도 하나의 독립변수가 종속 변수와 유의미한 관계를 갖고 있다.
      
      T-검정: 개별 변수를 대상으로 특정 독립 변수가 유의미한지 검정한다. p < 0.05이면 회귀모델이 유의미하다.
      * T-statistic: 각 독립 변수가 종속 변수 y에 유의미한 영향을 미치는지 판단.
      * 해당 변수의 회귀계수가 0이다.(귀무가설) ~> 종속변수에 유의미한 영향을 준다. (귀무가설 기각)    
   
    - 모델의 성능 평가(R-squared, Adjusted R-squared, AIC, BIC)
      * 통계지표 R square(결정계수)
        데이터의 총 변동 중에서 설명된 비율 -> 데이터의 평균을 예측값으로 사용할 때보다, 모델을 사용했을 때 얼마나 더 정확한 예측인가
        평균과의 차이
        1) 회귀선을 지나가는 데이터 ~ 회귀선에 의해 설명되는 편차(SSR) 
        2) 회귀선을 벗어나는 데이터 ~ 회귀선에 의해 설명되는 편차(SSR) + 회귀선에 의해 설명되지 않는 편차(SSE)
        ** SST = SSR + SSE
           R^2 = SSR / SST = 1 - SSE / SST
           모델이 얼마나 데이터를 잘 설명하는지 측정 (0,1) 범위.

      * Adjusted R-squared
        R^2는 독립 변수의 개수가 증가할수록 자연스럽게 증가... => 변수의 개수 증가에 덜 민감하도록 조정(샘플&독립 변수 개수를 추가)

      * AIC, BIC
        Akaike information criterion
        Schwarz Criterion(SC), Bayesian Information Criterion
        => 값이 낮을수록 좋은 모델로 평가. 
        

### 비선형 회귀
다항식 회귀모델
* 다항식 회귀 모델: $Y=a+bX+cX^2+dX^3+\cdot\cdot\cdot$ 
    - 데이터의 곡선적 특성을 모델링
    - 곡선적인 관계의 데이터 패턴에서 유용(성장률, 감소율 등 시간에 따라 변화하는 속도가 다른 경우)
    장점: 선형 모델에 비해 fit 한 회귀선
    단점: 너무 많은 feature 이용 시 결과가 overfitting될 수 있음 ~ 릿지, 라쏘 규제 이용 가능

* 지수 회귀모델: 종속변수가 지수적으로 변하는 관계 모델링 시: Y = ae^{bX}
    - 로그를 취하면 ln Y=ln a+bX
      => 로그를 취한 후 선형 회귀를 적용. 원래 데이터 형태로 복구하기 위해 예측 값에 지수함수 적용
    - 기술 발전, 투자 수익률 증가 등
    
* 로그 회귀모델: 종속변수와 독립 변수간의 관계가 로그 함수를 통해 더 잘 표현시: Y = a + blog(X)
    - 독립 변수 X에 로그함수 적용
    - 선형 회귀를 적용 ~> 예측값에 별도의 처리 X
    - 초기에 빠르게 증가... 점차 증가율 낮아지는 case ~ 인구 성장, 감염병의 확산
      
* 스플라인 회귀
    - 데이터를 구간별로 나누고 구간마다 선형 / 비선형 함수 적용하여 예측
    - 구간 경계에서의 연속성을 유지하는 것이 중요
    - 데이터 패턴이 구간마다 다르게 나타날 때 ~ 계절에 따른 판매량 변화


### 회귀 분석 프로세스
#### 사전 검증 -> 모델 생성 및 모델 fit -> 모델 평가 -> 모델 성능 개선
사전 검증: EDA / 회귀 분석의 기본 가정 6가지 검토
모델 생성 및 fit: 회귀 모델 생성, 모델 설정, 모델 학습 / 유의성 검정
모델 평가: R-Squared / 회귀 분석 후 검증 가능한 가설 검토
모델 성능 개선: 필요없는 변수 제거 / 비선형 모델 활용
