# 1. 통계란
### 1.1 들어가긴 전에
정확한 수치를 구하기 위한 확실한 방법은 ==> **모집단**을 조사하는 것

but 현실적인 이유로, 실제로는 **표본집단**만을 추출하여 분석하게 됨

- 표본집단 자체에 대한 관심 ==> 기술통계
- 표본집단이 속한 모집단에 대한 관심 ==> 추론통계

### 1.2 왜 통계를 공부해야 하나요
통계학은 데이터의 수집, 분석을 통해 패턴, 관계를 찾아내는 학문, 방법론

데이터를 통해 의미있는 정보를 얻기 위해서는 통계학적 지식이 필요하기 때문

### 1.3 데이터의 척도
#### 질적척도(Qualitative scale)
  - 명목척도(nominal)
    - 단순범주(이름)만 나타냄
    - 사칙연산 불가
    - EX) 색깔, 성별
  - 서열/순위척도(ordinal)
    - 순위 또는 서열이 있음
    - 얼마큼 차이가 나는지 알 수 없음
    - 사칙연산 불가
    - EX) 달리기 순위, 학년, 만족도   

#### 양적척도(Quantitative scale)
  - 구간/등간척도(interval)
    - 연속적인 수로 수량화 가능
    - 크고 작은 정도 표현 가능
    - 절대적인 원점(0) 존재 / 곱셈과 나눗셈 불가
    - EX) 온도, IQ
  - 비율척도(ratio)
    - 연속적인 수로 수량화 가능
    - 절대적인 원점(0)이 존재 / 사칙연산 가능
    - EX) 몸무게, 인구수, 길

# 2. (수치) 기술통계

### 2.1 기술통계란
**기술 통계학**
: 데이터의 주요특성을 요약하고 설명하는 방법을 다루는 분야

### 2.2 중심경향성
데이터의 중심적인 경향을 설명하는 기술 통계량
#### 평균
  - (데이터 값의 총합)/(데이터의 개수)
  - 이상치에 민감
  - ```series.mean()```
#### 중앙값
  - 데이터 값들을 순서대로 나열했을 때, 중앙에 위치한 값
  - 이상치에 덜 민감
  - ```series.median()```
#### 최빈값
  - 데이터 값들 중 가장 많이 나타난 값
  - 이산형 데이터에서 주로 사용
  - ```series.mode()```
#### +) 왜도
  - Positive Skewness: 오른쪽 꼬리가 길 때
    - 최빈값 < 중앙값 < 평균
  - Negative Skewness: 왼쪽 꼬리가 길 때
    - 평균 < 중앙값 < 최빈값
  - Normal Skeneww: 대칭일 때
    - 평균 = 중앙값 = 최빈값
  - ```series.skew()


### 산포도
데이터가 펴져있는 정도를 설명하는 기술 통계량
#### 분산
  - 데이터 값들이 평균에서 얼마나 멀리 퍼져있는지
  - 각 데이터 값과 평균간의 편차를 제곱한 값들의 평균
  - ```np.var(arr)```
#### 표준편차
  - 분산의 제곱근 -> 단위를 일치시켜줌
  - ```np.std(arr)```
#### 범위
  - (최대값)-(최소값)
  - ```np.max(arr) - np.min(arr)```
#### 사분위수
  - Q1: 하위 25%에 위치한 값, 하위 사분위수
  - Q2: 하위 50%에 위치한 값, 중앙값
  - Q3: 하위 75%(상위 25%)에 위치한 값, 상위 사분위수
  - ```np.percentile(arr, 25/50/75)```
#### 사분위간 범위(IQR)
  - 데이터의 중앙 50%가 퍼져있는 범위
  - 이상치 탐지할 때 사용됨
  - Q3 - Q1
#### 변동계수
  - 데이터의 상대적인 변동성
  - (표존편차/평균) * 100
  - ```np.std(arr) / np.mean(arr)*100```


# 3. 추론통계-추정

### 3.1 추론통계란
- **추론 통계학**
  - 표본집단을 통해 모집단의 정보 추측하는 과정
  - 통계적 가설검정, 신뢰구간 등

### 3.2 확률분포
- **확률분포**: 확률변수가 특정값을 가질 확률
- **확률변수**: 랜덤에서 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 특정 값으로 대응시키는 함수

이산확률 분포|연속확률분포
-------------|------------
확률변수가 이산적인 값을 가질 때 | 확률변수가 연속적인 값을 가질 때
베르누이분포, 이항분포, 기하분포, 음이항분포, 초기하분포, 포아송 분포,   | 정규분포, T분포/F분포, 지수분포, 카이제곱분포, 균등분포



### 3.3 대표적인 분포 모형

#### 정규분포
- 형태
  - 종모양
  - 평균을 기준으로 대칭
  - 표준편차 클 수록 평평
- 매개변수: 평균, 표준편차
- 특징
  - 가장 잘 알려짐
#### T 분포
- 형태
  - 종모양, 정규분포가 꼬리가 두꺼움
  - 평균을 기준으로 대칭
  - 자유도 커질수록, 표준정규분포 N(0,1)에 수렴
- 매개변수: 자유도
- 특징
  - 표본으로 모평균을 추정할 때, 사용
  - 표본크기가 작을 때, 정규분포보다 적합
#### 카이제곱 분포
- 형태
  - 0에서 시작해서, 오른쪽으로 긴꼬리를 가진 비대칭 분포
  - 자유도 커질수록, 정규분포에 가까워짐
- 매개변수: 자유도
- 특징
  - 독립적인 표준정규분포를 따르는 변수들의 제곱합
  - 독립성 검정, 적합도 검정 등에 사용


# 4. 추론통계-검정

### 4.1 가설검정의 절차

#### 4.1.1 가설설정
- **귀무가설 $H_0$**: 기각하고자 하는 명제/현재 통용되는 가설
- **대립가설 $H_1$**: 새롭게 주장하고자 하는 명제

#### 4.1.2 유의 수준 설정

통계적 결정/실제상황| $H_0$ 참|$H_0$ 거짓 
-------|--------|------------
$H_0$ 채택|옳은 결정|2종 오류($\beta$)
$H_0$ 기각|1종 오류($\alpha$)|옳은 결정

**유의수준**: 허용할 수 있는 오류의 최대치 => 유의수준은 1종 오류의 확률을 의미

#### 4.1.3 검정통계량 산출

- **Z값**
  - 표준정규분포를 통해 얻은 값
  - Z값의 절댓값이 크면, 평균으로부터 멀리 떨어져 있음을 의미
  - $Z=1.96$일 때, 유의수준 0.05(=신뢰수준 95%)
  - $Z=2.58$일 때, 유의수준 0.01(=신뢰수준 99%)
- **T값**
  - T 분포를 통해 얻은 값
  - T값의 절댓값이 크면, 모집단 평균으로부터 멀리 떨어져 있음을 의미
- **F값**
  - F 분포를 통해 얻은 값
  - 주로 분산분석(ANOVA)에서 그룹 간 평균 차이 검정할 때 사용
  - F값이 1에 가까우면, 두 표본의 분산이 비슷함을 의미


#### 4.1.4 가설기각/채택판단
유의확률(p-value) < 유의수준($\alpha$) ==> 귀무가설 *기각*

유의확률(p-value) > 유의수준($\alpha$) ==> 귀무가설 *채택*

### 4.2 자료형에 따른 검정방법

#### 4.2.1 [수치형-수치형] (피어슨) 상관분석

- $H_0$: 두 변수간 선형 상관관계가 존재하지 않는다.
- $H_1$: 두 변수간 선형 상관관계가 존재한다.
- ```python
  import scipy.stats as spst
  spst.pearsonr(df['col1'], df['col2'])
  ```
- 결과 형태: (상관계수, p-value)
  - 상관계수: -1 ~ 1 / 선형적인 관계를 나타냄
  - |상관계수| < 0.3 --> 상관관계가 약함
  - |상관계수| < 0.5 --> 상관관계가 중간
  - |상관계수| > 0.5 --> 상관관계가 강함
  


#### 4.2.2 [범주형-수치형] F-test(ANOVA), T-test

- **T-test**: 범주가 2개일 때
  - $H_0$: 두 그룹의 평균에는 차이가 없다.
  - $H_1$: 두 그룹의 평균에는 차이가 있다.
  ```python

  mincho = df.loc[df['민초여부'] == 'Yes', '수명']
  no_mincho = df.loc[df['민초여부'] == 'No', '수명']
  spst.ttest_ind(mincho, no_mincho)

  
  ```
  - 결과 형태: `TtestResult(statistic = t-값, pvalue = p값, df = 자유도)`
    - t값: 2를 기준으로 클 수록, 차이카 크다고 판단
    - p값: 0.05 기준

- **F-test(ANOVA)**: 범주가 3개 이상일 때
  - F값은 집단 간 분산/집단 내 분산 으로 계산
  - one way ANOVA 예시
    - $H_0$: 각 MBTI 그룹 별 수명 평균은 차이가 없다.
    - $H_1$: 각 MBTI 그룹 별 수명 평균은 차이가 있다.
    ```python

    ISTP = MBTI.loc[MBTI['mbti'] == 'ISTP', '수명']
    ESFP = MBTI.loc[MBTI['mbti'] == 'ESFP', '수명']    INFP = MBTI.loc[MBTI['mbti'] == 'INFP', '수명']
    spst.f_oneway(ISTP, ESFP, INFP)


    ```
    - 결과형태: `F_onewayResult(statistic = f값, p-value = p값)`
      - f값: 보통 2~3 이상이면, 차이가 있다고 판단
      - p값: f값 결과가 얼마나 유의한지 판단NFP)
    

#### 4.2.3 [범주형-범주형] Chi2-test
- 카이제곱 통계량 = {(관측빈도)-(기대빈도)}^2 / 기대빈도 의 합
  - 자유도의 2배이상일 때, 의미있는 차이가 있다고 판단
  - 자유도 = 범주 개수 - 1
- ```python
  table = pd.crosstab(df['col1'], df['col2'])
  spst.chi2_contingency(table)
  ```
