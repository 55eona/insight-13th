## 1. 통계란?

가장 정확한 평균 키를 구하는 방법은 당연히 우리나라 성인 전체에 대해 조사하는 것\
하지만 여러가지 이유로 인해 이는 실현이 매우 힘듦.
때문에 보통 모집단에 대해 일부 샘플(*=표본집단*)만 추출하여 분석하게 된다.

샘플을 구했을 때 우리의 관심사는 크게 2가지이다.
1. 내가 갖고 있는 샘플 자체에 대한 관심
2. 이 샘플의 모집단에 대한 관심

여기서 전자는 **기술통계**, 후자는 **추론통계**의 영역에 해당하게 된다.

기술통계란?\
기술통계는 수치와 시각화를 통해 샘플을 요약하여 표현하는 방법론!\
추론통계란?\
추정과 검정을 통해 샘플로부터 모집단을 추론하는 방법론!

*지난 시간의 EDA의 경우 기술통계에 해당하게 되고, 앞으로의 내용은 추론통계의 연장선이라고 볼 수 있다.*

----

통계학은 데이터를 수집 및 분석하여 그로부터 패턴과 관계를 찾아내고 결론을 도출해내는 학문이자 방법론이다.\
통계학의 본질은 **데이터로부터 의미를 찾는 것**이다.\
데이터 분석의 최종 목적은 과거로부터 나온 결과가 다시 반복된다 믿고, **미래를 예측하는 것**이다.\

오늘날 우리는 데이터 자체는 일상에서 쉽게 구할 수 있지만, 그 자체만으로는 의미 있는 정보를 얻을 수 없기 때문에 통계학적 지식을 활용해야 한다.

---

통계학에서 데이터란 관찰과 조사를 통해 얻은 수치적/비수치적 자료
숫자로 표현할 수 있는 데이터는 **양적 자료**(Quantitative, numerical data)\
숫자로 표현할 수 없는 데이터는 **질적 자료**(Qualitative, Categorical data)\
질적 자료는 다시 명목척도와 서열척도, 양적 자료는 다시 구간척도와 비율척도로 구분된다.

|양적 자료|질적 자료|
|---|---|
|구간척도|명목척도|
|비율척도|서열척도|

* 명목척도: 단순 범주(이름)만 나타내는 척도\
숫자로 표현할 수는 있지만, 여기서는 분류의 기능만 할 뿐, 양적인 의미는 전혀 가지지 않는다. 그렇기에 사칙연산을 수행할 수 없다.\
ex.색깔 (빨 : 0, 주 : 1, 노 : 2, 초 : 3), 성별 (남 : 0, 여 : 1)

* 서열척도: 순위 또는 서열이 있는 척도\
범주의 순서를 비교할 수 있지만, 얼만큼 차이가 난다고는 해당 자료를 통해서 말할 수 없다. 그렇기에 이 역시 사칙연산을 수행할 수 없다.\
ex. 달리기 순위 (1등, 2등, 3등), 학년 (1학년, 2학년)

*구간(등간)척도: 연속적인 수로 수량화할 수 있고, 숫자간의 간격이 동일하기에 크고 작은 정도를 표현할 수 있다. 다만 절대적인 원점(0)은 존재하지 않는다.\
덧셈과 뺄셈은 가능하나, 곱셈과 나눗셈은 불가능하다.\
ex. 온도 (0도가 존재는 하나, 온도가 없음을 의미하지 않는다.), 연도

* 비율척도: 연속적인 수로 수량화할 수 있으며, 절대적인 원점이 존재한다.\
모든 사칙연산이 가능하다.
ex. 몸무게 (0kg은 실제로 무게가 없음), 인구 수, 길이

데이터의 척도에 따라 접근방식이 약간씩 달라지게 되니, 척도에 대해 먼저 생각해볼 필요가 있다.

## 2. (수치) 기술 통계

### 기술통계란?

**기술 통계학**
* 데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야로 평균, 중앙값, 표준편차 등의 개념이 이에 해당한다.

→ 측정이나 실험을 통해 수지한 자료를 정보로 전환하기 위하여 요약하고 정리하여 그 특성을 기술하는 절차에 관한 개념

* 수치 기술 통계: 하나의 수치, 즉 기술통계량으로 데이터를 요약 & 설명하는 방법 (중심 경향성, 산포도 등)
* 시각화 기술 통계: 표와 그래프의 형태로 데이터를 시각화하는 방법

### 중심 경향성
데이터의 중심적인 경향을 나타내는 기술 통계량

* 평균 (mean)\
데이터 값의 총합을 데이터의 개수로 나눈 값으로, 데이터의 중심 경향성을 나타내는 가장 대표적인 지표\
평균은 총합치를 사용하기에, 이상치(outlier)에 민감할 수 있다.

* 중앙값 (median)\
데이터 값들을 크기 순서대로 나열했을 때 가운데 위치한 값\
평균과 달리 이상치에 덜 민감하며, 데이터의 중심 경향성을 나타내는 **대안적인 지표**

* 최빈값 (mode)\
데이터에서 가장 자주 나타나는 값\
이산형 데이터에서 사용되며, 데이터의 분포를 이해하는데 도움을 준다.


### 산포도
데이터가 퍼져있는 정도를 설명하는 기술 통계량

* 분산 (variance)\
데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 중요한 통계적 측정 지표. 분산이 높으면 데이터가 많이 퍼져있다는 것을 의미한다.

* 표준편차 (standard deviation)\
분산은 관측값 단위의 제곱이 되어있기에, 대상 데이터와 일치하지 않는다. 그렇기에 여기에 분산에 제곱근을 씌워주어야 단위가 일치된다.

* 범위 (range)\
데이터의 최대값과 최소값의 차이를 나타내는 지표로, 데이터의 전체 변동 범위를 알려준다.

* 사분위수 (quartile)\
데이터의 표본을 4개의 동일한 크기로 나눈 값을 의미한다.\
Q1 = 하위 25퍼센트에 위치한 값, '하위 사분위수'\
Q2 = 하위 50퍼센트에 위치한 값, 중앙값\
Q3 = 상위 25퍼센트에 위치한 값, '상위 사분위수'

* 사분위간 범위 (IQR)\
Q3에서 Q1 값을 제외한 범위, 즉 데이터의 중앙 50%가 퍼져 있는 범위이다. 이상치를 탐지할 때 유용하다.

아래쪽 이상치: Q1 - 1.5 * IQR 보다 작은 값\
위쪽 이상치: Q3 + 1.5 * IQR 보다 큰 값

* 변동계수 (coefficient of variation)\
데이터의 절대적인 변동성이 아닌 상대적인 변동성을 나타내는 통계적 지표.\
표준편차를 해당 데이터의 평균으로 나눈 값으로 계산된다.

**변동계수 = (표준편차 / 평균) * 100**

표준편차는 단위가 존재하기에, 평균값이 다른 두 표본집단을 비교하거나 단위가 다른 두 데이터를 비교할 때 사용하기 적합하지 않다.\
이 때 표준편차를 평균으로 나누면 단위가 사라지기에, 집단의 퍼진 정도를 동일한 기준으로 비교하는게  가능해진다.

## 3. 추론 통계 - 추정

### 추론통계란?

추론통계학: 수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전체 정보를 추론하고 결론을 도출하는 방법을 다루는 분야. 통계적 가설검정과 신뢰구간 등이 이에 속한다.

→ 표본 집단을 통해 모집단에 대한 정보를 추측하는 과정

### 확률분포

확률분포는 확률변수가 특정한 값을 가질 확률을 나타내는 함수를 뜻한다.

확률변수는 랜덤한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 잇는 모든 경우의 수를 특정 값으로 대응시키는 함수이다. 예를 들어 동전을 던지는 실험에서 "앞면"과 "뒷면"은 각각 1과 0으로 대응시킬 수 있다.

확률변수의 속성에 따라 이산확률분포와 연속확률분포로 범주를 나눌 수 있다.

* 이산확률분포는 확률변수가 이산적인 값을 가지는 경우에 해당한다.\
이산적이라는 것은 유한하거나 셀 수 있다는 뜻이며, 특정 값을 선택할 수 있다.

* 연속확률분포는 확률변수가 연속적인 값을 가지는 경우에 해당한다.\
연속적인 값은 무한한 범위를 지녀, 특정 값을 선택하는 것이 불가능하다.

### 대표적인 분포 모형

**정규분포**

정규분포는 종(bell)모양을 띄며 평균을 기준으로 대칭된다. 표준편차가 클수록 곡선이 평평해지는데, 다시 말하면 데이터가 넓게 퍼질수록 곡선이 평평해지는 것이다.

가장 잘 알려진 확률분포이며, 많은 자연 현상과 사회 현상이 정규분포를 따르는 것으로 알려져 있다.

중심극한정리에 의해 모집단의 분포와 상관없이 표본의 크기가 커질수록 표본평균이 분포는 정규분포에 근사하게 된다. (여기서 표본의 크기가 크다는 것은 대게 n >= 30 을 의미한다.)

**대수의 법칙**
표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상을 뜻하며, 취합하는 표본의 수가 많을수록 통계적 정확도는 올라가게 된다.

**중심극한정리**
표본의 크기가 크면 표본 평균의 분포가 모집단의 분포와 상관없이 정규분포에 가까워진다.

**T-분포**

T-분포 역시 종 모양을 띄나, 정규분포보다는 꼬리가 두껍다. 자유도가 커짐에 따라 표준정규분포 N(0,1)에 수렴한다.

표본으로 모평균(혹은 모평균 간의 차이)을 추정할 때 사용되는 분포이며, 표본 크기가 작을 때 정규분포보다 더 정확한 결과를 제공한다.

자유도라는 매개변수를 사용한다.

**자유도?**
자유도는 통계에서 얼마나 많은 정보가 독립적으로 변할 수 있는가를 나타내는 숫자이다. 특정 상황에서는 일부 정보가 이미 결정되어 있어, 다른 정보들이 그만큼 덜 자유롭게 변할 수 있다.

예컨대 3개의 수의 평균이 10이라는 것을 인지하고, 앞 두 숫자가 5와 10이라고 해보자, 그러면 자동으로 마지막 숫자는 15인것이 결정된다. 여기서 자유도는 3개의 숫자에서 1을 뺀 값, 2가 된다. 이는 2개의 숫자가 자유롭게 변할 수 있다는 의미이다.

T분포의 자유도는 표본 크기에 비례한다. 자유도가 높을수록 T분포는 정규분포에 가까워지고, 통계적 분석의 정확도가 높아질 가능성이 있다. 그러나 너무 많은 자유도를 가지면 오버피팅과 같은 문제가 발생할 수 있기에, 적절한 자유도를 선택하는 것이 중요하다.

**오버피팅?**
모델이 학습 데이터에 너무 맞추어져서 새로운 데이터에 대한 일반화 성능이 저하되는 현상

---

표준정규분포는 평균이 0, 표준편차가 1인 정규분포로, 모집단의 평균과 분산을 정확히 알고 있을 때 사용됩니다. 이는 주어진 모집단에서 직접 계산되므로, 불확실성이 적다.

t-분포는 모집단의 분산을 모를 때, 표본의 분산을 사용해 모집단 분산을 추정하여 만들어지기 때문에 표본의 크기가 작을수록 이 추정은 불확실성이 커지고, 이 불확실성 때문에 t-분포는 표준정규분포보다 중심으로부터 더 퍼져 있습니다. 이는 극단적인 값이 나올 가능성을 반영한 것이다.

통상적으로 모집단의 분산이 알려져 있지 않은 경우, 표분분포를 결정할 때 표본의 크기가 30 미만이면 t분포를, 크기가 30 이상이면 표준정규분포를 이용합니다.

---

**카이제곱분포**
카이제곱분포는 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭 분포이다.\
자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 대칭적으로 변화해, 정규분포에 가까워진다.

주로 독립성 검정, 적합도 검정 등에 사용된다.

## 4. 추론통계 - 검정

가설 검정은 모집단의 실제 값이 얼마가 된다는 주장과 관련해, 표본의 정보를 사용해서 가설의 합당성 여부를 판정하는 과정을 의미한다.

### 가설 검정의 절차 
---
1. 가설 설정
2. 유의 수준 설정
3. 검정통계량 산출
4. 가설 기각/채택 판단
---
1. 가설 설정 

귀무가설 H0과 대립가설 H1을 설정한다.\
주로 귀무가설은 기각하고자 하는 명제로 설정한다. 대게 '~는 차이가 없다'라는 명제이다.\
대립가설은 내가 주장하고자 하는 명제이다. 보통 '~는 차이가 존재한다'라는 명제이다.

2. 유의 수준 설정

유의 수준이란 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치를 의미한다. 주로 0.05, 즉 5%로 설정하는데, 유의 수준은 연구자가 어떤 결정이 얼마나 신뢰할 만한지를 판단하는 기준이 된다.

유의 수준을 5%로 설정한다는 것은 '귀무가설이 실제로 참인데도, 5%의 확률로 잘못 기각할 수 있다'는 것을 의미한다.

예를 들어, 어떤 약물의 효과를 검증할 때, 이 약물이 실제로는 효과가 없더라도 5%의 확률로 약물이 효과가 있다고 잘못 결론 내릴 가능성을 수용하는 것이다.

**1종, 2종 오류**
* 1종 오류: 귀무가설이 실제로 참이었는데 기각할 확률
* 2종 오류: 귀무가설이 실제로 거짓인데 기각하지 못할 확률

가설검정은 1종 오류를 더 치명적인 것으로 보기에 유의수준은 1종 오류의 확률을 의미하게 된다.

3. 검정통계량 산출

검정통계량은 가설 검정의 결과를 판단하는 데 사용되는 값이다.\
데이터의 분포나 특징에 따라 T값, F값 등 어떤 통계량을 사용할지 정해야 한다.

**Z값**

**Z = (관측값 - 평균) / 표준편차**

일반적으로 30 이상의 큰 표본 크기에 대한 가설 검정에 주로 사용된다. 예컨대 모집단의 평균에 대한 가설을 검정할 때 사용한다.

Z값의 절댓값이 크다면 해당 데이터가 평균으로부터 멀리 떨어져있다는 것을 뜻한다.

Z = 1.96일 때, 유의수준 0.05 (=신뢰수준 95%)\
Z = 2.58일 때, 유의수준 0.01 (=신뢰수준 99%)을 기준으로 판단한다.

**T값**

**T = (표본평균 - 연구자가 설정하는 값)/표본 표준편차/표본 크기의 제곱근**

주로 모집단의 평균에 대한 가설을 검정하거나, 두 표본 평균 간의 차이를 검정할 때 사용된다.

T값의 절댓값이 크다면 해당 데이터가 모집단 평균으로부터 멀리 떨어져있으며, 관측된 표본 평균이 모집단 평균과 유의미한 차이가 있을 가능성이 높다는 것을 나타냅니다.

**F값**

**F = 분산_1/분산_2**

주로 분산 분석에서 두 개 이상의 그룹 간 평균의 차이를 검정하는 데 사용된다. F값은 두 표본의 분산의 비율이기에, 이 값이 1에 가까우면 두 표본의 분산이 서로 비슷하다는 것을 의미한다. 즉 두 그룹이 퍼진 정도가 비슷하다는 것이다.

#### [수치형 - 수치형] (피어슨)상관분석

두 변수간 피어슨 상관계수가 얼마나 유의미한지를 검정하는 방법이다. 귀무/대립 가설은 아래와 같이 세워진다.

> H0: 두 변수간 선형 상관관계가 존재하지 않는다.\
> H1: 두 변수간 선형 상관관계가 존재한다.

* 공분산

두 변수가 **함께 어떤 방향으로 변화하는지와 그 크기**를 표현하는데 사용된다.

- Cov (X,Y) > 0: X와 Y는 양의 관계, X가 커지면 Y도 커짐
- Cov (X,Y) < 0: X와 Y는 음의 관계, X가 커지면 Y는 작아짐
- Cov (X,Y) = 0: X와 Y는 양도 음도 아닌 관계
- 선형적인 관계는 없으나, 비선형적인 관계가 존재할 수 있음
ex. X가 Y의 제곱에 비례하는 경우

* 상관계수

두 변수 간의 선형적인 관계를 나타내는 통계적 지표

상관계수는 **-1과 1 사이의 값**을 가지며, 상관관계와 인과관계는 다르다는 점을 주의해야 합니다.

|상관관계|인과관계|
|---|---|
|두 변수가 수치적으로 연관되어있는 관계|한 변수가 다른 변수를 발생시키는 관계|
|와인 가격의 증가 & 교수 월급의 증가|마케팅양의 증가 → 매출 증가|

---
```python
import scipy.stats as spst
spst.pearsonr(df['col1'], df['col2'])
```
---

결과는 __(상관계수, p-value)__의 튜플 형태로 나오게 된다.

상관계수 r에 대하여 (절댓값을 씌운다면), 0.3 이하는 상관관계가 약하고, 0.3~0.5는 상관관계가 중간, 0.5보다 크다면 상관관계가 강하다고 판단내릴 수 있다.

#### [범주형 - 수치형] F-test(ANOVA), T-test

남/여, 물/불 등 범주가 2개라면?\
→ T-test 사용

> H0: 두 그룹의 평균에는 차이가 없다.\
> H1: 두 그룹의 평균에는 차이가 있다.

*두 그룹의 평균에 차이가 없음을 주장하고 싶다면*

*t-test 검정법 자체가 두 변수간 차이를 기준으로 ‘얼마나 멀리 떨어져 있는지’를 체크하기 때문에, ‘차이가 없음’이라는 가설을 주장할 수 없습니다.*

*간접적으로, H0에 차이가 없음, Ha에 차이가 있음을 넣고 ‘귀무가설을 기각하지 못하였다’는 식으로 주장할 수 있습니다.*

---
```python
mincho = df.loc[df['민초여부'] == 'Yes', '수명']
no_mincho = df.loc[df['민초여부'] == 'No', '수명' ]

spst.ttest_ind(mincho, no_mincho)
```
---

결과는 **TtestResult(statistic = t-값, pvalue = p값, df = 자유도)**의 형태로 나오게 된다.

t값은 보통 절댓값 2를 기준으로 그보다 크면 차이가 크다고 생각한다. p값은 역시 이 결과가 얼마나 믿음직한지를 나타내며 보통 0.05를 기준으로 한다.

---

범주가 3개 이상이라면 F-Test(ANOVA)를 진행하는 것이 효율적이다.

*참고로 ANOVA에도 다양한 종류가 있다*

F-통계량은, 위에서 언급된 것처럼 분모 그룹과 분자 그룹의 분산 비입니다.

ANOVA 검정을 할 때엔, F값은 **집단 간 분산 / 집단 내 분산**으로 계산합니다.
즉 (전체 평균과 각 집단의 평균을 가지고 만든 분산) / (각 집단의 평균과 개별 값으로 만든 분산)으로 계산한다.

> H1: 각 MBTI 그룹 별 수명 평균은 차이가 있다.
