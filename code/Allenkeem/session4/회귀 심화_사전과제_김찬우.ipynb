## 1.2 다중선형회귀의 수식적 이해

선형회귀식: 종속변수 y에 대한 설명변수의 **가중평균**\
설명변수에 가중치를 곱해 더함으로써 y를 얼마나 잘 설명해내는지를 나타낸다.

다만 선형회귀를 모든 상황에 사용할 수는 없다.

## 1.3 다중선형회귀의 기본가정

### 1.3.1 선형성

* 종속변수와 설명변수 간의 관계가 선형적이어야 한다.\
*선형적: 함수 f(x+y)에 대해 f(x) + f(y)이고 f(ax)에 대해 af(x)를 성립시킬 때*

### 1.3.2 독립성

* 각각의 설명 변수가 서로 선형독립적이어야 한다.
* 선형독립적이지 않은 경우 다중공선성이 존재한다고 한다.

### 1.3.3 오차항의 평균은 0이다 (0에 가까워야 한다)

**오차항: 실제 값과 예측 값 사이의 차이**
* 오차항(aka.백색 잡음)은 특정 패턴 없이 랜덤하게 변하는 신호를 의미한다.
* 평균적으로 어느 방향에 치우치지 않는 것이 중요하다.
* *ex: 오차항의 평균이 3이라면, 회귀 모델이 항상 실제 값보다 3만큼 부족하게 or 넘치게 예측하는 문제가 발생한다. → 모델이 데이터의 패턴을 올바르게 반영하고 있지 못하다는 뜻*

### 1.3.4 등분산성

* 오차항의 분산이 일정해야한다.
* if 오차항의 분산이 일정하지 않다면 → 이분산성이 존재 → 회귀 분석의 결과가 정확하지 않을 수 있다. ∵ 모델의 예측이 일부 구간에서는 매우 정확하지만, 일부 구간에서는 오차가 매우 커질 수 있기에\
*cf. 잔차의 도표화 혹은 검정을 통해 등분산성을 알아볼 수 있다 .*

### 1.3.5 오차항은 자기상관되어 있지 않다.
* 오차항의 공분산은 항상 0이어야 된다는 가정
* 오차항의 공분산이 0이 아니라면, 자기 상관이 있다고 한다.\
*cf.자기상관: 한 변수의 현재 값이 과거 값과 상관관계를 가지는 현상. 모델이 데이터를 무작위로 예측하는 것이 아니라, 이전 데이터의 패턴을 학습하여 반복적인 오류를 만들어 낼 가능성이 높다.*
*과적합이랑 비슷한건가...?*

### 1.3.6 정규성

* 오차항이 정규분포를 따른다는 가정
* 가정 6은 위배되어도 다중선형회귀분석의 결과에 큰 영향을 주지는 않는다.

이러한 가정들은 회귀 분석 후에 검증 가능하기도 하다.

|기본 가정|회귀 분석 전|회귀 분석 후|검증 방법|
|---|---|---|---|
|선형성|O|O|scatter plot|
|독립성|O|O|VIF 지수, 상관계수|
|오차항 평균이 0이다|X|O|np.mean(residuals)|
|등분산성|X|O|잔차의 도표화|
|오차항은 자기상관이지 않다|△|O|Durbin-Watson 검정(분석 후)|
|정규성|X|O|Shapiro-Wilk 검증, Q-Q plot|

## 1.4 회귀분석 평가방법
>회귀분석의 결과가 얼마나 유효한지 평가하기 위해서는 회귀선을 데이터와 함께 시각화하거나 통계지표를 이용하면 된다.

### 1.4.1 시각화
* 데이터 전반을 잘 요약하여 보여준다.
* 다만 2개 이상의 회귀선을 서로 비교할 때는 어렵다.

### 1.4.2 통계지표

**모델의 유의성 검정**

|검정 종류|검정 대상|목적|해석 기준|
|---|---|---|---|
|F-검정|전체 모델|회귀 모델이 유의미한지 검정|p < 0.05이면 회귀 모델이 유의미하다|
|T-검정|개별 변수|특정 독립 변수가 유의미한지 검정|p < 0.05이면 해당 변수가 유의미하다|

**모델의 성능 평가**

**R^2**는 결정 계수의 약어로, 모델이 설명하는 데이터의 총 변동 (평균과의 차이) 중에서 설명된 비율을 나타낸다\
→ 데이터의 평균을 예측값으로 사용할 때보다, 모델을 사용할 때 얼마나 더 정확한 예측인지

* 1에 가까울수록 모델이 데이터를 잘 설명한다고 할 수 있다.
* 만약 값이 0이라면, 모델이 데이터를 전혀 설명하지 못하다고 할 수 있다.
* 다만 독립 변수의 개수가 증가하면 증가할수록 자연스레 증가한다. 따라서 이를 보완하기 위해 나온 지표가 조정된 결정계수(Adjusted R^2)이다.

**AIC, BIC** 두 지표 모두 정보기준이라고 불리며, 값이 낮을수록 좋다고 평가함.

# 2. 비선형회귀

회귀는 결국 독립변수와 종속변수 간의 함수관계를 나타내는 것이기에, 데이터셋이 선형적이지 않다면 선형적이지 않은 식을 작성하면 된다.

## 2.1 다항식 회귀모델

독립변수에 **거듭제곱 항**이 추가되어 곡선적 특성을 모델링하는 것

|장점|단점|
|---|---|
|선형 모델에 비해 회귀선을 잘 fit하도록 그릴 수 있다.|너무 많은 feature 사용시 결과가 과적합될 수 있다. (이를 방지하기 위해 L2, L1 규제 사용 가능)|

적합한 상황:
* 관계가 직선적이지 않고 곡선을 그리는 데이터 패턴에서 유용
* 성장률, 감소율 등이 시간에 따라 변화하는 속도가 다른 경우

## 2.2 지수 회귀모델

종속 변수가 지수적으로 변화하는 관계를 모델링할 때 사용할 수 있다.

사용방법
1. 종속 변수 Y에 로그함수 적용
2. 로그 변환된 Y와 기존의 독립변수 X에 대해 선형 회귀 적용 → 모델이 ln Y와 X의 선형 관계를 학습
3. Y값을 예측 시, 모델이 예측한 Y값에 지수함수 적용하여 로그 변환되기 전 Y로 변형

적합한 상황:
* 값이 시간에 따라 지수적으로 증가하거나 감소하는 경우 적합
* 기술 발전, 투자 수익률 증가 등

## 2.3 로그 회귀모델

종속 변수와 독립 변수 간의 관계가 로그 함수를 통해 더 잘 표현될 때 사용할 수 있다.

사용방법
1. 독립 변수 X에 로그함수를 적용
2. 선형 회귀를 적용 → 이렇게 예측된 값은 Y이기에 예측값에 별도의 처리 없이 바로 사용 가능하다.

적합한 상황:
* 데이터가 초기에 빠르게 증가하고 점차 증가율이 줄어드는 패턴을 보일 때
* 인구 성장, 감염병의 확산과 같이 초기 빠른 증가 후 점진적 감소를 보이는 데이터에 적합

## 2.4 스플라인 회귀

데이터를 구간별로 나누고 각 구간에서 다른 선형 또는 비선형 함수를 적용하여 예측하는 방식

사용방법
* 전체 데이터 범위를 여러 구간으로 나누고, 각 구간에 대해 별도의 회귀 모델을 적용\
**구간 경계에서의 연속성을 유지하는 것이 중요**

적합한 상황:
* 데이터 패턴이 여러 구간에서 서로 다르게 나타날 때 유용
* 계절에 따라 판매량이 변화하는 소매 데이터 분석 등에서 적합
