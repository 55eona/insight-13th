# 통계

## 1. 기술통계
데이터의 주요 특성 요약 및 설명

### 1.1 중심 경향성
데이터의 **중심적인** 경향을 나타내는 기술 통계량
- 평균 series.mean(): 이상치에 민감
- 중앙값 series.median(): 이상치에 덜 민감
- 최빈값 series.mode(): 이산형 데이터의 분포 이해

**왜도 series.skew()**: 데이터 분포의 비대칭성을 나타내는 통계적 지표
  
### 1.2 산포도
데이터가 **퍼져있는 정도**를 설명하는 기술 통계량
- 분산: 데이터 값들이 평균으로부터 **얼마나 멀리 퍼져있는지**를 나타냄
- 표준편차: 분산의 제곱근, 관측값의 단위와 일치
- 범위: 데이터의 **최대값과 최소값의 차이**
- 사분위수: 데이터의 표본을 **네 개의 동일 크기**로 나눈 값
- 사분위간 범위: 데이터의 **중앙 50%** 가 퍼져있는 범위, 이상치 탐지 시 유용
- 변동계수: 상대적인 변동성, **평균값이 다른 두 표본집단**의 비교

## 2. 추론통계
표본 집단을 기반으로 모집단에 대한 정보를 추측

### 2.1 추론통계 - 추정
**확률분포**: 확률분포가 특정한 값을 가질 확률을 나타내는 함수

- 정규 분포: **평균**을 기준으로 대칭, **표준편차**가 클수록 평평
- T분포: **자유도**가 커짐에 따라 표준정규분포 N(0,1)에 수렴
- 카이제곱분포: **자유도**가 증가함에 따라 중심 오른쪽으로 이동, 모양 대칭적, 정규분포에 가까워짐


### 2.2 추론통계 - 검정
**가설 설정 - 유의 수준 설정 -검정통계량 산출 - 가설 기각/채택 판단**
1. **가설 설정**

   귀무가설: 기각하고자 하는 명제

   대립가설: 주장하고자 하는 명제
   
3. **유의 수준 설정**
   
   유의수준: 연구자가 허용할 수 있는 오류의 최대치를 설정

   - 1종 오류: 참인데 기각할 확률 (유의수준)
   - 2종 오류: 거짓인데 기각 못 할 확률 
     
     
   
4. **검정통계량 산출**
   
   가설 검정의 결과 판단에 사용되는 값

   - Z-score(표준 정규 분포): **큰 표본**에 대해 가설 검정, 절댓값이 클수록 평균으로부터 떨어진 값.
   - T-score(T분포): **작은 표본**에서 **모집단의 평균** 검정, 자유도에 따른 모양 변화, 절댓값이 클수록 모편균으로부터 떨어진 값.
   - F-score(F분포): **그룹 간 평균의 차이** 검정, 1과 가까울수록 두 표본의 분산이 비슷하다. 
     
     
5. **가설 기각/채택 판단**

   유의확률: 유의 확률(p)이 유의 수준(α)보다 작을 시에 귀무가설을 기각

#### 데이터 자료형에 따른 검정 방법
**1. [수치형 - 수치형] (피어슨)상관분석**

- 귀무가설: 두 변수간 선형 상관관계가 존재하지 않는다.
- 대립가설: 두 변수간 선형 상관관계가 존재한다.

공분산: 두 변수의 변화 방향
- Cov (X,Y) > 0이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
- Cov (X,Y) < 0이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
- Cov (X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계

상관계수: 두 변수 간의 선형관계
- Corr (X,Y)이 양수면 양의 상관관계 (1이면 완전한 양의 상관관계)
- Corr (X,Y)이 음수면 음의 상관관계 (-1이면 완전한 음의 상관관계)


**2. [범주형 - 수치형] F-test(ANOVA), T-test**

**T-test**: 범주 2개
- 귀무가설: 두 그룹의 평균에는 차이가 없다.
- 대립가설: 두 그룹의 평균에는 차이가 있다.
 
**F-test(ANOVA)**: 범주가 3개 이상일 때

**3. [범주형 - 범주형] Chi2-test**
$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$
값이 클수록 기대값과 관측값 차이가 크다. 두 변수가 독립적이지 않다.
